## Machines, Programmation, Programmes, et le Programmeur

Les machines, en particulier les appareils numériques (ordinateurs, smartphones, robots), jouent un rôle primordial dans la société moderne. En essence, ces machines sont un assemblage de composants électroniques qui ne peuvent accomplir une tâche sans l’aide de ce que l’on appelle un **programme**. Un ensemble de programmes travaille en concert pour former la machine que l’on désigne sous le nom "d’ordinateur".

En gros, 'un programme demande à un ordinateur de faire quelque chose'. La création d’un programme nécessite une personne très spéciale que l’on appelle le **programmeur** ou le **codeur** (informellement). Cette personne effectue une activité appelée **programmation** ou **codage** (informellement).
En résumé, la programmation est le processus de création (écriture) d’une séquence d’instructions qui sera transformée (compilée) en un programme exécutable par l’ordinateur.

Vous avez tous et toutes l’ambition de devenir de grands programmeurs. Pour y arriver, les programmeurs doivent pratiquer la programmation ! Pour programmer, nous utilisons des langages de programmation.
Les langages de programmation se distinguent de plusieurs façons. Parmi ces différences, il y en a une que nous devons comprendre à ce stade de notre parcours vers notre objectif : devenir de grands programmeurs.
Il existe une bifurcation importante qui définit deux types principaux de langages de programmation : low-level (bas niveau) et high-level (haut niveau).

### Low-level
Les langages bas niveau, autrement dit les langages au niveau machine, sont spécifiques à une famille de processeurs. Un langage conçu pour une famille de processeurs ne fonctionnera pas sur un ordinateur doté d’un type de processeur différent.
Ces langages sont conçus pour contrôler un assemblage de composants électroniques, et non pour être facilement manipulables par le cerveau humain. Cependant, ils sont indispensables pour faire fonctionner un ordinateur. Pour un programmeur capable de penser plus comme une machine que comme un être humain, ces langages peuvent être utilisés afin d’obtenir un maximum de performance.

### High-level
Les langages de haut niveau sont des langages informatiques dont les mots-clés proviennent d’un langage humain (malheureusement pour vous, ce langage est généralement l’anglais !) et qui utilisent des concepts pour le contrôle de la machine plus compatibles avec le fonctionnement de vos cerveaux. Cela rend la création de programmes plus facile et agréable, surtout si vous êtes une personne “normale”.
Vous avez peut-être entendu parler de plusieurs de ces langages : C, Ruby, Java, et bien sûr Python. En réalité, ces langages doivent être convertis en code machine, essentiellement des programmes bas niveau avant exécution, car les ordinateurs ne comprennent que les instructions de bas niveau.
Les langages de haut niveau fonctionnent sur presque tous les types d’ordinateurs et sur de nombreux appareils numériques, comme les smartphones et les systèmes Arduino.
Bifurcation 2 : les langages de haut niveau peuvent être compilés ou interprétés. Nous l’expliquerons ci-dessous.

#### Langages compilés
Ce sont des langages qui nécessitent un logiciel appelé compilateur. Je sais, c’est un peu étrange d’avoir besoin d’un logiciel pour créer un autre logiciel, mais c’est ainsi.
Le compilateur transforme nos instructions joliment rédigées en langage presque humain en un programme que l’ordinateur peut exécuter.
Des exemples de langages compilés sont : C, C++, Java* (Java est un cas un peu particulier, mais nous attendrons le cours de Java pour entrer dans les détails).
Voici le schéma illustrant le fonctionnement d’un compilateur :

>>> fichier de code → compilateur → fichier d’execution → resulat du programme

#### Langages interpretés
Ce sont des langages qui utilisent un interpréteur, encore un logiciel, pour exécuter le programme. Un interpréteur et un compilateur accomplissent à peu près la même tâche : transformer nos instructions en langage machine.
La différence est que l’interpréteur lit les instructions une par une, exécutant immédiatement leur équivalent en langage machine, alors que le compilateur lit toutes les instructions du fichier programme, puis produit un programme en langage machine globalement optimisé.
Les langages compilés sont généralement plus performants que les langages interprétés, mais il est plus facile de développer et de tester des programmes avec un interpréteur.
Python est un exemple de langage interprété. Dans notre cas, nous utiliserons Jupyter Notebook comme interpréteur (en réalité, l’interpréteur est IPython, intégré à Jupyter Notebook).
Voici le schéma illustrant le fonctionnement d’un interpréteur :

>>> fichier de code → interpreteur → resulat du program

## Programmation et les bugs (les erreurs de programmation)

In every life, a little rain must fall (Dans toute vie, il faut qu’il tombe un peu de pluie).
Dans la programmation, les bugs sont comme la pluie, mais en vérité, ce n’est pas **un peu**, c’est une inondation à la malienne !
Tu ne feras jamais de programmation sans te heurter à un déluge de bugs.
Alors, qu’est-ce que sont ces fameux bugs dont on parle ?
Les bugs sont les erreurs que tu rencontreras lorsque tu crées tes programmes, lorsque tu les testes, ou pire encore, lorsque tes clients utilisent le logiciel que tu leur as vendu !
Il vaut mieux trouver les bugs tôt plutôt que tard ! Tu ne seras jamais un programmeur efficace sans avoir appris à comprendre les bugs et à les corriger, comme on tue des cafards !

En programmation, il existe trois types d’erreurs que tu vas rencontrer :

### Erreur de syntaxe (Syntax Error)

Les langages de programmation ont une structure et des règles bien définies. Python ne fait pas exception : si tu ne respectes pas les règles de Python, il ne te traitera pas gentiment !

```python
# Exemple16
print(2+4)
```

Ceci est notre premier exemple d’erreur liée à la syntaxe.
Les parenthèses sont comme maman et les portes : si tu ouvres la porte, il faut la refermer après. En Python, lorsque tu ouvres un bloc de parenthèses avec (, il faut ensuite le fermer avec ).

Les erreurs de syntaxe sont signalées immédiatement lorsque l’interpréteur tente d’exécuter le programme.

Une compréhension de base de l’anglais est très utile pour comprendre ces erreurs. Si l’anglais n’est pas encore ton point fort, essaie quand même de lire le message. Tu verras les mêmes messages encore et encore, et avec le temps, tu comprendras ce que chaque message signifie.
Après tout, n’avons-nous pas dit au début que l’anglais est incontournable pour le travail technique ?

Le format d’un message d’erreur de syntaxe est le suivant :

- Une ligne avec le nom du fichier et le numéro de la ligne où l’erreur s’est produite.

- Une copie de la ligne où l’erreur s’est produite, avec un **^** qui indique le point exact où l’erreur a été détectée (l’erreur peut être à cet endroit ou avant).

- Une ligne qui donne le type d’erreur, suivi de deux-points (**:**), puis d’une description de l’erreur.

### Erreur d'exécution (Runtime Error)

```python
# Exemple17
print(25/0)
```

Avec une erreur d’exécution, tu n’as pas enfreint les règles de syntax de Python, mais tu as tout de même commis une bavure que Python, une fois en train d’exécuter ton programme, ne peut pas supporter. Le programme s’arrêtera lorsqu’il atteindra la ligne fautive.
Entre amis, on peut parfois laisser passer une bavure, mais Python ne te pardonnera pas ! Ces erreurs sont également appelées exceptions.
Il existe aussi des exceptions définies par le programmeur, permettant d’arrêter le programme lorsqu’un événement particulier se produit, mais ce type d’exception n’est pas à notre problème pour l’instant.

Dans notre exemple, bien que la syntaxe soit correcte, une erreur est signalée. C’est parce que nous avons divisé un nombre par zéro, et ni toi, ni Python, ni le meilleur mathématicien du monde ne sait ce que cela donne. Python te signale simplement que ce que tu lui as demandé est impossible.

Le format du message de cette erreur est le même que pour une erreur de syntaxe.

Une liste complète des exceptions de Python se trouve ici :

https://docs.python.org/fr/3/library/exceptions.html#bltin-exceptions

Nous traiterons les exceptions plus tard dans le cours, jusqu’à ce que tu en aies marre.

### Erreur sémantique

```python
# Exemple18
print(12+4*2/3)
```

Ce sont les erreurs les plus difficiles à repérer, car elles ne sont pas vraiment des erreurs.
Tu as bien demandé à Python de faire quelque chose, et Python l’a fait fidèlement, mais le résultat n’est pas celui que tu attendais.
Pourquoi ? Parce que tu n’as pas donné les instructions correctes à Python.

Par exemple, dans Exemple 18, imaginons que nous voulons obtenir 10,66 comme résultat, mais notre code renvoie 14,67. C’est correct du point de vue syntaxique et exécutable, mais ce n’est pas ce que nous attendions. Peux-tu corriger cette erreur de sémantique ?

Ici, contrairement aux erreurs précédentes, l’erreur est facile à repérer.
Mais imagine que cette erreur sémantique, liée à un mauvais calcul, soit enfouie dans des centaines de lignes de code : on ne remarque le résultat erroné qu’à la fin, sans aucune idée de l’origine du problème.
Il faudra alors vérifier ligne par ligne pour s’assurer que le programme fonctionne correctement.
