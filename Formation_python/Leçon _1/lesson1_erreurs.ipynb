## 1. Question: Ecris une phrase qui contient des lettres, au moins un chiffre et une ponctuation

### Exemple d'erreur

```python
# Erreur1
print("ahd5245'")
print("rkbaws 11 !")
print("Je suis né en 2005")
```

Pour cette question, beaucoup d'entre vous ont oublié de lire attentivement la question. La question vous demandait d'écrire une phrase qui comporte au moins un chiffre et une ponctuation.

Comme tu peux le voir dans l'exemple ci-dessus (soumis par certains étudiants), ce ne sont pas des phrases correctes.

### Solution

```python
# Solution1
print("Je peux faire un kilomètre de jogging en 5 minutes.")
```

Cette phrase n'est pas seulement correcte sur le plan sémantique, nous avons également inclus un **5** comme nombre et un point **.** comme ponctuation.

## 2. Question: Fait ce qu'il faut faire afin de réussir la vérification. (Les codes avant et après la virgule doivent avoir les valeurs égales.)

### Exemple d'erreur

```python
# Bug8
print("Hello Mali, Hello World")
```

```python
# Bug9
print("je programme", "je Programme")
```

Ici, nous vous avons demandé de mettre le contenu entre parenthèses à égalité, c'est-à-dire de changer le contenu avant la virgule ou après la virgule jusqu'à ce qu'ils soient à égalité.

### Solution

On vous a demandé de faire fonctionner cette cellule en premier.

```python
# Vérification
verificateur = lambda x, y: 'Bien' if x == y else 'Oops, manqué! Reessayer.'
```

```python
# Bug8
verificateur("Hello Mali", "Hello Mali")
```

C'est la bonne réponse. Nous avons modifié la valeur après la virgule et l'avons rendue égale à la valeur avant la virgule. Notre code s'exécuterait maintenant correctement. 

```python
# Bug9
verificateur("je programme", "je programme")
```

De même, nous avons changé les valeurs au point qu'elles sont égales.
