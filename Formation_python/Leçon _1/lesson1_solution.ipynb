#### Langages compilés
Ce sont des langages qui nécessitent un logiciel appelé compilateur. Je sais, c’est un peu étrange d’avoir besoin d’un logiciel pour créer un autre logiciel, mais c’est ainsi.
Le compilateur transforme nos instructions joliment rédigées en langage presque humain en un programme que l’ordinateur peut exécuter.
Des exemples de langages compilés sont : C, C++, Java* (Java est un cas un peu particulier, mais nous attendrons le cours de Java pour entrer dans les détails).
Voici le schéma illustrant le fonctionnement d’un compilateur :

>>> fichier de code → compilateur → fichier d’execution → resulat du programme

#### Langages interpretés
Ce sont des langages qui utilisent un interpréteur, encore un logiciel, pour exécuter le programme. Un interpréteur et un compilateur accomplissent à peu près la même tâche : transformer nos instructions en langage machine.
La différence est que l’interpréteur lit les instructions une par une, exécutant immédiatement leur équivalent en langage machine, alors que le compilateur lit toutes les instructions du fichier programme, puis produit un programme en langage machine globalement optimisé.
Les langages compilés sont généralement plus performants que les langages interprétés, mais il est plus facile de développer et de tester des programmes avec un interpréteur.
Python est un exemple de langage interprété. Dans notre cas, nous utiliserons Jupyter Notebook comme interpréteur (en réalité, l’interpréteur est IPython, intégré à Jupyter Notebook).
Voici le schéma illustrant le fonctionnement d’un interpréteur :

>>> fichier de code → interpreteur → resulat du program

## Programmation et les Bugs (les erreurs de programmation, les bouges)

In every life, a little rain must fall (Dans toute vie, il faut que tombe un peu de pluie). Dans la programmation, les bugs sont la pluie, mais, en vérité, c'est pas **un peu** mais une inondation à la malienne ! Tu ne feras jamais la programmation pas sans se heurter à un déluge de bugs. Alors, c'est quoi, ces fameux bugs dont on parle ? Les bugs sont les erreurs que tu rencontreras quand tu es train de créer tes programmes et quand tu les teste ou, pire encore, quand tes clients utilisent le logiciel que tu les as vendu ! Mieux trouver les bugs tôt plutôt tard ! Tu ne seras jamais un programmeur sans ayant appris à comprendre les bugs et les tuer comme les cafards !

En programmation, il y a trois types d'erreurs que tu vas rencontrer :

### Erreur de syntaxe (Syntax Error)

Les langages de programmation ont une structure et des règles qui sont bien définis. Et Python fait pas exception, quand tu ne respectes pas les règles de Python, Python ne te traitera pas gentiment !

```python
# Exemple16
print(2+4
```

Ceci est notre premier exemple d'erreur lié à la syntaxe. Les parenthèses sont comme maman et les portes - si tu ouvres la porte, il faut la fermer après. En Python, quand tu ouvres un bloc de parenthèse avec (, il faut plus tard le fermer avec ).

Les erreurs de syntaxe sont signalées immédiatement quand l'interpréteur essaye d'exécuter le programme.

Une compréhension de base de l'anglais est plus qu'utile pour comprendre ces erreurs. Si l'anglais n'est pas encore ton point fort, il faut, quand même, essayer de lire le message. Tu vas voir les mêmes messages encore et encore, et avec le temps, tu vas arriver à comprendre ce que le message veut dire. Ou non, nous n'avons pas dit au début que l'anglais est incontournable pour le travail technique ? 

Le format du message d'erreur de syntaxe est le suivant :

- Un ligne avec le nom du fichier, et le nombre de la ligne où l'erreur s'est produit.

- Un copie de la ligne où l'erreur s'est produit, avec **^** qui indique le point exact où l'erreur a été détecté (l'erreur peut être là ou avant cette position)

- Un ligne qui donne le type d'erreur, suit par un deux points(**:**) et une description de l'erreur.

### Erreur d'exécution (Runtime Error)

```python
# Exemple17
print(25/0)
```

Avec une erreur d'exécution, tu n'as pas enfreint les règles de syntaxe de Python, mais, quand même, tu as commis une bavure que Python, une fois en cours de l'exécution de ton programme, ne peut pas supporter. Le programme arrêtera quand il arrive à la ligne fautive. Entre amis, peut-être, on peut laisser passer une bavure, mais Python ne te pardonnera pas ! Ces erreurs sont aussi appelés **exception**. Il y a aussi les exceptions que le programmeur peut définir afin d'arrêter le programme quand quelque chose bizarre arrive, mais ce type d'exception n'est pas nos oignons pour l'instant. 

Dans notre exemple, bien que la syntaxe soit correcte, on a signalé une erreur. C'est parce que on a divisé un numéro par zéro et ni toi, ni Python, ni le meilleur mathématicien du monde sait ce que cela donne. Python te dit simplement que ce que tu l'as demandé de faire est idiot.

Le format de message de cette erreur est le même qu'une erreur de syntaxe.

Une liste complète des exceptions de Python se trouve ici :

https://docs.python.org/fr/3/library/exceptions.html#bltin-exceptions

Nous allons traiter les exceptions plus tard dans le cours jusque tu les en auras marre. 

### Erreur de semantique

C'est les erreurs qui sont les plus difficiles à répérer parce qu'ils ne sont pas en réalité les erreurs. Parce que tu n'as pas donné les instructions qu'il faut à Python. 

Par exemple, dans Exemple18, disons que nous voulons 10.66 comme résultat, mais avec notre code nous avons 14.67. C'est correcte, mais pas ce que nous avons attendu. Peux-tu corriger cette erreur de sémantique ?

Ici, contrairement à ce que nous avons dit, l'erreur est facile à repérer. Mais il faut imaginer que l'erreur sémantique avec le mauvais calcul est enfouies dans des centaines de lignes de code et on remarque le résultat erroné à la fin, sans aucune idée de l'origine du problème. On sera obligé de vérifier ligne par ligne la bonne exécution de programme. Ca, c'est l'enfer de programmeurs !
Ce sont les erreurs les plus difficiles à repérer, car elles ne sont pas vraiment des erreurs.
Tu as bien demandé à Python de faire quelque chose, et Python l’a fait fidèlement, mais le résultat n’est pas celui que tu attendais.

Par exemple, dans Exemple 18, imaginons que nous voulons obtenir 10,66 comme résultat, mais notre code renvoie 14,67. C’est correct du point de vue syntaxique et exécutable, mais ce n’est pas ce que nous attendions. Peux-tu corriger cette erreur de sémantique ?

Ici, contrairement aux erreurs précédentes, l’erreur est facile à repérer.
