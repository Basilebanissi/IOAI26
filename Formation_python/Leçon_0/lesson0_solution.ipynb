# Guide de Jupyter Notebook

# Introduction
La plateforme Jupyter Notebook est un espace de travail pour la programmation en Python. On travaille avec des fichiers notebook. Dans Jupyter Notebook il y a deux interfaces principales avec lesquelles nous allons interagir, elles sont le Tableau de bord et le Modificateur de Notebook

**L'interface de Jupyter Notebook est en anglais. Il faut apprendre à te débrouiller avec l'anglais, pour maîtriser le langage Python et même maîtriser l'interface que nous utiliserons pour apprendre le langage Python. Courage !**

## Comment Naviguer le Tableau de bord
Après le lancement de Jupyter Notebook depuis 'Ananconda Navigator', Notre navigateur (chrome préféré) s'ouvrira sur la page que l'on appelle le Tableau de bord. Cette page nous donne une liste comme celle-ci(l’image en-dessous). Cette liste est la liste de nos dossiers présents sur notre ordinateur. De là-bas on doit naviguer à nos fichiers notebook(qui ont la forme: nom_du_ficher**.ipynb**)



En total on a quatre sections dans cette page pour la navigation de la plateforme:

- En haut de la page sur la droite il y a deux boutons 'Quit' (Quitter) et 'Logout' (Se déconnecter), on utilise ces boutons pour respectivement quitter la session de Jupyter Notebook et se déconnecter 

- Les options en-dessous sur la gauche sont: 'Files' (onglet) qui est présélectionné, 'Running' (onglet), et 'Clusters' (onglet).
    - L'onglet 'Files' liste les contenus des dossiers dans le système, c'est dans cette option qu'on navigue à nos fichiers notebook pour les leçons
    - L'onglet 'Running' liste les fichiers notebook en cours d'exécution
    - L'onglet 'Clusters' ne nous concerne pas trop à ce niveau

- Sur la droite un peu en-dessous on a trois boutons pour: 'Upload' (*ajouter*), 'New' (Nouveau), 'Refresh' (Actualiser)
    - 'Upload' nous permet d’ajouter des fichier directement à partir du disque de notre ordinateur
    - 'New' nous permet de créer un nouveau fichier notebook
    - 'Refresh' nous permet d'actualiser la liste des fichiers, pour voir s'il y a des mises à jour
    
- Et finalement on a une section qui varie par onglet sélectionner::
    - Comme mentionné en-haut quand l’onglet ‘Files’ est sélectionné on a un tableau qui contient la liste des dossiers et documents
    - Quand l’onglet 'Running' est sélectionné on a deux tableaux: 
        - La première qui liste les ‘Terminals’ qui sont actifs (Les terminals ne nous concernent pas à ce niveau)
        - La deuxième qui liste les notebooks qui sont actifs

- **Note que tu dois cliquer sur le 'nom' du fichier que tu veux ouvrir pour y accéder
**-
