### 1.1 Création et manipulation de tableaux

Rappelons les bases de NumPy et explorons plus en profondeur certaines fonctions essentielles, en détaillant leur fonctionnement, leurs cas d’usage et quelques subtilités importantes.

**Nouvelles fonctions importantes :**

- `np.linspace(debut, fin, n)` : crée un tableau de `n` valeurs **également espacées** entre `debut` et `fin`
- `np.zeros(shape)`  
  Crée un tableau rempli de zéros avec une forme donnée (`shape`)  
  `shape` peut être :
  - un entier → tableau 1D
  - un tuple → tableau multidimensionnel (ex: `(3, 4)`)

- `np.zeros_like(arr)` :   Crée un tableau rempli de zéros **ayant exactement la même forme et le même type** que `arr`.  
  Cela évite de devoir redéfinir manuellement la dimension ou le type

- `arr.reshape(shape)` :  Permet de changer la forme d’un tableau **sans modifier ses données**, uniquement leur organisation en mémoire.  

- `np.concatenate([arr1, arr2])` : concatène plusieurs tableaux en un seul 

- `np.sort(arr)` : Retourne une **copie triée** du tableau (ne modifie pas l’original).  


```python
# Exemple 1 — Création de tableaux avec diverses fonctions

# linspace : très utile pour tracer des graphiques
x = np.linspace(0, 10, 5)   # 5 points entre 0 et 10
print("np.linspace(0, 10, 5) :", x)

# zeros et ones
zeros = np.zeros((3, 4))
ones  = np.ones((2, 3))
print("\nnp.zeros((3, 4)) :\n", zeros)
print("\nnp.ones((2, 3)) :\n", ones)

# reshape — changer la forme d'un tableau
arr = np.arange(12)          # [0, 1, 2, ..., 11]
mat = arr.reshape((3, 4))    # matrice 3 lignes x 4 colonnes
print("\nnp.arange(12) :", arr)
print("reshape((3, 4)) :\n", mat)

# concatenate
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.concatenate([a, b])
print("\nnp.concatenate([a, b]) :", c)
```

### 1.2 Opérations mathématiques — Broadcasting

NumPy permet d'appliquer des opérations mathématiques directement sur des tableaux entiers sans boucle. C'est le **broadcasting** (diffusion). C'est ce qui rend NumPy extrêmement rapide comparé aux listes Python.

**Opérations disponibles :**

- `arr + valeur`, `arr - valeur`, `arr * valeur`, `arr / valeur` : opération sur chaque élément
- `arr1 + arr2` : addition élément par élément (Les tableaux doivent avoir les mêmes dimensions.)
- `np.sqrt(arr)` : racine carrée de chaque élément
- `np.exp(arr)` : exponentielle ($e^x$) de chaque élément
- `np.abs(arr)` : valeur absolue de chaque élément
- `np.clip(arr, min, max)` : borne les valeurs entre min et max
- `np.round(arr, n)` : arrondi à n décimales

```python
# Exemple 2 — Broadcasting et opérations mathématiques

notes = np.array([12, 8, 15, 10, 18, 7, 14, 11, 16, 9])
print("Notes originales    :", notes)
print("Notes + 2 points    :", notes + 2)
print("Notes * 5 (sur 100) :", notes * 5)
print("Notes au carré      :", notes ** 2)
print("Racine des notes    :", np.round(np.sqrt(notes), 3))
print("Notes clippées 10-20:", np.clip(notes, 10, 20))

# Opérations entre deux tableaux
poids  = np.array([0.4, 0.6])    # coefficients
scores = np.array([14.0, 12.0])  # [contrôle, examen final]
moyenne_ponderee = np.sum(scores * poids)
print("\nMoyenne pondérée :", moyenne_ponderee)
```

### 1.3 Statistiques avec NumPy

NumPy possède des fonctions statistiques essentielles pour analyser des données. Elles permettent de calculer rapidement des mesures comme la moyenne, la médiane, la variance, ou encore de travailler sur les sommes et corrélations.

| Fonction | Description |
|---|---|
| `np.mean(arr)` | Moyenne |
| `np.median(arr)` | Médiane |
| `np.std(arr)` | Écart-type |
| `np.var(arr)` | Variance |
| `np.min(arr)` / `np.max(arr)` | Minimum / Maximum |
| `np.percentile(arr, q)` | Percentile q |
| `np.sum(arr)` | Somme |
| `np.cumsum(arr)` | Somme cumulée |
| `np.argmin(arr)` / `np.argmax(arr)` | Index du min / max |
| `np.corrcoef(x, y)` | Matrice de corrélation de Pearson |

```python
# Exemple 3 — Statistiques descriptives

np.random.seed(42)
tailles = np.random.randint(155, 200, size=20)  # 20 tailles en cm

print("Tailles (cm) :", tailles)
print(f"\nMoyenne        : {np.mean(tailles):.2f} cm")
print(f"Médiane        : {np.median(tailles):.2f} cm")
print(f"Écart-type     : {np.std(tailles):.2f} cm")
print(f"Min / Max      : {np.min(tailles)} / {np.max(tailles)} cm")
print(f"Index du max   : {np.argmax(tailles)} → valeur {tailles[np.argmax(tailles)]}")
print(f"Q1 (25%)       : {np.percentile(tailles, 25):.2f} cm")
print(f"Q3 (75%)       : {np.percentile(tailles, 75):.2f} cm")
print(f"Somme cumulée (4 premiers) : {np.cumsum(tailles)[:4]}")
```

### 1.4 Indexation avancée et filtrage

Lorsque l’on travaille avec des tableaux NumPy, il ne suffit pas toujours d’accéder aux éléments un par un. NumPy permet d’utiliser des masques booléens, des conditions combinées et des opérations vectorisées pour sélectionner, catégoriser ou modifier des données de manière rapide et efficace.

**Masques booléens**

Un masque booléen est un tableau de la même taille que le tableau d’origine, constitué de *True* ou *False*. Il indique quels éléments répondre à une condition.
Par exemple, si on a un tableau de températures journalières, on peut créer un masque qui identifie les jours où la température dépasse 30°C.

Une fois le masque créé, il suffit de l’utiliser pour sélectionner les éléments correspondant à la condition. NumPy retourne alors un nouveau tableau contenant uniquement ces valeurs.

**Conditions combinées**

Il est possible de combiner plusieurs conditions pour filtrer les données de manière plus précise.
- L’opérateur & correspond à un *ET logique*,
- L’opérateur | correspond à un *OU logique*,
- L’opérateur ~ correspond à un *NON logique*.

Ces opérations sont appliquées élément par élément sur le tableau et permettent de construire des expressions complexes pour sélectionner exactement ce que l’on souhaite.

**Vectorisation avec np.where**

La fonction *np.where* permet de créer de nouveaux tableaux selon une condition : c’est une version vectorisée d’un if/else.

Chaque élément est évalué :
- si la condition est vraie → on prend la première valeur
- sinon → on prend la deuxième valeur

**Modification conditionnelle**

NumPy permet de modifier certains éléments d’un tableau directement, selon une condition, sans boucle.

```python
# Exemple 4 — Indexation avancée

np.random.seed(7)
salaires = np.random.randint(150000, 900000, size=15)  # en FCFA
print("Salaires :", salaires)

# Filtre simple
print("\nSalaires > 500 000 :", salaires[salaires > 500000])

# Filtre combiné
milieu = salaires[(salaires > 300000) & (salaires < 600000)]
print("Entre 300k et 600k :", milieu)

# np.where : version vectorisée du if/else
categories = np.where(salaires >= 500000, 'haut', 'bas')
print("Catégories         :", categories)

# Remplacement conditionnel
sal2 = salaires.copy()
sal2[sal2 > 700000] = 700000
print("Plafonné à 700k    :", sal2)
```

### 1.6 Génération de données aléatoires

NumPy offre plusieurs **distributions statistiques** pour générer des données réalistes.

| Fonction | Description |
|---|---|
| `np.random.normal(mu, sigma, size)` | Distribution normale (en cloche) |
| `np.random.uniform(low, high, size)` | Distribution uniforme |
| `np.random.randint(low, high, size)` | Entiers uniformes |
| `np.random.choice(arr, size, replace)` | Tirage dans un tableau |
| `np.random.seed(n)` | Graine pour reproductibilité |

| Fonction | Description |
|---|---|
| `np.random.normal(mu, sigma, size)` | Génère des valeurs suivant une **distribution normale (gaussienne)** centrée autour de `mu` (moyenne) avec une dispersion contrôlée par `sigma` (écart-type). |
| `np.random.uniform(low, high, size)` | Génère des valeurs réparties de manière **uniforme** entre `low` et `high` (toutes les valeurs ont la même probabilité). |
| `np.random.randint(low, high, size)` | Génère des **entiers aléatoires** entre `low` (inclus) et `high` (exclu). |
| `np.random.choice(arr, size, replace)` | Tire aléatoirement des éléments dans `arr`, avec ou sans remise (`replace=True` ou `False`). |
| `np.random.seed(n)` | Initialise le générateur aléatoire pour obtenir des résultats **reproductibles**. |

```python
# Exemple 6 — Distributions aléatoires

np.random.seed(10)

# Distribution normale : notes d'une classe
notes_A = np.clip(np.random.normal(loc=14, scale=2, size=30), 0, 20)
notes_B = np.clip(np.random.normal(loc=11, scale=3, size=30), 0, 20)

print(f"Classe A — Moy: {np.mean(notes_A):.2f}, Std: {np.std(notes_A):.2f}")
print(f"Classe B — Moy: {np.mean(notes_B):.2f}, Std: {np.std(notes_B):.2f}")

# Distribution uniforme
durees = np.random.uniform(low=1.5, high=4.0, size=20)
print(f"\nDurées — Min: {durees.min():.2f}h, Max: {durees.max():.2f}h, Moy: {durees.mean():.2f}h")

# Tirage sans remise
equipe = np.array(['Amadou', 'Fatoumata', 'Moussa', 'Awa', 'Ibrahim', 'Kadiatou', 'Seydou', 'Mariam'])
tirage = np.random.choice(equipe, size=3, replace=False)
print("\nTirage au sort (3 personnes) :", tirage)
```

## Partie 2 : Pandas — Manipulation de Données

**Pandas** est la bibliothèque de référence pour manipuler des données tabulaires. Elle gère des données hétérogènes (texte, nombres, dates) avec des étiquettes.

Les deux structures fondamentales :
- **`Series`** :  est un tableau 1D avec un index, comparable à une colonne unique où chaque valeur est étiquetée.
- **`DataFrame`** : est un tableau 2D avec lignes et colonnes, comparable à un tableur, et chaque colonne est une Series.

### 2.1 La Series Pandas

Dans les Series, on peut accéder aux données de deux façons principales :

- par *label (étiquette)* : avec le nom de l’index
- par *position* : avec l’indice numérique (comme une liste)

```python
# Exemple 7 — Création d'une Series Pandas

temperatures = pd.Series(
    [34, 36, 38, 35, 37, 40],
    index=['Lun', 'Mar', 'Mer', 'Jeu', 'Ven', 'Sam'],
    name='Température Bamako (°C)'
)
print(temperatures)
print("\nType :", type(temperatures))
print("\nAccès par label    : temperatures['Mer'] =", temperatures['Mer'])
print("Accès par position : temperatures[2]     =", temperatures [2])
print("\nSlice 'Mar':'Ven' :")
print(temperatures['Mar':'Ven'])
print("\nMoyenne :", temperatures.mean())
print("\ndescribe() :\n", temperatures.describe())
```

Dans une Series Pandas, chaque valeur est associée à un label (index). Si ces labels sont textuels comme **'Lun', 'Mar', 'Mer'**, on ne peut pas accéder aux valeurs directement avec un nombre, car Pandas interprète toujours **[]** comme un label et non comme une position. Pour accéder à un élément par sa position numérique, on utilise **.iloc (integer location)**, qui permet de récupérer la valeur en indiquant sa position dans la Series, en commençant à 0 pour le premier élément.

### 2.4 Ajout, modification et suppression de colonnes

Dans un DataFrame, on peut créer une colonne calculée à partir de colonnes existantes afin de générer de nouvelles valeurs ou indices utiles pour l’analyse.

On peut également ajouter une colonne imbriquée en utilisant *np.where* avec des conditions multiples, ce qui permet de catégoriser les données selon différents seuils ou critères.

Par ailleurs, il est possible de renommer des colonnes avec *rename()* pour améliorer la clarté et la lisibilité du DataFrame.

Enfin, on peut supprimer des colonnes avec *drop()* afin de simplifier le DataFrame et se concentrer uniquement sur les informations pertinentes.

```python
# Exemple 10 — Modifier le DataFrame

df2 = df.copy()
import numpy as np

# Colonne calculée
df2['note_sur_100'] = df2['note'] * 5

# Colonne catégorielle imbriquée
df2['mention'] = np.where(df2['note'] >= 16, 'Très bien',
                 np.where(df2['note'] >= 14, 'Bien',
                 np.where(df2['note'] >= 12, 'Assez bien',
                 np.where(df2['note'] >= 10, 'Passable', 'Insuffisant'))))

# Renommer
df2 = df2.rename(columns={'heures_etude': 'h_etude'})

# Supprimer
df2 = df2.drop(columns=['ville'])

print(df2)
```

### 2.5 Groupement et agrégation

Dans un DataFrame, la méthode groupby() permet de regrouper les données selon une ou plusieurs colonnes, comme ici par exemple la colonne ville.

On peut ensuite appliquer des fonctions d’agrégation sur les groupes :

- Calculer la moyenne, le minimum, le maximum ou le nombre d’éléments d’une colonne avec des fonctions comme *mean(), min(), max(), count() ou avec agg()*.
- On peut aussi sélectionner plusieurs colonnes et appliquer des agrégations sur chacune, par exemple calculer la moyenne des notes et des heures d’étude par ville.

Ces opérations permettent de résumer et analyser les données par groupes, ce qui est très utile pour comparer des sous-ensembles du DataFrame de manière synthétique.

```python
# Exemple 11 — groupby

print("=== Moyenne des notes par ville ===")
print(df.groupby('ville')['note'].mean())

print("\n=== Nombre d'étudiants par ville ===")
print(df.groupby('ville')['nom'].count())

print("\n=== Statistiques complètes par ville ===")
print(df.groupby('ville')['note'].agg(['mean', 'min', 'max', 'count']))

print("\n=== Plusieurs colonnes par ville ===")
print(df.groupby('ville')[['note', 'heures_etude']].mean().round(2))
```

### 2.6 Valeurs manquantes

Dans un DataFrame, certaines cellules peuvent être vides ou absentes, et Pandas les représente par **NaN (Not a Number).**
Ces valeurs manquantes peuvent poser problème pour les calculs ou les analyses statistiques, il est donc important de les détecter et de les traiter.

La méthode **isnull()** permet d’identifier les cellules NaN.
En combinant avec sum(), on peut connaître le nombre de valeurs manquantes par colonne

**dropna()** supprime toutes les lignes qui contiennent au moins un NaN.
Cela permet de travailler uniquement avec les données complètes, mais les lignes supprimées sont perdues.

**fillna()** permet de remplacer les NaN par une valeur spécifique ou une statistique calculée.
Par exemple, on peut remplacer les âges manquants par la moyenne de la colonne et les notes manquantes par la médiane de la colonne.
Cette méthode permet de conserver toutes les lignes tout en rendant le DataFrame complet pour les calculs et analyses.

```python
# Exemple 12 — Valeurs manquantes

data_inc = {
    'nom':  ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Fati'],
    'age':  [25, None, 30, 22, None, 28],
    'note': [14.5, 12.0, None, 16.0, 11.5, None]
}
df_inc = pd.DataFrame(data_inc)

print("DataFrame :\n", df_inc)
print("\nNaN par colonne :\n", df_inc.isnull().sum())
print("\n% NaN par colonne :\n", (df_inc.isnull().sum() / len(df_inc) * 100).round(1))

print("\nAprès dropna() :\n", df_inc.dropna())

df_rem = df_inc.copy()
df_rem['age']  = df_inc['age'].fillna(df_inc['age'].mean())
df_rem['note'] = df_inc['note'].fillna(df_inc['note'].median())
print("\nAprès fillna (moy/médiane) :\n", df_rem)
```

### 2.7 Tri, apply et opérations avancées

Dans un DataFrame, on peut trier les lignes avec sort_values() selon une ou plusieurs colonnes, en choisissant l’ordre croissant ou décroissant pour chaque colonne. Cela permet, par exemple, de voir rapidement les meilleurs élèves par ville.

La méthode value_counts() sert à compter le nombre d’occurrences de chaque valeur unique dans une colonne. Elle est utile pour connaître la répartition des étudiants par ville ou par mention.

Enfin, apply() permet d’appliquer une fonction personnalisée à chaque élément d’une colonne afin de créer de nouvelles informations, comme une colonne mention qui classe les notes en TB (Très bien), B (Bien), AB (Asez bien), P(Passable) ou INS(Insufissant).

```python
# Exemple 13 — Tri, apply, value_counts

# Trier sur plusieurs colonnes
df_trie = df.sort_values(by=['ville', 'note'], ascending=[True, False])
print("=== Trié par ville puis note décroissante ===")
print(df_trie[['nom', 'ville', 'note']])

# value_counts
print("\n=== Étudiants par ville ===")
print(df['ville'].value_counts())

# apply : appliquer une fonction personnalisée
def mention(note):
    if note >= 16: return 'TB'
    elif note >= 14: return 'B'
    elif note >= 12: return 'AB'
    elif note >= 10: return 'P'
    return 'INS'

df['mention'] = df['note'].apply(mention)
print("\n=== Mentions ===")
print(df[['nom', 'note', 'mention']])
print("\nRépartition :\n", df['mention'].value_counts())
```

Ce tableau résume les principales méthodes de matplotlib pour personnaliser un graphique.

- **Taille de la figure** : ajuste la dimension globale du graphique.

### 3.2 Histogrammes

En plus des methodes de matplotlib vu dans la lecon 9, nous aons egalement : 
- *plt.subplots()* : crée une figure et plusieurs axes (sous-graphes) pour placer plusieurs graphiques dans la même figure.

- *axes[i].hist()* : trace un histogramme sur un axe spécifique, en définissant le nombre de classes, la couleur, la transparence et les contours des barres.

- *axes[i].axvline()* : ajoute une ligne verticale à une position donnée, avec couleur, style, épaisseur et étiquette, utile pour marquer la moyenne ou la médiane.

- *axes[i].set_title()* : définit le titre d’un sous-graphique spécifique.

- *axes[i].set_xlabel()* : définit le nom de l’axe des X pour un sous-graphique spécifique.

- *axes[i].set_ylabel()* : définit le nom de l’axe des Y pour un sous-graphique spécifique.

- *axes[i].legend()* : affiche la légende sur le sous-graphique correspondant, à partir des labels définis pour chaque élément.

- *plt.tight_layout()* : ajuste automatiquement les espacements entre les sous-graphes pour éviter que les titres, axes ou légendes se chevauchent.

```python
# Exemple 15 — Histogrammes

np.random.seed(42)
notes_cls = np.clip(np.random.normal(13, 3, 80), 0, 20)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(notes_cls, bins=12, color='steelblue', edgecolor='black', alpha=0.7)
axes[0].axvline(np.mean(notes_cls),   color='red',    linestyle='--', linewidth=2,
                label=f'Moy = {np.mean(notes_cls):.1f}')
axes[0].axvline(np.median(notes_cls), color='orange', linestyle='-.', linewidth=2,
                label=f'Méd = {np.median(notes_cls):.1f}')
axes[0].set_title('Distribution des notes')
axes[0].set_xlabel('Note (/20)')
axes[0].set_ylabel('Fréquence')
axes[0].legend()

notes_A = np.clip(np.random.normal(14, 2, 50), 0, 20)
notes_B = np.clip(np.random.normal(11, 3, 50), 0, 20)
axes[1].hist(notes_A, bins=10, alpha=0.6, color='green', label='Classe A')
axes[1].hist(notes_B, bins=10, alpha=0.6, color='red',   label='Classe B')
axes[1].set_title('Comparaison de deux classes')
axes[1].set_xlabel('Note (/20)')
axes[1].legend()

plt.tight_layout()
plt.show()
```

### 3.5 Subplots et tableaux de bord

Un **dashboard** (tableaux de bord) est un outil interactif qui convertit des données brutes en informations utiles, facilitant la compréhension et la prise de décision rapide grâce à des graphiques et visualisations claires. Il permet de suivre les tendances, contrôler les performances et explorer les données, en regroupant plusieurs sources sur une même interface.

- **plt.subplots(nrows, ncols, figsize=(w,h))** : crée une figure avec plusieurs sous-graphes organisés en lignes et colonnes, et définit la taille de la figure.

- **fig.suptitle(texte, fontsize, fontweight)** : ajoute un titre principal pour toute la figure ou le dashboard.

- **ax.bar(x, y, ...)** : crée un histogramme ou un graphique en barres pour représenter des valeurs par catégorie.

- **ax.axhline(y)** : trace une ligne horizontale à une valeur donnée, utile pour marquer un seuil.

- **ax.fill_between(x, y, alpha, color)** : remplit la zone sous une courbe pour visualiser un cumul ou un intervalle.

- **ax.pie(sizes, labels, autopct, colors, startangle)** : crée un diagramme circulaire pour représenter la proportion des valeurs, avec pourcentages et couleurs.

```python
# Exemple 18 — Tableau de bord (dashboard)

np.random.seed(1)
mois = ['Jan','Fév','Mar','Avr','Mai','Jun','Jul','Aoû','Sep','Oct','Nov','Déc']
ventes    = np.random.randint(50, 200, 12)
depenses  = np.random.randint(30, 150, 12)
benefices = ventes - depenses
clients   = np.cumsum(np.random.randint(5, 30, 12))

fig, axes = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Tableau de Bord Annuel', fontsize=15, fontweight='bold')

x = np.arange(len(mois))
axes[0,0].bar(x - 0.2, ventes,   0.4, label='Ventes',   color='steelblue')
axes[0,0].bar(x + 0.2, depenses, 0.4, label='Dépenses', color='salmon')
axes[0,0].set_xticks(x)
axes[0,0].set_xticklabels(mois, rotation=45, fontsize=8)
axes[0,0].set_title('Ventes vs Dépenses')
axes[0,0].legend()

couleurs = ['green' if b >= 0 else 'red' for b in benefices]
axes[0,1].bar(mois, benefices, color=couleurs)
axes[0,1].axhline(0, color='black', linewidth=0.8)
axes[0,1].set_title('Bénéfices mensuels')
axes[0,1].tick_params(axis='x', rotation=45, labelsize=8)

axes[1,0].plot(mois, clients, marker='o', color='purple', linewidth=2)
axes[1,0].fill_between(range(12), clients, alpha=0.2, color='purple')
axes[1,0].set_title('Clients cumulés')
axes[1,0].tick_params(axis='x', rotation=45, labelsize=8)

axes[1,1].pie(ventes[-3:], labels=['Oct','Nov','Déc'],
              autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'],
              startangle=90)
axes[1,1].set_title('Répartition ventes Q4')

plt.tight_layout()
plt.show()
```
