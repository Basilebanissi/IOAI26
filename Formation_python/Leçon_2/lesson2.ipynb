## Changer la valeur d'une variable

**Tu peux généralement mettre à jour une variable en lui attribuant simplement une nouvelle valeur.**

```python
# Exemple18
compteur = 10
écrire(compteur)
compteur = 20
écrire(compteur)
```

Nous voyons ici que la valeur de la variable **compteur** est passée de 10 à 20.
C’est aussi simple que ça !

Dans la cellule ci-dessous, nous incrémentons la valeur de la variable compteur de 1.
L’expression à droite, **compteur + 1**, prend la valeur actuelle de compteur (20), ajoute 1 (20 + 1), puis attribue le résultat (21) à la même variable **compteur**.

```python
#Exemple19
compteur = compteur + 1
écrire(compteur)
```

L’incrémentation d’une variable est une opération si courante que Python, pour le plus grand plaisir des paresseux, propose un raccourci : **+=**.

```python
#Exemple20
compteur += 9 # cela est équivalent à compteur = compteur + 9
écrire(compteur)
```

Et parce que les paresseux sont si nombreux parmi les amateurs de Python, le même raccourci s’applique à tous les opérateurs mathématiques.

## Notes en marge: Les Commentaires

Tu les as peut-être déjà remarqués et saisis leur utilité.
Il est maintenant temps de parler des commentaires.

Les commentaires servent de zone d’explication libre pour le programmeur, afin de fournir aux autres et à soi-même les informations nécessaires pour comprendre les objectifs du programme et le fonctionnement du code.

Les bons commentaires sont le signe d’un programmeur professionnel.
La capacité de bien expliquer un programme via les commentaires est une qualité vivement recherchée par les employeurs.

Pourquoi écrit-on des commentaires ?

- Pour les autres programmeurs.
- Pour soi-même, lorsqu’on a oublié ce que l’on a fait il y a plusieurs mois.
- Pour documenter les objectifs et le fonctionnement du programme pour les utilisateurs.

En Python, les commentaires sont précédés par un ('#').
Ils peuvent être placés avant le code, après le code, ou sur la même ligne, directement après le code.

- Les commentaires en ligne sont généralement brefs et servent, par exemple, à expliquer l’utilisation d’une variable.
- Les commentaires précédant un bloc de code sont souvent plus détaillés et expliquent le but du code, l’algorithme, ou comment utiliser le code.

```python
# Exemple22
# Ce programme est une réalisation d'une petite calculatrice.
# Il demande que l'utilisateur entre deux numéros entiers et il
# écrit leur somme sur l'écran.

premier, deuxieme, resultat = 0, 0, 0  # initalisation des deux numéros d'entrée et le résultat

premier = int(input("Entrer le premier numero de l'addition: "))
deuxieme = int(input("Entrer le deuxieme numero de l'addition: "))
resultat = premier + deuxieme
# écrire(resultat)
print(resultat)
```

Le titre du programme (Exemple22) et son but sont expliqués en haut dans des commentaires multi-lignes.
L’initialisation des variables **premier, deuxieme et resultat** est indiquée.

Enfin, le code utilisant **écrire** à la place de print est désactivé.
Si le programmeur souhaite utiliser **écrire**, il peut retirer le symbole **#** devant la ligne de code correspondante et mettre la ligne suivante en commentaire.
