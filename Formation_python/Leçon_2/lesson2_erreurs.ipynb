<b>Au-dessous  On écrit la déscription du programme en haut, au-dessus du programme, pas après le programme. La fonction recrit ne dit pas forcement écrire, la ligne execute une fonction qui s'appelle "recrit". -1</b>

```python
# Problème9
def recrit(valeur): #Python doit inserer une valeur
    try: #il dit d'essayer
        assert valeur is not None #met une valeur non nulle
        print(valeur) #calculer la valeur
    except: #excepté
        print("Malin(e)!") #excecuter la valeur Malin(e)

recrit("Yacouba") #écrit Yacouba
recrit(None) #n'écrit rien
# on dit d'écrire valeur sans donner la valeur puis on ditb d'écrire Malin(e) etYacouba. 
```
