## Les Regles pour Formation des Noms de Variables en Python

### Les règles

#### Les règles requises pour la dénomination des variables

- **Les variables ne peuvent pas commencer par des chiffres, mais elles peuvent contenir des chiffres.**

```python
# Exemple10
1angle = 45 # Forme Incorrecte
```

La cellule de code ci-dessus donnera une erreur de syntaxe à l'exécution.

```python
# Exemple11
angle1 = 45 # Forme Correcte
a1ngle = 45 # Forme Correcte
```

- **les noms de variables ne doivent pas contenir d'espaces**

```python
# Exemple12
mon angle = 60 # Forme Incorrecte
```

La cellule de code ci-dessus donnera une erreur de syntaxe à l'exécution.

```python
# Example13
monangle = 60  # Forme Correcte
```

- **À l'exception du _**, des lettres et des chiffres, aucun autre caractère n'est accepté dans le nom des variables.

```python
# Example14
?diametre = 35 # Forme Incorrecte
mon!symbole = 90 # Forme Incorrecte
```

- **L'utilisation des mots-clés de Python est à éviter**

```python
# Exemple15
achat1 = 25
mon_compte = 50
return = mon_compte - achat1 # illegale
assert = mon_compte * 2 # illegale
```

## Changer la valeur d'une variable

**Tu peut généralement mettre à jour une variable en lui attribuant simplement une nouvelle valeur**

```python
# Exemple18
compteur = 10
écrire(compteur)
compteur = 20
écrire(compteur)
```

Nous voyons ici que la valeur de la variable **compteur** est passée de 10 à 20. C'est aussi simple que ça.

Dans la cellule au-dessous, nous incrementons la valeur de la variable compteur par 1. L'expression sur la droite, **compteur + 1** prend la valeur actuelle de compteur (20), ajoute 1 (20 + 1) et, puis, attribue le résultat (21) à la même variable **compteur**.

```python
#Exemple19
compteur = compteur + 1
écrire(compteur)
```

L'incrémentation d'une variable est une opération si commune que Python, pour la grande joie de paresseux, a une raccourci **+=**.

```python
#Exemple20
compteur += 9 # cela est équivalent à compteur = compteur + 9
écrire(compteur)
```

## Notes en marge: Les Commentaires

** Tu as déjà les remarqué et peut-être, tu as déjà saisi leur utilité. Il est temps de faire les commentaires sur les commentaires. Les commentaires servent comme un espace libre pour le programmeur d'expliquer aux autres et à lui-même les informations nécessaires pour comprendre les objectifs du programme et le fonctionnement du code. Les bons commentaires sont le signe d'un programmeur professionnel, la capacité de bien expliquer un programme pars les commentaires est une qualité vivement sollicité par les employeurs. Pourquoi on écrit les commentaires ?

- On les écrits pour d'autres programmeurs
- On les écrits pour nous-même (dans l'avenir, quand tu as oublié ce que tu as fait il y a 6 mois )
- On les écrits pour documenter les objectifs et le fonctionnement du programme pour les utilisateurs

En Python, les commentaires sont précédé par un ('#'). Ils peuvent être mis avant ou après le code ou mis sur la même ligne directement après le code. Les commentaires en ligne sont normalement brèves, par exemple, ils peuvent être utilisés pour expliquer l'utilisation d'une variable. Les commentaires qui précéde un bloc de code sont souvents les explications plus générel ou plus détailles comme le but du code, l'algorithme, ou comment de l'utiliser.

```python
# Exemple22
# Ce programme est une réalisation d'une petite calculatrice.
# Il demande que l'utilisateur entre deux numéros entiers et il
# écrit leur somme sur l'écran.

premier, deuxieme, resultat = 0, 0, 0  # initalisation des deux numéros d'entrée et le résultat

premier = int(input("Entrer le premier numero de l'addition: "))
deuxieme = int(input("Entrer le deuxieme numero de l'addition: "))
resultat = premier + deuxieme
# écrire(resultat)
print(resultat)
```

Le titre du programme (Exemple22) et son but sont expliqués en haut dans les commentaires multi-lignes. L'initialisation des variables **premier, deuxieme, resultat** est signalée. Finalement, le code utilisant **écrire** pour print est désactivé. Si le programmeur décide qu'il préfère d'utiliser **écrire** il peut enlever le symbole **#** devant le code et changer la ligne suivante en commentaire.
