En général, les calculs effectués par Python fonctionnent comme tu les as appris à l’école, avec une exception importante : la division.
Lorsqu’on effectue une division avec des entiers, Python ne renvoie pas un entier avec son reste, mais un nombre à virgule flottante, même lorsque la division est exacte et que le reste est zéro.

```python
# Exemple4
# Ici, le résultat sera 2.5 et de type float et pas 2  avec type int
print(10/4)
print(type(10/4))
# Ici, le résultat sera 2.0 et type float et pas 2 avec type int
print(10/5)
print(type(10/5))
```

Mais, en fait, Python est gentil et il peut donner un résultat entier et un reste si tu insistes.
Il existe des opérateurs mathématiques spéciaux pour ton bonheur.**//** et **\%**

### Quotient d'une division entière (//)

```python
# Exemple5
print(10//4)
print(type(10//4))
```

L’exemple divise 10 par 4, tronque la partie fractionnaire, ce qui donne toujours un entier.
Parfois, nous ne voulons pas nous occuper de ces nombres décimaux déroutants.

### Reste d’une division entière (%)

```python
# Exemple6
mod = 10 % 4
print(mod)
print(type(mod))
```

Ici, nous avons divisé 10 par 4 et obtenu le reste du calcul.
Cette opération renvoie également un résultat entier.

## Les nombres à virgule flottante (décimaux) — Floating Point (float)

Tu as déjà vu que les nombres à virgule flottante n’utilisent pas de virgule en Python, mais un point.
Ils représentent essentiellement les nombres décimaux.
(Pas exactement, mais pour l’instant, faisons comme si c’était le cas.)

```python
# Exemple11
var1 = 2.7
print(var1)
print(type(var1))
```

Voici un exemple de valeur à virgule flottante.
Nous pouvons également effectuer des opérations mathématiques avec elle.

Nos deux opérations ont donné un résultat décimal, ce qui du au fait que nous avons inclus un nombre décimal dans le calcul.

## Les chaînes de caractères ctring (<code>str</code>)

Une chaîne de caractères est une séquence de caractères adjacents.
Comme nous l’avons vu dans les leçons précédentes, nous obtenons une chaîne de caractères en plaçant la séquence de caractères entre guillemets.

```python
# Exemple15
mon_string = "Ceci est un String"
mon_string2 = "Ca c'est un string aussi"
```

En fait, les guillemets ne sont pas le seul moyen de créer une chaîne de caractères ; nous pouvons également utiliser l’apostrophe.

```python
# Exemple16
texte1 = 'Hello, World'
texte2 = "Hello, World"
print(texte1)
print(texte2)
```

Les variables texte1 et texte2 contiennent le même texte : le fait que l’une ait été créée avec " et l’autre avec ' ne change rien.

Le guillemet et l’apostrophe sont des caractères qui peuvent être utilisés dans les chaînes de caractères.
Mais ils servent également à délimiter une chaîne.
Alors, comment faire la différence entre un guillemet ou une apostrophe qui est un caractère à l’intérieur de la chaîne et un guillemet ou une apostrophe qui démarque le début ou la fin de la chaîne ?

```python
#Exemple17
amuser = 'Le Python utilise les guillemets anglais "Python est mon bonheur"'
print(amuser)
amuser = "Le Python utilise les apostrophes aussi 'Python est mon bonheur'"
print(amuser)
```

Une solution consiste à créer la chaîne avec des guillemets lorsque la chaîne contient des apostrophes, et à utiliser des apostrophes lorsque la chaîne contient des guillemets, comme l’exemple l’a montré.

```python
#Exemple18
amuser = 'Le Python utilise les guillemets anglais "Python, c'est mon bonheur"'
print(amuser)
```

Que faire lorsqu’une chaîne de caractères contient à la fois des guillemets et des apostrophes ?
Il existe une solution plus générale : placer un caractère d’échappement \\ directement devant le caractère pour indiquer à Python que ce caractère fait partie de la chaîne et n’est pas la marque de clôture.

## Booléen (bool)

Pour notre dernier type de base en Python, nous avons l’honneur de vous présenter le fameux booléen (bool).
En programmation, le booléen représente la valeur de vérité d’une expression en Python.
Il peut prendre l’une des deux valeurs possibles : True (Vrai) ou False (Faux).
Les valeurs booléennes sont binaires, ce qui signifie qu’il n’y a que deux valeurs possibles, un peu comme le bit informatique, qui peut être 0 ou 1.

```python
# Example24
affirm1 = True
print(type(affirm1))
print(affirm1)
affirm1 = False
print(type(affirm1))
print(affirm1)
```

En Python, toutes les valeurs ont une valeur de vérité.
Tout est considéré comme vrai, sauf : **False, None, 0, et les chaînes de caractères vides ('' et "")**, qui sont fausses par défaut.
Nous pouvons utiliser la fonction **bool()** pour vérifier la valeur de vérité d’une variable.

```python
# Example25
vérité1, vérité2, vérité3, vérité4, vérité5 = bool(None), bool(False), bool(0), bool(''), bool("")
print(vérité1, vérité2, vérité3, vérité4, vérité5)
vérité1, vérité2, vérité3, vérité4, vérité5, vérité6 = bool(-1), bool(2e-100), bool(True), bool('Thanos'), bool(" "), bool(23.65)
print(vérité1, vérité2, vérité3, vérité4, vérité5, vérité6)
```

### Opérateurs Booléens (not, and, or)

Pour les types int et float, nous utilisons les opérateurs mathématiques.
Pour le type str, nous avons les opérateurs de concaténation et de répétition.
Les booléens ont également leurs opérateurs : **not, and, or** (négation, et, ou).

- **not (négation)**: donne la négation de la valeur qui le suit.

```python
# Example26
a = True
b = not a
c = not b
d = not c
print(a, b, c, d)
e = not not not a
print(e)
```

- **and (et)** : donne la valeur de vérité de deux expressions booléennes.

Si les deux expressions sont vraies, leur valeur **and** est vraie .

Si une ou les deux expressions sont fausses, leur valeur **and** est fausse.

Voir l’exemple ci-dessous :

```python
# Example27
a, b = True, True
c, d = False, False
case1 = a and b
case2 = a and c
case3 = c and a
case4 = c and d
print(case1, case2, case3, case4)
a, b, c, d, e, f, g, h, i, j = True, True, True, True, True, True, True, True, True, False
case1 = a and b and c and d and e and f and g and h and i and j
case2 = a and b and c and d and e and f and g and h and i and not j
print(case1, case2)
```

- **or (ou)**: donne la valeur de vérité de deux expressions booléennes.

Si l’une, l’autre ou les deux expressions sont vraies, leur valeur **or**  est vrai.

Si les deux expressions sont fausses, leur valeur **or** est faux.

Voir l’exemple ci-dessous :

```python
# Example28
a, b = True, True
c, d = False, False
case1 = a or b
case2 = a or c
case3 = c or a
case4 = c or d
print(case1, case2, case3, case4)
a, b, c, d, e, f, g, h, i, j = True, False, False, False, False, False, False, False, False, False
case1 = a or b or c or d or e or f or g or h or i or j
case2 = not a or b or c or d or e or f or g or h or i or j
print(case1, case2)
```

Comme pour les maths, tu peux utiliser des parenthèses pour décider dans quel ordre Python va calculer une expression booléenne.
La partie la plus à l’intérieur des parenthèses est calculée en premier, ensuite les parenthèses plus externes, et enfin ce qui n’est pas dans les parenthèses.

Essaie de deviner la valeur de l’expression ci-dessous avant d’exécuter la cellule.

```python
# Example29
a, b = True, False
laVéritéSiJeMens = ((not (a and b)) or not b) and (not (not a or b) and a)
print(laVéritéSiJeMens)
```

3. Passe au niveau suivant : **(not (a and b)) et (not (not a or b) and a)**. Calcule leurs valeurs et remplace-les par True ou False.
4. Il reste ensuite la dernière couche de parenthèses : ((not (a and b)) or not b). Calcule-la et remplace-la par sa valeur.
5. Enfin, il ne reste que la partie de l’expression en dehors des parenthèses, avec l’opérateur and.
