## Les nombres en virgule flottante (décimaux) Floating Point (float)

Tu as déjà vu que les nombres en virgule flottante n'ont pas le virgule en Python, mais le point. Ils s'agissent, essentiellement, de la représentation des nombres décimaux. (Pas vraiment, mais pour l'instant, faisons comme si c'était le cas)

```python
# Exemple11
var1 = 2.7
print(var1)
print(type(var1))
```

## Booléen (bool)

Pour notre dernier type Python de base, nous avons l'honneur de vous présenter le fameux Booléen (bool). En programmation, le booléen est la valeur de vérité d'une expression en Python. Il peut avoir une de deux valeurs possibles : True (Vrai) ou False (Faux). Les valeurs booléennes sont binaires, cela veut dire qu'il n'y a que deux valeurs possible, comme le bit informatique avec ses valeurs de 0 ou de 1.

```python
# Example24
affirm1 = True
print(type(affirm1))
print(affirm1)
affirm1 = False
print(type(affirm1))
print(affirm1)
```

### bool

En Python, toutes les valeurs ont une valeur de vérité. Tout est vrai sauf **False, None, 0, et les chaînes de caractères vides '' et ""** qui sont faux par défaut. Nous pouvons utiliser **bool** pour vérifier la valeur de vérité d'une variable.

```python
# Example25
vérité1, vérité2, vérité3, vérité4, vérité5 = bool(None), bool(False), bool(0), bool(''), bool("")
print(vérité1, vérité2, vérité3, vérité4, vérité5)
vérité1, vérité2, vérité3, vérité4, vérité5, vérité6 = bool(-1), bool(2e-100), bool(True), bool('Thanos'), bool(" "), bool(23.65)
print(vérité1, vérité2, vérité3, vérité4, vérité5, vérité6)
```

### Opérateurs Booléens (not, and, or)

Pour les types int et float, nous avons les opérateurs mathématiques. Pour le type str, nous avons les opérateurs de concaténation et de répétition. Les Booléens ont aussi leurs opérateurs : **not, and, or** (négation, et, ou)

- **not (négation)**: donne la négation de la valeur qui le suit.

```python
# Example26
a = True
b = not a
c = not b
d = not c
print(a, b, c, d)
e = not not not a
print(e)
```

- **and (et)**: donne la valeur de vérité de deux expressions booléennes où si les deux expressions sont vraies leur valeur **and** est vraie et si une expression ou les deux expressions sont fausses leur valeur **and** est fausse. Voir :

```python
# Example27
a, b = True, True
c, d = False, False
case1 = a and b
case2 = a and c
case3 = c and a
case4 = c and d
print(case1, case2, case3, case4)
a, b, c, d, e, f, g, h, i, j = True, True, True, True, True, True, True, True, True, False
case1 = a and b and c and d and e and f and g and h and i and j
case2 = a and b and c and d and e and f and g and h and i and not j
print(case1, case2)
```

- **or (ou)**: donne la valeur de vérité de deux expressions booléennes où si l'une ou l'autre ou les deux des expressions est vraie leur valeur **or** est vraie et si l'une ou l'autre est fausse leur valeur **or** est fausse. Voir : 

```python
# Example28
a, b = True, True
c, d = False, False
case1 = a or b
case2 = a or c
case3 = c or a
case4 = c or d
print(case1, case2, case3, case4)
a, b, c, d, e, f, g, h, i, j = True, False, False, False, False, False, False, False, False, False
case1 = a or b or c or d or e or f or g or h or i or j
case2 = not a or b or c or d or e or f or g or h or i or j
print(case1, case2)
```

Comme tu as fait avec les maths, tu peux utiliser les parenthèses pour gérer l'ordre de l'évaluation de l'expression booléenne. Il faut se souvenir que les parenthèses rendre l'expression embarquée, cela veut dire que la partie de l'expression entouré par les parenthèses la plus interne est évalué avant que les parties de l'expression entourées par les parenthèses extérieures soient évaluées et les parties qui ne sont pas entourées par les parenthèses sont évalués à la fin. Essayer de calculer la valeur de l'expression au-dessous avant que tu exécutes la cellule.

```python
# Example29
a, b = True, False
laVéritéSiJeMens = ((not (a and b)) or not b) and (not (not a or b) and a)
print(laVéritéSiJeMens)
```
