## Entier (int)
Les entiers sont la représentation de la valeur des nombres en base 10 (0-9). En théorie, les entiers éntendent entre l'infini négatif et l'infini positif, mais en raison des limites de l'ordinateur, nous ne pouvons pas faire le calcul avec les nombres extrêmement grands.

## Les nombres en virgule flottante (décimaux) Floating Point (float)

## Les Chaînes de Caractères String (<code>str</code>)

La chaîne de caractères est une séquence de caractères adjacents. Comme nous l'avons vu dans les leçons précédentes, nous obtenons une chaîne de caractères en mettant la séquence de caractères entre guillemets.

```python
# Exemple15
mon_string = "Ceci est un String"
mon_string2 = "Ca c'est un string aussi"
```

En fait, les guillemets ne sont pas le seul moyen de former une chaîne de caractères, nous pouvont également utiliser l'apostrophe.

```python
# Exemple16
texte1 = 'Hello, World'
texte2 = "Hello, World"
print(texte1)
print(texte2)
```

Les variables texte1 et texte2 ont le même contenu, le fait que l'un a été fait avec " et l'autre avec ' ne change rien.

Le guillemet et l'apostrophes sont les caractères qui peut être utilisé dans les chaînes des caractères. Mais ils sont aussi utilisé pour créer les chaînes de caractères. Alors, comment de faire la différence entre un guillemet ou un apostrophe qui est un caractère et un guillemet ou un apostrophe qui démarque une chaîne ?

```python
#Exemple17
amuser = 'Le Python utilise les guillemets anglais "Python est mon bonheur"'
print(amuser)
amuser = "Le Python utilise les apostrophes aussi 'Python est mon bonheur'"
print(amuser)
```

Une solution est créer la chaîne avec les guillemets quand il y a d'apostrophes dans la chaîne et les apostrophes quand il y a de guillements dans la chaîne, comme l'exemple a montré.

```python
#Exemple18
amuser = 'Le Python utilise les guillemets anglais "Python, c'est mon bonheur"'
print(amuser)
```

```python
#Exemple18 Corrigé
amuser = 'Le Python utilise les guillemets anglais "Python, c\'est mon bonheur"'
print(amuser)
```

## Booléen (bool)

Pour notre dernier type Python de base, nous avons l'honneur de vous présenter le fameux Booléen (bool). En programmation, le booléen est la valeur de vérité d'une expression en Python. Il peut avoir une de deux valeurs possibles : True (Vrai) ou False (Faux). Les valeurs booléennes sont binaires, cela veut dire qu'il n'y a que deux valeurs possible, comme le bit informatique avec ses valeurs de 0 ou de 1.

```python
# Example24
affirm1 = True
print(type(affirm1))
print(affirm1)
affirm1 = False
print(type(affirm1))
print(affirm1)
```

### bool

En Python, toutes les valeurs ont une valeur de vérité. Tout est vrai sauf **False, None, 0, et les chaînes de caractères vides '' et ""** qui sont faux par défaut. Nous pouvons utiliser **bool** pour vérifier la valeur de vérité d'une variable.

```python
# Example25
vérité1, vérité2, vérité3, vérité4, vérité5 = bool(None), bool(False), bool(0), bool(''), bool("")
print(vérité1, vérité2, vérité3, vérité4, vérité5)
vérité1, vérité2, vérité3, vérité4, vérité5, vérité6 = bool(-1), bool(2e-100), bool(True), bool('Thanos'), bool(" "), bool(23.65)
print(vérité1, vérité2, vérité3, vérité4, vérité5, vérité6)
```

### Opérateurs Booléens (not, and, or)

Pour les types int et float, nous avons les opérateurs mathématiques. Pour le type str, nous avons les opérateurs de concaténation et de répétition. Les Booléens ont aussi leurs opérateurs : **not, and, or** (négation, et, ou)

- **not (négation)**: donne la négation de la valeur qui le suit.

```python
# Example26
a = True
b = not a
c = not b
d = not c
print(a, b, c, d)
e = not not not a
print(e)
```

- **and (et)**: donne la valeur de vérité de deux expressions booléennes où si les deux expressions sont vraies leur valeur **and** est vraie et si une expression ou les deux expressions sont fausses leur valeur **and** est fausse. Voir :

```python
# Example27
a, b = True, True
c, d = False, False
case1 = a and b
case2 = a and c
case3 = c and a
case4 = c and d
print(case1, case2, case3, case4)
a, b, c, d, e, f, g, h, i, j = True, True, True, True, True, True, True, True, True, False
case1 = a and b and c and d and e and f and g and h and i and j
case2 = a and b and c and d and e and f and g and h and i and not j
print(case1, case2)
```

- **or (ou)**: donne la valeur de vérité de deux expressions booléennes où si l'une ou l'autre ou les deux des expressions est vraie leur valeur **or** est vraie et si l'une ou l'autre est fausse leur valeur **or** est fausse. Voir : 

```python
# Example28
a, b = True, True
c, d = False, False
case1 = a or b
case2 = a or c
case3 = c or a
case4 = c or d
print(case1, case2, case3, case4)
a, b, c, d, e, f, g, h, i, j = True, False, False, False, False, False, False, False, False, False
case1 = a or b or c or d or e or f or g or h or i or j
case2 = not a or b or c or d or e or f or g or h or i or j
print(case1, case2)
```

Comme tu as fait avec les maths, tu peux utiliser les parenthèses pour gérer l'ordre de l'évaluation de l'expression booléenne. Il faut se souvenir que les parenthèses rendre l'expression embarquée, cela veut dire que la partie de l'expression entouré par les parenthèses la plus interne est évalué avant que les parties de l'expression entourées par les parenthèses extérieures soient évaluées et les parties qui ne sont pas entourées par les parenthèses sont évalués à la fin. Essayer de calculer la valeur de l'expression au-dessous avant que tu exécutes la cellule.

    **Affirmations**

>>> - 1 la negation de la phrase 'a'

```python
# Ce programme demande si l'utilisateur est un étudiant, s'il a un travail et s'il a un diplôme. 
# Puis, il imprime la valeur de vérité pour plusieurs conditions de combination des 
# valeurs de vérité de ses réponses.
a = "Je suis un étudiant." # prompt pour vérité a
b = "J'ai un travail."     # prompt pour vérité b
c = "J'ai un diplôme."     # prompt pour vérité c
print("Répondre aux questions avec 1 pour Vrai (True) ou 0 pour Faux (False)")
vérité_a = bool(int(input(a)))  # valeur de vérité pour a
vérité_b = bool(int(input(b)))  # valeur de vérité pour b
vérité_c = bool(int(input(c)))  # valeur de vérité pour c

print("La négation de \"" + a + "\" est " + str((not vérité_a)))
print("\"" + a + "\" or \" " + b + "\" est " + str((not vérité_a or vérité_b)))
print("\"" + a + "\" and \" " + c + "\" est " + str((not vérité_a and vérité_c)))
print("\"" + c + "\" or \" " + b + "\" est " + str((not vérité_c or vérité_b)))
complexe = ((vérité_a or vérité_b) and (vérité_a and vérité_c) and ((not vérité_b) or vérité_c) )
print('("' + a + '" or "' + b + '") AND')
print('"' + a + '" and "' + c + '") AND')
print('"(not ' + b + '") or "' + c + '") EST ' + str(complexe))
```

```python
# Calculer le volume d'un cône utilisant le formule 1/3*pi*r-squared*h où r est le rayon et h est l'hauteur
PI = 3.14 # Valeur de π (PI)
rayon = 5
hauteur = 10

print("Le volume d'un cône de rayon ", rayon, "et de hauteur", hauteur, " est ", 1/3*PI*rayon*rayon*hauteur)
```
