# Leçon 4 : Les Conditions et Les instructions Conditionnelles


## Bloc de code

En Python, un bloc de code est un groupe de lignes qui fonctionnent ensemble et qui ne sera exécuté que si sa condition est évaluée à **True** (vrai).
Si la condition est évaluée à **False** (faux), le bloc de code ne sera pas exécuté.
Il a la forme suivante :

- La condition est une ligne de code composée d’une expression Python qui peut être évaluée en une valeur booléenne (**True** ou **False**).
Cela peut sembler un peu compliqué, mais cela deviendra clair comme de l’eau de roche, comme le dit l’expression française.
La condition doit être suivie de <b>:<b>. Tu vas certainement oublier plusieurs fois ce petit deux-points, tu es averti 😌 !
Si la condition est True, la ligne ou les lignes situées en dessous seront exécutées.

- Toutes les lignes qui suivent le <b>:<b> et qui font partie du bloc doivent être indentées, soit avec des espaces, soit avec une tabulation.

**L’indentation** est le fait de décaler une ligne de code vers la droite afin d’indiquer qu’elle appartient à un bloc de code particulier. En Python, l’indentation est très importante, car elle détermine la structure du programme.

Habituellement, 4 espaces sont utilisés, mais ce n’est pas obligatoire.
Toutes les instructions d’un même bloc doivent être indentées exactement au même niveau (c’est-à-dire décalées vers la droite du même nombre d’espaces).

L’indentation va aussi te piéger, nous te le garantissons, même si tu es encore une fois averti 😌!
Les blocs de code peuvent également être imbriqués (un bloc dans un autre bloc), et la règle d’indentation s’applique à chaque niveau.
Tu dois donc mettre systématiquement tes 4 espaces ou ta tabulation après chaque condition suivie de <b>:<b>.
Encore une fois, chaque bloc de code, même à l’intérieur d’un autre bloc, doit être indenté.
C’est comme ça que fonctionne Python.

Mais restons concrets : nous allons bientôt voir des exemples de blocs de code dans cette leçon.



## Les Conditions

Les déclarations conditionnelles ont une valeur de vérité, c’est-à-dire les valeurs booléennes de Python, comme nous l’avons vu dans la leçon précédente.
Elles peuvent être **True** (vrai) ou **False** (faux).

Avant d’exécuter la cellule ci-dessous, essaie de calculer les valeurs de vérité qui seront affichées, afin de faire une petite révision des booléens.

```python
# Exemple1
p = True
q = False
print("p : ", p)
print("q : ", q)
print ("(p or q) : ", p or q)
print ("(not q) : ", not q)
print ("(p and q) : ", p and q)
cond1 = (p or q) and (not q) or (p and q)
print("(p or q) and (not q) or (p and q) : ", cond1) 
```

Dans cet exemple, cond1 est un exemple de déclaration conditionnelle.
Nous avons étudié les opérateurs booléens not, and et or dans la leçon précédente.

Maintenant, nous allons apprendre les opérateurs de comparaison pour former des déclarations conditionnelles.
Ils seront extrêmement utiles pour prendre des décisions et contrôler le flux de nos programmes.

| opérateurs | Description |
|:----------:|:-----------:|
| **==** | Opérateur pour tester l'égalité, sous la forme de **a == b**, il demande si a est égal à b |
| **<** | Opérateur pour tester l'inégalité inférieure à, sous la forme de **a < b**, il demande si a est inférieur à b |
| **>** | Opérateur pour tester l'inégalité plus que, sous la forme **a > b**, il demande si a et plus que b |
| **<=** | Opérateur pour tester une inégalité inférieure ou égale, sous la forme **a <= b**, il demande si a est inférieur ou égal à b |
| **>=** | Opérateur pour tester une inégalité supérieure ou égale à, sous la forme **a >= b**, il demande si a est supérieur ou égal à b |
| **!=** | Opérateur pour tester une inégalité non égale, sous la forme **a != b**, il demande si a n'est pas égal à b |
| **in** | Opérateur pour tester s'il est membre d'une collection, sous la forme **a in b** où b est une séquence ou une collection, il demande si a est membre de b |
| **is** | Opérateur pour tester l'égalité des objets, sous la forme **a is b**, il demande si a est une référence du même objet que b |

Dans les exemples ci-dessous, essaie de déterminer les valeurs de vérité qui seront affichées avant d’exécuter la cellule.

```python
# Exemple2.1
# Opérateur pour tester l'égalité, sous la forme de a == b, il demande si a est égal à b
# valide pour les nombres et les chaînes
print("3 == 3 : ", 3 == 3)
print("3 == 4 : ", 3 == 4)
print("3 == int('3') : ", 3 == int(3))
print("None == 0 : ", None == 0)
print('"three" == "three" : ', "three" == "three")
print('3 == "3" : ', 3 == "3")
print("False == -3 : ", False == -3)
print('None == "None" : ', None == "None")
print("True == True : ", True == True)
print("True == 1 : ", True == 1)
print("False == 0 :", False == 0)
print("3 == 3.0000000000001 : ", 3 == 3.0000000000001)
print('"three" == "Three" : ', "three" == "Three")
print("True == False : ", True == False)
print("3 == 3.0 : ", 3 == 3.0)
print("True == 3 : ", True == 3)
```

```python
# Example 2.2
# Opérateur pour tester si deux valeurs sont différentes, sous la forme a != b, Il vérifie si a n’est pas égal à b et renvoie une valeur booléenne
# valide pour les nombres ou les chaînes
print("3 != 3 : ", 3 != 3)
print("3 != 4 : ", 3 != 4)
print("3 != int('3') : ", 3 != int(3))
print("None != 0 : ", None != 0)
print('"three" != "three" : ', "three" != "three")
print('3 != "3" : ', 3 != "3")
print("False != -3 : ", False != -3)
print('None != "None" : ', None != "None")
print("True != True : ", True != True)
print("True != 1 : ", True != 1)
print("False != 0 :", False != 0)
print("3 != 3.0000000000001 : ", 3 != 3.0000000000001)
print('"three" != "Three" : ', "three" != "Three")
print("True != False : ", True != False)
print("3 != 3.0 : ", 3 != 3.0)
print("True != 3 : ", True != 3)
```

```python
# Example 2.3
# Opérateur pour tester l'inégalité « inférieur à », sous la forme a < b, il vérifie si a est inférieur à b
# Opérateur pour tester l'inégalité « supérieur à », sous la forme a > b, il vérifie si a est supérieur à b
# Opérateur pour tester l'inégalité « inférieur ou égal à », sous la forme a <= b, il vérifie si a est inférieur ou égal à b
# Opérateur pour tester l'inégalité « supérieur ou égal à », sous la forme a >= b, il vérifie si a est supérieur ou égal à b
# Utilise pour les nombres
print("3 > 3 : ", 3 > 3)
print("3 < 3 : ", 3 < 3)
print("3 >= 3 : ", 3 >= 3)
print("3 <= 3 : ", 3 <= 3)
print("3 > 4 : ", 3 > 4)
print("3 < 4 : ", 3 < 4)
print("3 >= 4 : ", 3 >= 4)
print("3 <= 4 : ", 3 <= 4)
print("3.01 > 3 : ", 3.01 > 3)
print("3.01 < 3 : ", 3.01 < 3)
print("3.01 >= 3 : ", 3.01 >= 3)
print("3.01 <= 3 : ", 3.01 <= 3)
```

```python
# Exemple 2.3
# Opérateur pour tester si deux objets sont identiques, sous la forme a is b, il vérifie si a et b font référence au même objet
x = "chaîneDeCaractères"
y = "chaîneDeCaractères"
print("x == y : ", x == y)
x1 = x
print("y is x : ", y is x) # question 1
print("x1 is x : ", x1 is x) # question 2
x1 = y
print("x1 is x : ", x1 is x) # question 3
print("x1 is y : ", x1 is y) # question 4
```

## Les instructions Conditionnelles

Les instructions conditionnelles sont essentielles, car elles permettent au programmeur de contrôler la logique d’un programme.

Cela signifie que le programme peut prendre des décisions en fonction de certaines conditions et exécuter du code parmi plusieurs branches possibles.
On peut dire que le branchement conditionnel est au cœur de la programmation informatique.
Il permet à un programme de produire des résultats différents selon les circonstances.

Peut-être trouves-tu cette discussion “as clear as mud” (c’est-à-dire aussi limpide que de la boue).
Bon, comme d’habitude, nous allons donner un exemple concret.

Supposons que tu as construit un train autonome, où ton logiciel remplace le chef de train. Autrement dit, un train qui se conduit tout seul.

Vous arrivez à un point où le train doit aller soit à droite, soit à gauche.
Normalement, les deux directions mènent à la gare, mais il paraît qu’aujourd’hui des cheminots travaillent sur les rails du côté gauche.

Tu n’as pas prévu cette situation et tu as 30 secondes pour améliorer ton programme avant que le train  prenne la décision d’aller à droite ou à gauche.

Comment vas-tu écrire le programme pour sauver la vie des cheminots et préserver la confiance de tes actionnaires ?

```python
# Exemple3
CheminotsGauche = True

if CheminotsGauche:
    print("Virer à droite")
```

Mais c’est quoi ce **if** (si en français) ?
**if** signifie : exécute le bloc de code uniquement si la condition est vraie.

Ici, nous avons un exemple d’un bloc de code indenté de 4 espaces, même si le bloc ne contient qu’une seule ligne de code après les deux-points.

**if** est un mot-clé en Python. Il est suivi d’une condition, et si cette condition est évaluée à True, le bloc qui suit est exécuté.
La condition après le **if** peut être entourée de parenthèses. Elles sont facultatives, mais recommandées, car elles aident les lecteurs humains à comprendre le code.

Supposons maintenant que, lors d’un deuxième essai, les cheminots se déplacent vers la droite.
Nous voulons toujours nous assurer que, si les gens sont soit à gauche, soit à droite, le train prend la voie libre.

```python
# Example4
CheminotsGauche = False

if ( CheminotsGauche ) :
    print("Virer à droite")
else:
    print("Virer à gauche")
```

**else** (sinon en français) fait partie de la déclaration if. Nous l’utilisons lorsque la condition suivant la déclaration if n’est pas **True**, et que nous voulons que le programme exécute un autre bloc (un branchement).

En fait, **else** n’est pas seul : nous avons aussi **elif** à notre disposition.
**elif** est une contraction de else if (sinon si en français).
On utilise **elif** lorsqu’il y a des conditions supplémentaires que nous voulons tester.
Les **elif** peuvent s’enchaîner ; si nécessaire, nous pourrions avoir 100 conditions (le nombre est illimité), l’une après l’autre.
Le programme va tester chaque condition successivement.
Une fois qu’il trouve une condition **True**, il exécute son bloc de code puis arrête les tests de conditions, terminant la déclaration **if**.

Si toutes les conditions sont **False** et qu’il y a, à la fin, un **else**, le programme exécutera son bloc de code.
S’il n’y a pas de **else**, le test des conditions se termine et rien n’est exécuté.

Pour résumer, voici le format complet des instructions conditionnelles :

**if** : une ligne qui est requise pour toutes les déclarations de if.

**elif** : suit soit une déclaration if, soit d’autres déclarations elif. Une déclaration if peut avoir 0 ou autant de déclarations elif.

**else** : doit être le dernier bloc dans une déclaration **if**. Il est appelé lorsque la condition de **if** ne tient pas, et aucune clause **elif** n’est vraie. Il peut y avoir 0 ou 1 clause else dans une déclaration if.
