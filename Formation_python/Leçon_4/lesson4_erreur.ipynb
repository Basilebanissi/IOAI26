3. Tu vas modifier Exemple 6. Les trois voies sont aménagées pour gérer les trains à différentes vitesses, la voie gauche pour les trains lents qui roule à moins que 100km/heure, la voie du milieu pour les trains de moyenne vitesse qui roule entre 100km/heure et 200km/heure, et la voie droite pour les trains qui roule à plus 200km/heure. Tu vas diriger le train vers la voie qui correspond à sa vitesse, mais si cette voie n'est pas dégagée, tu diriges le train vers la voie avec la plage des valeurs de vitesse plus proche à la valeur de vitesse actuelle du train. Bien sûr, si toutes les voies sont bloquées, le programme est inchangé. Tu devrais demander à l'utilisateur la vitesse du train et aussi les valeurs de vérité pour les cheminots sur chaque voie. Copie Exemple 6 à la cellule au-dessous avant de commencer de le modifier.


```python
# Erreur: Le programme ne fait pas ce qui a été demandé par le probléme, la logique est défecteuse. Il y a aussi les fautes
# au niveau de Python. Par exemple, les valeurs de CheminotsGauche, Droite, Milieu seront les chaînes de caractères, pas les
# booléens ... la valeur d'une chaîne de caractères est True, les tests pour les voies dégagées échoueront toujours. Il n'y
# a pas de logique pour choisir la voie selon la vitesse du train.

# Example6 Modifier : Réponse aux questions
# Programme de frainage d'un Train via la detection des Cheminots
# Nous allons demander à l'utilisateur d'entrer les valeurs de vérité pour les cheminots sur chaque voie et aussi la vitesse du train
# Déclaraition des valeur de vérités initiale
CheminotsGauche = True
CheminotsDroite = True
CheminotsMilieu = True
# Déclaraition des valeur de vérités par l'utilisateur
CheminotsGauche = input("Veuillez entrer une valeur de vérité pour les cheminot de la voie gauche (1 = oui = True, 0 = non = False): ")
CheminotsDroite = input("Veuillez entrer une valeur de vérité pour les cheminot de la voie droite (1 = oui = True, 0 = non = False): ")
CheminotsMilieu = input("Veuillez entrer une valeur de vérité pour les cheminot de la voie du milieu (1 = oui = True, 0 = non = False): ")
# Variables de calcule
Vitesse = float(input('Veuillez entrer la vitesse du train: ')) # km/heure
Décélération = 4 # km/heure / sec
DistanceACheminots = 2.0 # km
VitesseArrivée = Vitesse - (DistanceACheminots / Vitesse * 3600 * Décélération) # formule de la Vitesse en km/h
#  Déclarations des Instructions Conditionnelles
if(CheminotsGauche == False  ): # SI CheminotsGauche est évaluer à False ALORS
    print("Virer à gauche") # Résultat
elif(CheminotsDroite == False): # SINON SI CheminotsDroite est évaluer à False
    print("Virer à droite") # Résultat
elif(CheminotsMilieu == False): # SINON SI CheminotsMilieu est évaluer à False
    print("Prendre la voie du milieu") # Résultat
elif (VitesseArrivée >= 0) : # SI VitesseArrivée est superieur ou égale à 0 ALORS       
    print("La vitesse du trein sera " + str(VitesseArrivée) + "km/h au site de travail.") # Résultat
elif (VitesseArrivée < 10.0): # SI VitesseArrivée est inferieur à 10 ALORS
    print("Freinez le train.") # Résultat
elif (VitesseArrivée > 10.0 and VitesseArrivée < 20.0 ): # SINON SI VitesseArrivée est superieur à 10.0 et que VitesseArrivée inferieur 20.0
    print("Freinez le train et déclencher la sirène d'alarme") # Résultat
elif (VitesseArrivée < 40.0 ): # SINON SI VitesseArrivée est inferieur 40.0
    print("Freinez le train et appelez l'ambulance") # Résultat
else:   # SINON VitesseArrivée est plus que 40 km/h
    print("Proposez la vente de la totalité de ta société aux actionneurs immediatement !") # Résultat
```
