## Bloc de code

En Python, un bloc de code est un groupe de lignes qui fonctionnent ensemble et qui ne sera pas exécuté sauf dans le cas où sa condition est évaluée **True** (vrai). Si sa condition est évaluée **False** (faux) le bloc de code ne sera pas exécuté. Il a la forme suivante :
- La condition est une ligne de code composée d'une expression Python qui peut être résolue à une valeur booléen (**True** ou **False**). Cela semble, peut-être, compliqué, mais il sera clair comme l'eau de roche, comme le veut l'expression française. La condition doit être suivie de <b>:</b>. Tu vas certainement oublier beaucoup de fois ce petit double point, tu es en averti ! Si la condition est TRUE, la ligne ou les lignes, au-dessous, seront exécutées.
- Toutes les lignes qui suivent le <b>:</b>, et qui font partie du bloc doivent être décalées ou par d'espaces ou par une touche de tabulation. Habituellement, 4 espaces sont utilisés mais ce n’est pas obligatoire. Toutes les instructions d’un même bloc doivent être mis en retrait exactement au même niveau (c’est-à-dire décalées à droite d’un même nombre d’espaces). L'indentation va aussi te piéger, nous te le garantissons, même si tu es encore une fois averti ! Les blocs de code peuvent certainement être embarqué dans les blocs de code, le règle d'indentation s'applique chaque fois. Tu dois mettre systématiquement ou tes 4 espaces ou ta tabulation après chaque condition et <b>:</b>. Encore une fois, chaque bloc de code, même sous un autre bloc de code, doit être décalé. Ca, c'est Python.

On ne reste pas dans les discours abstraits, on passe bientôt aux exemples de blocs de code dans cette leçon.

## Les Conditions

Les déclarations conditionnelles ont une valeur de vérité, les valeurs booléennes de Python, comme nous l'avons vu dans la leçon précédente. Elles sont (**True**) Vrai ou (**False**) Faux. Avant d'exécuter la cellule au-dessous, calcule les valeurs de vérité qui seront imprimées afin de faire une petite révision des valeurs booléennes.

```python
# Exemple1
p = True
q = False
print("p : ", p)
print("q : ", q)
print ("(p or q) : ", p or q)
print ("(not q) : ", not q)
print ("(p and q) : ", p and q)
cond1 = (p or q) and (not q) or (p and q)
print("(p or q) and (not q) or (p and q) : ", cond1) 
```

Dans cet exemple, *cond1* est un exemple de déclaration conditionnelle. Nous avons étudié les opérateurs booléennes **not**, **and**, **or** dans la leçon précédente. Maintenant, nous aurons le bonheur d'apprendre des opérateurs de comparaison pour former des déclarations conditionnelles qui seront extrêmement utiles pour effectuer les décisions de contrôle dans nos programmes.

| opérateurs | Description |
|:----------:|:-----------:|
| **==** | Opérateur pour tester l'égalité, sous la forme de **a == b**, il demande si a est égal à b |
| **<** | Opérateur pour tester l'inégalité inférieure à, sous la forme de **a < b**, il demande si a est inférieur à b |
| **>** | Opérateur pour tester l'inégalité plus que, sous la forme **a > b**, il demande si a et plus que b |
| **<=** | Opérateur pour tester une inégalité inférieure ou égale, sous la forme **a <= b**, il demande si a est inférieur ou égal à b |
| **>=** | Opérateur pour tester une inégalité supérieure ou égale à, sous la forme **a >= b**, il demande si a est supérieur ou égal à b |
| **!=** | Opérateur pour tester une inégalité non égale, sous la forme **a != b**, il demande si a n'est pas égal à b |
| **in** | Opérateur pour tester s'il est membre d'une collection, sous la forme **a in b** où b est une séquence ou une collection, il demande si a est membre de b |
| **is** | Opérateur pour tester l'égalité des objets, sous la forme **a is b**, il demande si a est une référence du même objet que b |

Dans les exemples au-dessous, trouve les valeurs de vérité qui seront imprimées avant que tu exécutes la cellule

## Les Instructions Conditionnels

Les instructions conditionnelles sont essentielles pour donner la facilité de la contrôle de logique au programmeur. C'est-à-dire que le programme a la capacité de prendre des décisions sur la base des conditions, et d'exécuter le code parmi de multiples branchements du code. On peut dire que le branchement conditionnel est au cœur de la programmation informatique. Il permet à notre programme de nous donner des résultats basés sur les circonstances. 

Peut-être, tu trouves cette discussion "as clear as mud" (aussi limpide que de la boue). Bon, comme d'habitude, nous donnerons un exemple concret. Supposons que tu as construit un train où ton logiciel remplace le chef de train, cela veut dire, un train d'auto-conduite. Tu vas gagner des millions de dollars avec ton invention ! Toi et tes actionnaires (les personnes qui ont financé le projet) soyez à bord pour le test initial. Vous êtes arrivé à un point où le train doit aller soit à droite, soit à gauche. Normalement, les deux sens mènent à la gare, mais il paraît qu'aujourd'hui il y a des cheminots travaillent sur les rails sur le côté gauche.

Tu n'as pas prévu cette situation et tu as 30 secondes pour améliorer ton programme avant que le train doit prendre la décision d'aller à droite ou à gauche. Comment vas-tu rédiger le programme pour sauver les vies des cheminots et la bonne foi de tes actionnaires ? (Nous allons rendre le programme plus simple en donnant les commandes au train par les imprimer à l'écran.)

```python
# Exemple3
CheminotsGauche = True

if CheminotsGauche:
    print("Virer à droite")
```

Voilà ! T'as sauvé ta société ! Mais c'est quoi, ce **if** (si en français) ? **if** veut dire, exécute son bloc de code uniquement si la condition est vraie. Nous avons ici aussi un exemple d'un bloc de code décalé par les 4 espaces, même si le bloc n'est plus qu'une seule ligne de code après les deux points.

**if** est un mot-clé en Python. Il est suivi d'une condition, et dans le cas où la condition est évaluée à **True**, le bloc qui la suit est exécuté. La condition après l'**if** pourrait être entourée par les parenthèses. Les parenthèses sont facultatives, mais elles sont recommandables, car elles aident les lecteurs humains à comprendre ton code.

Supposons maintenant que dans notre exemple, lors d'un deuxième essai, les cheminots se déplacent vers la droite, et nous voulons toujours nous assurer que si les gens sont soit à gauche soit à droite, le train prend la route libre.

```python
# Example4
CheminotsGauche = False

if ( CheminotsGauche ) :
    print("Virer à droite")
else:
    print("Virer à gauche")
```

**else** (sinon en français) fait partie de la déclaration if, nous l'utilisons dans le cas où la condition suivant la déclaration if n'est pas **True**, et nous voulons que le programme exécute un autre bloc (un branchement).

En fait, **else** n'est pas seul, nous avons aussi **elif** à notre disposition. **elif** est une contraction de else if (sinon si en français). On utilise **elif** s'il y a des conditions supplémentaires que nous voulons tester. Les **elif** peuvent enchaîner, le cas échéant, nous pourrions avoir 100 conditions (le nombre est illimité), l'une après l'autre. Le programme va tester chaque condition successivement. Une fois qu'il trouve une qui est **True** il exécute son bloc de code et, puis, il arrête d'exécuter les tests de conditions, terminant l'**if** déclaration. Si toutes les conditions sont **False** et il y a, à la fin, un **else** le programme va exécuter son bloc de code. S'il n'y a pas un **else** le test des conditions se termine et rien n'est pas exécute.

Pour résumer, voici le format complet des instructions conditionnelles :

- **if**: une ligne qui est requise pour toutes les déclarations de **if**.
- **elif**: suit soit une déclaration if ou d'autres déclarations elif, une déclaration if peut avoir 0 ou autant de déclarations elif.
- **else**: doit être le dernier bloc dans une déclaration de **if**, il est appelé lorsque **if** ne tient pas, et aucune clause elif ne tient également. Il peut y avoir 0 ou 1 **else** clause dans une déclaration **if**.

Le programme doit le prendre en compte. On saura le placement de cheminots et la distance depuis le train aux cheminots, la vitesse du train, et le taux de la décélération du train. Il faut que le programmeur, l'ancienne stagiaire qui tu as donné une promotion à Assistant Programmeur, ajoute les conditions suivantes :

Si toutes les voies sont bloquées et la vitesse à l'arrivée au site de travail est zéro ou plus que zéro et moins que 10km par heure, engage les freins. Les cheminots peuvent se sauver en sautant des rails. Si la vitesse est 10 km/h ou plus que 10 km/h et moins que 20km par heure, déclenche la sirène de l'alarme. Avec l'alarme il y a toute chance que les cheminots seront avertis en temps. Si la vitesse est 20km par heure ou plus que 20km/h et moins que 40km par heure, appelle le secours parce qu'il est probable que des cheminots seront blessés. Finalement, si, sur l'arrivée au site de travail la vitesse du train est 40 km/h ou plus, bof, il n'y a pas grande chose à faire. Voici le programme que ton Programmeur Assistant a écrit : 

```python
# Example6
CheminotsGauche = True
CheminotsDroite = True
CheminotsMilieu = True
Vitesse = 200 # km/heure
Décélération = 4 # km/heure / sec
DistanceACheminots = 2.0 # km

if(CheminotsGauche== False):
    print("Virer à gauche")
elif(CheminotsDroite == False):
    print("Virer à droite")
elif(CheminotsMilieu == False):
    print("Prendre la voie du milieu")
else:
    VitesseArrivée = Vitesse - (DistanceACheminots / Vitesse * 3600 * Décélération)
    if (VitesseArrivée < 0) : 
        VitesseArrivée = 0
    print("La vitesse du trein sera " + str(VitesseArrivée) + "km/h au site de travail.")
    if (VitesseArrivée < 10.0):
        print("Freinez le train.")
    elif (VitesseArrivée > 10.0 and VitesseArrivée < 20.0 ):
        print("Freinez le train et déclencher la sirène d'alarme")
    elif (VitesseArrivée < 40.0 ):
        print("Freinez le train et appelez l'ambulance")
    else :  # VitesseArrivée est plus que 40 km/h
        print("Proposez la vente de la totalité de ta société aux actionneurs immediatement !")
```

**Questions**  <br>

1. Le formule "Vitesse - (DistanceACheminots / Vitesse * 3600 * Décélération)" est un peu compliqué. L'Assistant Programmeur n'a pas fait un travail de qualité parce qu'elle n'a pas ajouté les commentaires pour expliquer la formule. Améliore son travail. <br>

3. Il y un bug dans le code. Trouve-le et le corrige. Astuce : Il s'agit d'une action imprévue quand la vitesse d'arrivée a une valeur particulière.

```python
# Example6 avec Modifications en Réponse aux Questions au-dessus
CheminotsGauche = True
CheminotsDroite = True
CheminotsMilieu = True
Vitesse = 200 # km/heure
Décélération = 4 # km/heure / sec
DistanceACheminots = 2.0 # km

if(CheminotsGauche== False):
    print("Virer à gauche")
elif(CheminotsDroite == False):
    print("Virer à droite")
elif(CheminotsMilieu == False):
    print("Prendre la voie du milieu")
else:
    # Réponse à Question 1
    # Formule pour obtenir la vitesse à l'arrivée
    #   Une approximation est utilisée pour calculer la décélération appliquée par les freins
    #   DistanceACheminots/Vitesse donne le temps, en heures, pour arriver aux cheminots sans décélération
    #   DistanceACheminots/Vitesse * 3600 donne le temps, en secondes, poura arriver aux chiminots sans décéleration
    #   Ce dernier numéro multiplie par la décéleration, en km/h/sec, donne la réduction de la vitesse au long du
    #   temps d'arriver aux cheminots à plein vitesse
    DiminuationVitesse = DistanceACheminots / Vitesse * 3600 * Décélération
    # Réponse à Question 2
    #   Le trein peut arrêter avant qu'il ait parcouru la distance aux cheminots. Si cela est le cas, la valeur
    #   pour la diminuation de vitesse sera plus grande que la vitesse au début, mais, en réalité le train arrête
    #   quand la valeur de la diminuation de la vitesse est egal à la vitesse initiale. Le programmeur a permis le
    #   calcul d'une vitesse d'arrivée negative et il a changé la valeur après à zéro. Nous avons exprimé directement
    #   dans le programme que la diminuation de la vitese ne peut pas dépasser la vitesse.
    if (DiminuationVitesse > Vitesse ) :
        VitesseArrivée = 0
    else :
    #   Nous soustractions cette valeur de la vitesse au début pour trouver la vitesse au numéro de secondes estimé 
    #   pour l'arrivée.           
        VitesseArrivée = Vitesse - DiminuationVitesse     
    print("La vitesse du trein sera " + str(VitesseArrivée) + "km/h au site de travail.")
    # Réponse à Question 3
    #   Le bug est que le programmeur a omis la valeur exacte de 10.0 dans la condition. Une valeur de 10.0
    #   aura été pris en compte par la condition else. Nous avons changé le premier condition de tester pour
    #   une valueur egal ou moins que 10.0
    if (VitesseArrivée <= 10.0):
        print("Freinez le train.")
    elif (VitesseArrivée > 10.0 and VitesseArrivée < 20.0 ):
        print("Freinez le train et déclencher la sirène d'alarme")
    elif (VitesseArrivée < 40.0 ):
        print("Freinez le train et appelez l'ambulance")
    else :  # VitesseArrivée est plus que 40 km/h
        print("Proposez la vente de la totalité de ta société aux actionneurs immediatement !")    
```

La voiture peut comprendre aussi les textes tapés sur l'écran. Écris un programme qui dira à la voiture ce qu'elle doit faire en approchant un feu. N'oublie pas le cas, trop courant au Mali, des feux qui ne marchent pas.

```python
# Problème2 - Solution
feu_etat = input("Est ce que le feu marche[oui/non]: ")

if(feu_etat == "oui"):
    feu_couleur = input("C'est quoi la couleur du feu[vert/jaune/rouge]: ")
    if(feu_couleur == "vert"):
        print("Vous pouvez aller.")
    elif(feu_couleur == "jaune"):
        print("Vous devez ralentir.")
    elif(feu_couleur == "rouge"):
        print("Arretez maintenant.")
    else:
        print("Option, est invalide.")
elif(feu_etat == "non"):
    print("Oops! Bonne chance...")
else:
    print("Option, est invalide.")
```

3. Tu vas modifier Exemple 6. Les trois voies sont aménagées pour gérer les trains à différentes vitesses, la voie gauche pour les trains lents qui roule à moins que 100km/heure, la voie du milieu pour les trains de moyenne vitesse qui roule entre 100km/heure et 200km/heure, et la voie droite pour les trains qui roule à plus 200km/heure. Tu vas diriger le train vers la voie qui correspond à sa vitesse, mais si cette voie n'est pas dégagée, tu diriges le train vers la voie avec la plage des valeurs de vitesse plus proche à la valeur de vitesse actuelle du train. Bien sûr, si toutes les voies sont bloquées, le programme est inchangé. Tu devrais demander à l'utilisateur la vitesse du train et aussi les valeurs de vérité pour les cheminots sur chaque voie. Copie Exemple 6 à la cellule au-dessous avant de commencer de le modifier.

```python
# Problème3 - Solution
Vitesse = float(input("Entrez la vitesse du train: ")) # km/heure

# Zero(0) pour dégagée et Un(1) pour pas dégagée
CheminotsGauche = int(input("Voie Gauche 0=voie dégagée 1=cheminots présents : "))
CheminotsMilieu = int(input("Voie Millieu 0=voie dégagée 1=cheminots présents : "))
CheminotsDroite = int(input("Voie Droite 0=voie dégagée 1=cheminots présents : "))

Décélération = 4 # km/heure / sec
DistanceACheminots = 2.0 # km

# Vitesse moins que 100 km/h, prendre gauche si dégagée
#   sinon prendre milieu si dégagée pcq elle est la voie plus proche à 100 km/h
#   sinon prendre droite si dégagée
if(Vitesse < 100):
    if(CheminotsGauche == 0):
        print("Virer à gauche")
    elif(CheminotsMilieu == 0):
        print("Prendre la voie du milieu")
    elif(CheminotsDroite == 0):
        print("Virer à droite")
# Vitesse entre 100 et 200 km/h, prendre milieu si dégagée
#   sinon, si juste une autre voie est dégagée, la prend
#   sinon, si la deux sont dégagée, choisir la voie pour la vitesse la plus proche à la vitesse actuelle
elif(Vitesse >= 100 and Vitesse <= 200):
    if(CheminotsMilieu == 0):
        print("Prendre la voie du milieu")
    elif(CheminotsGauche == 0 and CheminotsDroite == 1):
        print("Virer à gauche")  
    elif(CheminotsDroite == 0 and CheminotsGauche == 1):
        print("Virer à droite")
    elif(CheminotsGauche == 0 and CheminotsDroite == 0):
        if(Vitesse <= 149.5):
            print("Virer à gauche")
        else:
            print("Virer à droite")
# Vitesse supérieure à 200 km/h prendre la droite si dégagée
#   sinon prendre la milieu si dégagée sinon prendre la gauche si dégagée
elif(CheminotsGauche == 0 or CheminotsMilieu == 0 or CheminotsDroite ==0):
    if(CheminotsDroite == 0):
        print("Virer à droite")
    elif(CheminotsMilieu == 0):
        print("Prendre la voie du milieu")
    else:
        print("Virer à gauche")
#  Si aucun voie n'est pas dégagée, freiner le train
if (CheminotsGauche == 1 and CheminotsMilieu == 1 and CheminotsDroite == 1 and Vitesse > 0):
    # Formule pour obtenir la vitesse à l'arrivée
    #   Une approximation est utilisée pour calculer la décélération appliquée par les freins
    #   DistanceACheminots/Vitesse donne le temps, en heures, pour arriver aux cheminots sans décélération
    #   DistanceACheminots/Vitesse * 3600 donne le temps, en secondes, poura arriver aux chiminots sans décéleration
    #   Ce dernier numéro multiplie par la décéleration, en km/h/sec, donne la réduction de la vitesse au long du
    #   temps d'arriver aux cheminots à plein vitesse
    DiminuationVitesse = DistanceACheminots / Vitesse * 3600 * Décélération
    #   Le trein peut arrêter avant qu'il ait parcouru la distance aux cheminots. Si cela est le cas, la valeur
    #   pour la diminuation de vitesse sera plus grande que la vitesse au début, mais, en réalité le train arrête
    #   quand la valeur de la diminuation de la vitesse est egal à la vitesse initiale. Le programmeur a permis le
    #   calcul d'une vitesse d'arrivée negative et il a changé la valeur après à zéro. Nous avons exprimé directement
    #   dans le programme que la diminuation de la vitese ne peut pas dépasser la vitesse.
    if (DiminuationVitesse > Vitesse ) :
        VitesseArrivée = 0
    else :
    #   Nous soustractions cette valeur de la vitesse au début pour trouver la vitesse au numéro de secondes estimé 
    #   pour l'arrivée.           
        VitesseArrivée = Vitesse - DiminuationVitesse     
    print("La vitesse du trein sera " + str(VitesseArrivée) + "km/h au site de travail.")
    if (VitesseArrivée <= 10.0):
        print("Freinez le train.")
    elif (VitesseArrivée > 10.0 and VitesseArrivée < 20.0 ):
        print("Freinez le train et déclencher la sirène d'alarme")
    elif (VitesseArrivée < 40.0 ):
        print("Freinez le train et appelez l'ambulance")
    else :  # VitesseArrivée est plus que 40 km/h
        print("Proposez la vente de la totalité de ta société aux actionneurs immediatement !")    
```
