## print()

Comme nous l'avons vu tout au long du cours, la fonction **print()** affiche ce que tu lui demandes d'afficher.

Cela a été très utile jusqu'à présent, mais tu n'as pas encore vu tout ce que l'on peut faire avec **print()**

````python
# Exemple 1
print("Hot")
print("Dogs")
````

Ici, nous avons utilisé **print()** deux fois. Les textes s’afficheront sur deux lignes.

Pour les mettre sur la même ligne, nous pourrions utiliser notre vieil ami, l’opérateur de concaténation **+**.

Mais l’opérateur de concaténation ne fait pas toujours exactement ce que nous voulons, comme dans la situation suivante :

````python
# Exemple 2
print("J'ai" + 15 + "ans.")
````

Comme tu peux le voir, ceci affiche une erreur. Pourquoi ? C’est simple si tu comprends l’anglais :
"TypeError: can only concatenate str (not 'int') to str".

Pour ceux qui ne maîtrisent pas la langue de Shakespeare, cela veut dire : Erreur de type : on ne peut concaténer qu’une chaîne de caractères avec une autre chaîne de caractères, et non avec un entier.

Pas de panique, Python propose plusieurs solutions à ce problème.

````python
# Exemple 3
print("J'ai", 15, "ans.")
````

C’est aussi simple que ça : tout ce que tu as à faire est de mettre une virgule entre les valeurs. Pour la joie des paresseux, Python ajoute même un espace entre les valeurs pour toi.

Avec notre Python de niveau débutant des leçons précédentes, nous t’avions conseillé de faire cela manuellement afin de rendre tes phrases lisibles. Mais, si tu es un apprenant attentif, tu aurais remarqué que nous avons également utilisé cette manière d’utiliser print() dans les leçons précédentes.

Maintenant, retournons à notre premier exemple. Supposons que nous ne voulons pas qu’un saut de ligne soit ajouté après l’exécution de **print()**, mais plutôt seulement un espace.

Nous garderons nos deux **print()**, mais après la sortie **Hot** nous voulons voir "Dogs", avec un espace entre les deux mots.

Et rien n’est plus facile…

````python
# Exemple 4
print("Hot", end=" ")
print("Dogs")
````

**end**, fin en français, est un argument. Cela signifie une valeur que tu peux passer à la fonction **print** pour changer son comportement.

La valeur que l’on passe avec l’argument **end** est un caractère ou une chaîne de caractères qui sera émise à la fin de la sortie de print.

En réalité, **end** est toujours utilisé, mais il possède une valeur par défaut. Si l’argument **end** n’est pas ajouté à print, sa valeur par défaut est utilisée.
Et quelle est cette fameuse valeur par défaut ? Tu la connais déjà : le saut de ligne.

Le saut de ligne, ou newline en anglais, est en réalité un caractère spécial qui peut être écrit en Python ```\n```. C’est un caractère comme les lettres a, b ou c, sauf qu’on ne voit pas une lettre lorsqu’il est affiché : il provoque simplement un passage à la ligne suivante.

Dans l’exemple précédent, nous avons remplacé le newline par un espace. Nous pouvons demander à print de terminer la sortie avec un newline (remplacer la valeur par défaut par la même valeur), avec deux newlines, avec un tiret, rien du tout, ou une chaîne de caractères, comme tu peux le voir ci-dessous.

````python
# Exemple 5
print("Hot", end="\n")
print("Dogs")
print("Hot", end="\n\n")
print("Dogs")
print("Hot", end="-")
print("Dogs")
print("Hot", end="")
print("Dogs")
print("Hot", end="!!!!!!!!!!!!!!!!!!!")
print("Dogs")
````

## Mise en page de chaînes de caractères

### Caractères d'échappement

Tu as déjà vu que le newline s’écrit ```\n``` et que le retour arrière s’écrit `\b`. Il y a un lien entre les deux :

1. Chacun commence par une barre oblique inverse (**\\** ).
2.  La barre oblique inverse est suivie d’un caractère normal pour former ensemble un caractère spécial.

C’est tout à fait normal ! Ces caractères sont appelés caractères d’échappement et ils seront très utiles pour gérer l’affichage des chaînes de caractères.

Pour résumer, un caractère d’échappement a la forme suivante :

- Une barre oblique inverse (\\)
- Suivie par un caractère d’échappement

Par exemple, nous pouvons créer une chaîne de caractères sur plusieurs lignes en utilisant l’échappement pour le saut de ligne : **\n**.

````python
# Exemple 6
multiligne = "Je\nsuis\nune\nmachine"
print(multiligne)
````

Comme le montre la sortie de l’exemple précédent, c’est comme si nous avions utilisé quatre exécutions de la fonction **print**

Les caractères d’échappement les plus utiles dans ce cours sont :
- **\\n**: c’est le saut de ligne. Quand Python rencontre ce caractère dans une chaîne, il passe automatiquement à la ligne suivante, comme si tu avais écrit un nouvel appel à print().

- **\\t**: c’est le caractère de tabulation (tab en anglais). Il avance le texte d’une distance fixe, ce qui est pratique pour aligner des mots ou des nombres.

- **\\"**: Cela donne la représentation littérale de la double apostrophe. C’est un moyen d’éviter que Python considère la double apostrophe comme la fin de la chaîne de caractères.

- **\\'**: Cela donne la représentation littérale de l’apostrophe simple. C’est un moyen d’éviter que Python considère l’apostrophe simple comme la fin de la chaîne de caractères.

````python
# Exemple 7

PI = 3.1415

print("id\trang\téchantillon1\téchantillon2\n")

for i in range(10):
    samp2 = 313*((i**3)+((PI+(i*0.952))**3))
    print(str(i) + "\t" + str(i+1) + "\t" + str(i**i) + "\t" + str(samp2))

print()
````

Comme tu peux le constater, le caractère de tabulation nous permet de mettre en page les informations de manière assez bien organisée (en quelque sorte).

### Création de modèles

Python nous permet de créer des modèles génériques pour nos chaînes de caractères et de les manipuler avec toutes les valeurs que nous voulons.
Cette fonctionnalité rend le formatage du texte très simple, comme nous allons le découvrir dans cette section.

Pour les amateurs du Web, c’est une technique souvent utilisée pour générer du code HTML et ajouter ou mettre à jour dynamiquement des valeurs sur une page Web.

### Mise en page, style langage C

C est un langage de programmation plus ancien que Python, mais toujours largement utilisé et très apprécié par de nombreux programmeurs.
C est un langage de bas niveau (low-level), ce qui veut dire que programmer en C est un peu comme programmer directement le processeur. Cela peut être plus difficile pour les humains, car nous ne pensons pas comme un ordinateur, mais c’est plus performant, car le programmeur peut optimiser au maximum l’utilisation du matériel informatique.

Mais ici, l’important n’est pas le langage C lui-même. Ce que nous voulons montrer, c’est que Python a emprunté à C sa façon de créer des modèles pour les chaînes de caractères, ce qui permet de faire facilement des mises en page et du formatage de texte.

Dans une chaîne de caractères, le caractère **%** suivi d’un s, d ou f représente un paramètre qui sera remplacé par une valeur spécifique. Une chaîne contenant ces paramètres est appelée un modèle.

+ **%s** - représente un paramètre qui sera remplacé par une chaîne de caractères (str)
+ **%d** - représente un paramètre qui sera remplacé par un entier (int)
+ **%f** - représente un paramètre qui sera remplacé un nombre décimal (float)

Un exemple vaut mille mots.

````python
# Exemple 9
c_format = "Je suis %s, j'ai %d ans et je pèse %f kilogrammes. Je suis un peu gourmand." % ("Samory Toure", 38, 165.47)
print(c_format)
````

Dans l'exemple ci-dessus, nous avons une mises en forme complète en utilisant le style C.

- On trouve les trois types de paramètres **(%s, %d, %f)** dans le modèle texte.
- Le modèle est suivi par l’opérateur de formatage, le caractère **(%)**
- Enfin, entre parenthèses, on indique les valeurs à insérer, dans le même ordre que les paramètres dans le modèle.

La création d'un modèle et le remplacement de ses paramètres avec de valeurs peuvent être effectués dans différentes instructions, comme le montre l’exemple ci-dessous.

````python
# Exemple 10
modèle = "Le %s a une population de %d et un âge médian de %f ans en %d."
Mali2020 = modèle % ("Mali", 20416780, 16.3, 2020)
Japon2020 = modèle % ("Japon", 125507472, 48.4, 2020)
print(Mali2020)
print(Japon2020)
````

C’est bien, mais la mise en forme pourrait être améliorée, n’est-ce pas ? Par exemple, on n’a pas toujours besoin d’afficher tous les chiffres après le point décimal. Heureusement, Python nous facilite la tâche.

- Pour les nombres décimaux, nous pouvons indiquer la taille de la partie entière et de la partie décimale.- - Entre le caractère % et le caractère f dans le modèle, on met la taille des deux parties séparées par un point.
- Si la taille de la partie décimale n'est pas spécifiée, le défaut 6.

Par exemple, **\%2.1f** utilisé dans l'exemple ci-dessous. Fais des expériences en ajustant les tailles des parties entières et décimales pour voir ce qui passe.

````python
# Exemple 11
modèle = "Le %s a une population de %d et un âge médian de %2.1f ans en %d."
Mali2020 = modèle % ("Mali", 20416780, 16.3, 2020)
Japon2020 = modèle % ("Japon", 125507472, 48.4, 2020)
print(Mali2020)
print(Japon2020)
````

Et encore ...

Pour aligner une chaîne de caractères ou un entier dans un champ de taille fixe :
- On indique la taille du champ entre le caractère **%** et le **s** (pour chaîne) ou le **d** (pour entier).
- L’alignement à gauche se fait avec le signe moins (-), et l’alignement à droite est obtenu avec une valeur positive (par défaut).

Par exemple :
- **%8d** : aligne un entier à droite dans un champ de 8 caractères.
- **%12s** : aligne une chaîne de caractères à droite dans un champ de 12 caractères.

Comme illustré dans l’exemple ci-dessous.

````python
# Exemple 12
text = "%8d" % (1000)
print(text)
text = "%12s" % ("Mali")
print(text)
````

Et finalement ...

On peut contrôler le nombre de zéros de remplissage ou (ou zéros non-significatifs) pour les entiers dans un champ de taille fixée.

- Entre le **\%** et le **d**, on met la taille du champ, suivi d’un point, puis du nombre de chiffres à afficher.
- Si le nombre de chiffres à afficher est plus grand que le nombre réel, des zéros non-significatifs seront ajoutés à gauche pour atteindre le nombre de chiffres demandé.

Dans l'exemple, ci-dessous, Python ajoutera entre 0 et 4 zéros non-significatifs.

````python
# Exemple 13
text = "%8.4d" % (1000)
print(text)
text = "%8.5d" % (1000)
print(text)
text = "%8.6d" % (1000)
print(text)
text = "%8.7d" % (1000)
print(text)
text = "%8.8d" % (1000)
print(text)
````

OK, maintenant, nous avons tout ce qu’il faut pour faire des merveilles avec le formatage.
Refaisons l’Exemple 7 en utilisant nos super-pouvoirs de formatage !

````python
# Exemple 14
PI = 3.1415

print("%15s %15s %15s %15s" % ("id", "rang", "échantillon1", "échantillon2"))
print()

for i in range(10):
    samp2 = 313*((i**3)+((PI+(i*0.952))**3))
    print("%15d %15.2d %15.9d %15.3f" % (i, i+1, i**i, samp2))

print()
````

Comme tu peux le constater, par rapport à l’Exemple 7, la sortie est maintenant bien formatée, plus lisible et plus claire.

### .format()

Python nous offre de nombreuses possibilités pour le formatage des chaînes de caractères.
Le style inspiré du langage C est un peu ancien.

La méthode moderne consiste à utiliser la classe intégrée des chaînes de caractères (built-in string class en anglais).
C’est une approche orientée objet, et nous n’en avons pas encore étudié tous les détails.
Mais nous pouvons apprendre à l’utiliser simplement sans entrer dans la complexité de l’orienté objet.

C’est assez simple : il suffit d’ajouter **.format()** à nos modèles de chaînes de caractères.
Comment ça marche ? Découvrons-le !

````python
# Exemple 16

mon_texte = "La date est: {}/{}/{}.".format(25, 9, 2020)
print(mon_texte)
````

OK, cela ne semble peut-être pas si simple au premier abord, mais ça ira.

**.format()** fonctionne en plaçant des accolades **{}** dans la chaîne de caractères à l’endroit où tu veux insérer une valeur.
Ensuite, dans **.format()**, on met les valeurs dans le même ordre que les accolades.

Mais Python nous permet aussi de changer l’ordre.
Les arguments à l’intérieur des parenthèses de **.format()** sont numérotés, en commençant par 0.
On peut mettre ce numéro à l’intérieur des accolades pour spécifier quelle valeur doit remplacer quelle accolade.

Comme toujours, un exemple expliquera cela mieux que des mots.

````python
# Exemple 17

anniversaire = "Je m'appelle {3}, mon anniversaire est le {2}/{1}/{0}.".format(2005, 9, 22, "Abdoulaye Samake")
print(anniversaire)
````

Une autre fonctionnalité très intéressante de **.format()** est que nous pouvons utiliser des noms au lieu de numéros dans le modèle, ce qui rend la chaîne plus lisible et compréhensible.

````python
# Exemple 18

cout = 11000000
discount_pourcentage = 0.05
disc = int(cout - (cout * discount_pourcentage))
cond = "il est entièrement payé sur place"
format_ = "Le coût de la voiture est {vc}, avec un prix réduit de {pf} si {c}.".format(pf=disc, vc=cout, c=cond)
print(format_)
````

Maintenant, tu te demandes peut-être si nous pouvons ajouter des zéros pour remplir les nombres, aligner à droite ou aligner à gauche.
La réponse est, bien sûr, oui ! C’est Python : Python peut tout faire.
Et nous pouvons faire tout cela avec **.format()**.

````python
# Exemple 19

# alignement à droite - 10 espaces
print("{:>10d}".format(2020))

# alignement à gauche - 10 espaces
print("{text:<10s}".format(text="Afrique"))

# alignement au centre - 12 espaces
print("{:^12.1f}".format(2.7))

# remplissage de texte avec le caractère (-) - 10 espaces
print("{:->10s}".format("Sahara")) # Nous pouvons remplacer (-) avec d'autres caractères pour remplir notre texte.
````

C’est ainsi que le formatage se fait avec **.format()**.
Bien qu’il soit similaire au style du langage C, il nous donne un nouveau sac de techniques (bag of tricks) pour la mise en forme.

Est-ce que c’est tout ? Non, pas du tout !
Il y a encore une dernière technique à découvrir : les **f-strings**, la méthode moderne pour le formatage des chaînes de caractères.

### f-string

**f-string** est la dernière méthode ajoutée pour le formatage des chaînes de caractères.

Le grand avantage des **f-strings** est qu’elles permettent de mettre directement des variables ou des expressions Python à l’intérieur d’une chaîne.
C’est incroyablement simple : il suffit de coller un f devant la chaîne de caractères, puis de placer les variables ou expressions Python entre des accolades **{}**.
Le code Python à l’intérieur des accolades sera exécuté, et le résultat remplacera les accolades.

Comme toujours, un petit exemple parlera mieux que des mots.

````python
# Exemple 20

nombre = input("Entre un nombre pour voir son carré : ")
modèle1 = f"Le nombre est {nombre}."
modèle2 = f"Son carré est {int(nombre)**2}."
print(modèle1)
print(modèle2)
````

**Problème_4.**

Le grand polymathe Isaac Newton a proposé la formule suivante pour estimer les racines carrées des nombres:

    - $Y = \frac{x + \frac{S}{x}}{2}$

<strong>S</strong> est le numéro dont tu veux trouver la racine carrée.<br>
<strong>x</strong> est un numéro de supposition, cela veut dire, une estimation ou une supposition de la racine carrée de <strong>S</strong><br>
<strong>Y</strong> est la racine carrée de <strong>S</strong> calculée par la formule de Newton. Il faut retenir que le résultat est basé sur une supposition, donc il n'est pas forcement la bonne réponse.

Peut-être que tu n’as pas encore compris l’utilité de la formule de Newton.
En effet, c’est un algorithme numérique qui permet d’atteindre, ou de se rapprocher, de la bonne réponse grâce aux itérations.
À chaque répétition du calcul, on se rapproche toujours davantage de la solution correcte. Comment ? Chaque fois que le calcul est fait, après la première itération, tu utilises la valeur d'<strong>Y</strong> que tu as trouvé dans l'itération précédente comme la valeur de <strong>x</strong>.

On donne un exemple simple. Disons que tu voudrais calculer la racine carrée de 25. Pour amorcer les itérations, tu as deviné 6. Alors, <strong>S</strong> est 25 et <strong>x</strong> est 6. La première itération donne **5,083333333333334** pour le résultat <strong>Y</strong>, proche mais pas correct. Pour la deuxième itération, tu prends <strong>Y</strong> d'itération 1 pour la nouvelle <strong>x</strong> et tu calcules encore une fois une nouvelle <strong>Y</strong>. La deuxième <strong>Y</strong> est **5.0006830601092895**, ce qui est encore plus proche du bon résultat mais pas correct. On continue, la troisième itération donnerait un <strong>Y</strong> de **5.000000046650738**.

Avec <strong>25</strong>, c'est facile de savoir quand nous avons trouvé le bon résultat (après 5 itérations). Mais pour un autre nombre ? On peut savoir que l'on est très proche quand les résultats commencent à converger, cela veut dire que la différence entre <strong>Y</strong> après chaque itération est de plus en plus petite. Donc, on peut arrêter les itérations une fois que les différences entre <strong>Y</strong> sont devenues extrêmement petites. C'est à toi de décider ce que "petit" veut dire. Une technique très facile est de faire un bon nombre d'itérations et voir par l'œil si les différences sont petites. Si non, on augmente le nombre d'itérations. Ou, on peut comparer les résultats et une fois que la différence arrive à un seuil, on s'arrête.
