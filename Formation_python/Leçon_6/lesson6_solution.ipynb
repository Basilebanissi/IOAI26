## print()

Comme nous l'avons vu tout au long du cours, la fonction **print()** affiche ce que tu lui demandes d'afficher. Cela a été très utile jusqu'à maintenant, mais tu n'as pas encore vu tout que l'on puisse faire avec **print()** 

```python
# Exemple 1
print("Hot")
print("Dogs")
```

Ici, nous avons utilisé **print()** deux fois. Les textes vont s'afficher en deux lignes. Pour mettre les deux sur la même ligne, nous pourrions utiliser notre vieil ami, l'opérateur de concaténation **+**. Mais l'opérateur de concaténation ne fait pas toujours exactement ce que nous voulons qu'elle fasse, comme dans la situation suivante :

```python
# Exemple 2
print("J'ai" + 15 + "ans.")
```

Comme tu peux le voir, ceci affiche une erreur. Pourquoi ? C'est simple si tu comprends l'anglais "TypeError: can only concatenate str (not "int") to str". Pour les pauvres qui ne piègent pas la langue de Shakespeare, cela veut dire, Erreur de Type, on ne peut pas concaténer les chaînes de caractères aux ints. 

Pas de crise, Python donne plusieurs solutions au problème. 

```python
# Exemple 3
print("J'ai", 15, "ans.")
```

C'est aussi simple que ça, tout ce que tu as à faire est de mettre une virgule entre les valeurs. Pour la joie des paresseux, Python a même ajouter un espace entre les valeurs pour toi. Avec notre Python niveau nul des leçons précédentes, nous t'avions conseillé de faire cela manuellement afin de faire tes phrases lisibles. Mais, si tu es un apprenant attentive, tu aurais remarqué que nous avons utilisé cette méthode d'utilisation de print() aussi dans les leçons précédentes.

Maintenant, retournons à notre premier exemple. Supposons ne voulons pas qu'un saut de ligne soit mise après l'exécution de **print()**, mais, plutôt, juste un espace. Nous garderons nos deux **print()** mais après la sortie **Hot** nous voulons voir "Dogs" avec un espace entre les deux mots. Mais rien n'est plus facile ... 

```python
# Exemple 4
print("Hot", end=" ")
print("Dogs")
```

**end**, fin en français, est un argument, cela veut dire, une valeur que tu peux passer à la fonction **print** pour changer son comportement. La valeur que l'on passe par l'argument **end** est un caractère ou une chaîne de caractères qui sera émit à la fin de la sortie de **print**. En effet, **end** est toujours utilisé, mais il a une valeur défaut - si l'argument **end** n'est pas ajouté à la **print** sa valeur de défaut est utilisé. Et c'est quoi, cette fameuse valeur de défaut ? Tu la sais, c'est le saut de ligne. Le saut de ligne, ou, en anglais, newline, est, en réalité, un caractère qui peut être écrit en Python **\n**. C'est un caractère comme abc, sauf on ne voit pas un lettre quand il est imprimé ou affiché, mais un saut de ligne.

Dans l'exemple précédent, nous avons remplacé le newline avec un espace. Nous pouvons demander **print** de terminer la sortie avec un newline (remplace le défaut par le défaut), par deux newlines, par un tiret, par rien, par une chaîne de caractères, comme tu peux voir au-dessous.

```python
# Exemple 5
print("Hot", end="\n")
print("Dogs")
print("Hot", end="\n\n")
print("Dogs")
print("Hot", end="-")
print("Dogs")
print("Hot", end="")
print("Dogs")
print("Hot", end="!!!!!!!!!!!!!!!!!!!")
print("Dogs")
```

## Mise en page de chaîne de caractères

### Caractère d'échappement

Tu as juste vu que le newline est écrit **\n** et le retour arrière par **\b**. Il y a un lien entre les deux. D'abord, chacun commence avec une barre oblique inverse **\\** et, deuxièmement, la barre oblique inverse est suivie par un caractère normal pour faire, ensemble, un caractère un peu anormal. C'est normal ! Ces sont les caractères d'échappement et ils seront très utiles pour gérer l'affichage des chaînes de caractères. Pour le résume, les caractères d'échappements ont la forme suivante :

- Une barre oblique inverse(\\)
- Suivi par un caractère d'échappement

Par exemple, nous pouvons réaliser une chaîne de caractères multi-ligne en utilisant un échappement du saut de ligne, que nous avons mentionné précédemment **\\n**.

```python
# Exemple 6
multiligne = "Je\nsuis\nune\nmachine"
print(multiligne)
```

Comme le montre la sortie de l'exemple précédent, c'est comme si nous avions utilisé quatre exécutions de la fonction **print**. Les caractères d'échappement qui seront le plus utile à nous dans cette classe sont :

- **\n**: c'est le caractère d'échappement du saut de ligne, il nous position au début de la ligne suivante.

- **\t**: c'est la caractère de tabulation, tab en anglais Il émit le caractère de tabulation pour avancer la position une distance fixée.

- **\\"**: cela donne la représentation littérale du guillemet. C'est un moyen d'éviter que le guillemet soit traité comme la terminaison d'une chaîne de caractères.

- **\\'**: cela donne la représentation littérale du guillemet simple. C'est un moyen d'éviter que le guillemet simple soit traité comme la terminaison d'une chaîne de caractères.

```python
# Exemple 7

PI = 3.1415

print("id\trang\téchantillon1\téchantillon2\n")

for i in range(10):
    samp2 = 313*((i**3)+((PI+(i*0.952))**3))
    print(str(i) + "\t" + str(i+1) + "\t" + str(i**i) + "\t" + str(samp2))

print()
```

Comme tu peux le constanter, la caractère de tabulation nous permettre de mettre en page les informations assez bien rangées (en quelque sorte).

### Création de modèles

Python nous permet de créer les modèles génériques pour nos chaînes et de manipuler celles-ci avec toutes les valeurs que nous puissons vouloir. Cette fonctionnalité rend le formatage de texte à point un jeu d'enfant, comme nous le découvrirons dans cette section. Pour les amateurs du Web, c'est la technique utilisée très souvent pour générer du code source HTML et ajouter / mettre à jour dynamiquement des valeurs d'une page Web.

### Mise en page, style langage C

C est un langage de programmation, plus ancien que le Python, mais toujours largement utilisé et très apprécie par beaucoup de programmeurs. C est un langage de bas niveau, low-level, cela veut dire, la façon de programmer en C est un peu comme la programmation directe du processeur, peut-être plus difficile pour les êtres humains qui ne pensent pas comme les ordinateurs mais plus performant parce que le programmeur peut faire la meilleure utilisation de la matériel informatique.

Ca, c'est un peu trop sur le langage C. La chose que nous avons voulu de dire ici est juste que C a prêté à Python sa façon de créer les modèles de chaînes de caractères. Donc, il y a la mise en page en Python utilisant le style du langage C.

A l'intérieur d'une chaîne de caractères, la caractère **%** suivie par une s, d, ou f est un paramètre d'un certain type qui sera remplacé par une valeur. Une chaîne de caractères avec ces paramètres est un modèle.

+ **%s** - représente un paramètre qui sera remplacé par une chaîne de caractères (str)
+ **%d** - représente un paramètre qui sera remplacé par un entier (int)
+ **%f** - représente un paramètre qui sera remplacée un nombre décimal (float)

Un exemple vaut mille mots.

```python
# Exemple 9
c_format = "Je suis %s, j'ai %d ans et je pèse %f kilogrammes. Je suis un peu gourmand." % ("Samory Toure", 38, 165.47)
print(c_format)
```

Dans l'exemple ci-dessus, nous avons une mise en forme complète en utilisant le style C.

- On trouve les trois types de paramètres **(%s, %d, %f)** dans le modèle texte.
- Le modèle est suivi par l'opérateur de motif, le caractère **(%)** 
- Et enfin, entre parenthèses, les valeurs dans l'ordre des paramètres dans le modèle

La création d'un modèle et le remplacement de ses paramètres avec de valeurs peuvent être effectués dans différentes instructions, comme l'exemple, au-dessous, montre.

```python
# Exemple 10
modèle = "Le %s a une population de %d et un âge médian de %f ans en %d."
Mali2020 = modèle % ("Mali", 20416780, 16.3, 2020)
Japon2020 = modèle % ("Japon", 125507472, 48.4, 2020)
print(Mali2020)
print(Japon2020)
```

C'est bien, mais, la mise en forme pourrait être mieux, n'est pas ? Par exemple, on n'a pas besoin de tous les chiffres après le point décimal. Python vient à notre secours.

- Pour les nombres décimaux, nous pouvons indiquer la taille de la partie entière et de la partie décimale. Entre le caractère **\%** et le caractère <b>f</b> dans le motif nous mettons la taille de chacune des deux parties séparées par un point. Si la taille de la partie décimale n'est pas spécifiée, le défaut de 6 sera utilisé. Par exemple, **\%2.1f** utilisé dans l'exemple au-dessous. Fais des expériences en ajustement les tailles des parties entières et décimales pour voir ce qui arrive.

```python
# Exemple 11
modèle = "Le %s a une population de %d et un âge médian de %2.1f ans en %d."
Mali2020 = modèle % ("Mali", 20416780, 16.3, 2020)
Japon2020 = modèle % ("Japon", 125507472, 48.4, 2020)
print(Mali2020)
print(Japon2020)
```

Et encore ...

Nous pouvons aligner une chaîne de caractères ou un entier à gauche ou à droit. 

- Pour aligner une chaîne de caractères ou un entier sur un champ de taille fixée, nous mettons entre le **\%** et le **s ou le d** la taille du champ et nous indiquons l'alignement à gauche par le signe minus et l'alignement à droite par une valeur positive. Par exemple, **\%8d** alignera un entier à gauche dans un champ de taille de 8 et **\%12s** alignera une chaîne de caractères à gauche dans un champ de taille de 12, comme au-dessous.

```python
# Exemple 12
text = "%8d" % (1000)
print(text)
text = "%12s" % ("Mali")
print(text)
```

Et finalement ...

On peut contrôler le nombre de zéros de remplissage ou zéros non-significatifs pour les entiers dans un champ de taille fixée.

- Entre le **\%** et le **d** nous metton la taille du champ, suivi par un point, suivi par le nombre de chiffres d'afficher. Si le nombre de chiffres d'afficher et plus le grand que le nombre sans zéros non-significatifs, les zéros non-significatifs seront ajoutés jusqu'au nombre de chiffres demandés. L'exemple, au-dessous, ajoutera entre 0 et 4 zéros non-significatifs.

```python
# Exemple 13
text = "%8.4d" % (1000)
print(text)
text = "%8.5d" % (1000)
print(text)
text = "%8.6d" % (1000)
print(text)
text = "%8.7d" % (1000)
print(text)
text = "%8.8d" % (1000)
print(text)
```

OK, maintenant, nous avons tous qu'il faut pour faire de merveilles de formatage. Refaisons l'Exemple 7 en utilisant nos superpowers.

```python
# Exemple 14
PI = 3.1415

print("%15s %15s %15s %15s" % ("id", "rang", "échantillon1", "échantillon2"))
print()

for i in range(10):
    samp2 = 313*((i**3)+((PI+(i*0.952))**3))
    print("%15d %15.2d %15.9d %15.3f" % (i, i+1, i**i, samp2))

print()
```

Comme tu le vois, par rapport à Exemple 7, il s'agit d'une sortie bien formatée, plus lisible et plus claire. C'est la beauté du formatage de chaînes de caractères; nous pouvons jouer avec notre sortie jusqu'à ce que nous ayons quelque chose de cohérent.

### .format()

Python nous donne un embarras de choix pour la mise en forme des chaînes de caractères. Le style de C est un peu à l'ancienne. La méthode moderne est d'utiliser une classe intégrée dans le langage Python pour le traitement de chaînes de caractères (built-in string class ... plus simple en anglais !). Il s'agit de l'orientation objet et nous ne sommes pas encore arrivé là-bas. Mais, nous pouvons apprendre son utilisation sans d'entrer dans tous les détails. C'est assez simple. Nous allons juste appose **.format()** à nos modelés de chaînes de caractères. Comment ça ? Apprenons.

```python
# Exemple 16

mon_texte = "La date est: {}/{}/{}.".format(25, 9, 2020)
print(mon_texte)
```

OK, peut-être pas si simple comme ça. Mais, ça ira.  **.format()** marche en mettant des accolades **{}** dans la chaîne de caractères où tu veux mettre une valeur. Ensuite, dans **.format()** nous mettons les valeurs dans l'ordre des accolades.

Sauf, Python nous donne la possibilité de changer l'ordre. Les arguments entre les parenthèses de **.format()** sont numérotés, commençant avec 0. On peut mettre ce numéro entre les accolades pour spécifier son remplacement à n'importe quel argument de **.format()**. Comme toujours, un exemple va l'expliquer mieux que nos paroles.

```python
# Exemple 17

anniversaire = "Je m'appelle {3}, mon anniversaire est le {2}/{1}/{0}.".format(2005, 9, 22, "Abdoulaye Samake")
print(anniversaire)
```

Une autre fonctionnalité très intéressante de **.format()** est le fait que nous pouvons utiliser les noms plutôt les numéros dans le modèle pour rendre le modèle plus lisible.

```python
# Exemple 18

cout = 11000000
discount_pourcentage = 0.05
disc = int(cout - (cout * discount_pourcentage))
cond = "il est entièrement payé sur place"
format_ = "Le coût de la voiture est {vc}, avec un prix réduit de {pf} si {c}.".format(pf=disc, vc=cout, c=cond)
print(format_)
```

Maintenant, tu te demandes peut-être si nous pouvons faire le remplissage des nombres, l'alignement à droite et l'alignement à gauche. La réponse est, bien sûr, oui. C'est Python, Python peut tout faire. Et nous pouvons le faire avec **.format()**.

```python
# Exemple 19

# alignement à droite - 10 espaces
print("{:>10d}".format(2020))

# alignement à gauche - 10 espaces
print("{text:<10s}".format(text="Afrique"))

# alignement au centre - 12 espaces
print("{:^12.1f}".format(2.7))

# remplissage de texte avec le caractère (-) - 10 espaces
print("{:->10s}".format("Sahara")) # Nous pouvons remplacer (-) avec d'autres caractères pour remplir notre texte.
```

C'est ainsi que le formatage est fait avec **.format()**, bien qu'il soit similaire au style de C; cela nous donne un nouveau bag of tricks (sac de tours ?, cela veut dire : techniques) pour la mise en forme. 

C'est donc fini ? Non, pas de tout. Il y en a encore. Jetons maintenant un œil à la technique finale de formatage de chaînes de caractères, **f-string**.

### f-string

**f-string** est le dernier ajout aux méthodes de formatage de chaînes de caractères. Le grand avantage de **f-string** est qu'il nous permit de mettre les variables ou les expressions Python directement dans la chaîne de caractères. C'est follement simple, il faut juste coller une **f** devant la chaîne de caractères et, puis, on peut mettre d'expressions Python et les variables entre les accolades. Le Python entre accolades sera exécuté, et le résultat le remplacera. Comme toujours, un petit exemple will speak volumes (dira long sur le sujet).

```python
# Exemple 20

nombre = input("Entre un nombre pour voir son carré : ")
modèle1 = f"Le nombre est {nombre}."
modèle2 = f"Son carré est {int(nombre)**2}."
print(modèle1)
print(modèle2)
```

5. Le grand polymathe Isaac Newton a proposé la formule suivante pour estimer les racines carrées des nombres:

    - $Y = \frac{x + \frac{S}{x}}{2}$

<strong>S</strong> est le numéro dont tu veux trouver la racine carrée.<br>
<strong>x</strong> est un numéro de supposition, cela veut dire, une estimation ou une supposition de la racine carrée de <strong>S</strong><br>
<strong>Y</strong> est la racine carrée de <strong>S</strong> calculée par la formule de Newton. Il faut retenir que le résultat est basé sur une supposition, donc il n'est pas forcement la bonne réponse.

Peut-être, tu n'as pas encore piégé l'utilité de la formule de Newton. En effet, c'est un algorithme numérique pour arriver à ou proche à la bonne réponse par les itérations. Chaque fois on répète le calcul, on arrive toujours plus proche à la bonne réponse. Comment ? Chaque fois que le calcul est fait, après la première itération, tu utilises la valeur d'<strong>Y</strong> que tu as trouvé dans l'itération précédente comme la valeur de <strong>x</strong>.

On donne un exemple simple. Disons que tu voudrais calculer la racine carrée de 25. Pour amorcer les itérations, tu as deviné 6. Alors, <strong>S</strong> est 25 et <strong>x</strong> est 6. La première itération donne **5,083333333333334** pour le résultat <strong>Y</strong>, proche mais pas correct. Pour la deuxième itération, tu prends <strong>Y</strong> d'itération 1 pour la nouvelle <strong>x</strong> et tu calcules encore une fois une nouvelle <strong>Y</strong>. La deuxième <strong>Y</strong> est **5.0006830601092895**, ce qui est encore plus proche du bon résultat mais pas correct. On continue, la troisième itération donnerait une <strong>Y</strong> de **5.000000046650738**. 

Avec <strong>25</strong>, c'est facile de savoir quand nous avons trouvé le bon résultat (après 5 itérations). Mais pour un autre nombre ? On peut savoir que l'on est très proche quand les résultats commencent à converger, cela veut dire que la différence entre <strong>Y</strong> après chaque itération est plus et plus petite. Donc, on peut arrêter les itérations une fois que les différences entre <strong>Y</strong> est devenues extrêmement petites. C'est à toi de décider ce que "petit" veut dire. Un technique très facile est de faire un bon nombre d'itérations et voir par l'œil si les différences sont petites. Si non, on augmente le nombre d'itérations. Ou, on peut comparer les résultats et une fois que la différence arrive à un seuil, on s'arrête.

Écris un programme qui demandera à un utilisateur un nombre dont il veut la racine carrée, et un nombre deviné. Ton programme doit appliquer cette méthode à ces valeurs jusqu'à ce que tu obtiens le bon résultat. Les deux nombres devraient être de types entiers. La sortie de ton programme doit ressembler à l'image ci-dessous, tous les nombres sont remplis de 8 chiffres.



```python
# Problème 5

valeur = int(input("Entre la valeur dont tu veux la racine carrée: "))
supposx = int(input("Entre un nombre de supposition: "))

steps = 1
resultat = 0

while(True):
    print(f"valeur de x à l'étape {steps:^8d} = {supposx:>8.10f}")
    Y = (supposx + (valeur/supposx)) / 2
    if(Y == supposx):
        resultat = Y
        break
    supposx = Y
    steps += 1

print(f'Le résultat final *{resultat:^8.2f}* est atteint en {steps:^8d} étapes')
```
