## Utilisation des fonctions

Les fonctions en programmation sont des commandes qui effectuent des instructions spécifiques. Les fonctions nous permettent de regrouper un certain nombre d'instructions sous une seule commande. Elles nous permettent aussi de nommer un gros bloc de code afin de le réutiliser plus tard. Le format d’une fonction est généralement le suivant :

- **nom de la fonction**: toutes les fonctions ont un nom qui décrit ce que fait la fonction.
- **(\[paramètres\])**: le nom est directement suivi d'une paire de parenthèses. 
Entre les parenthèses, il peut y avoir des paramètre. Les parametres sont les données que la fonction utilisera dans ses calculs. Si la fonction n'a pas besoin de paramètres, on garde les parenthèses, mais elles restent vides.

"Appeler une fonction" consiste à placer la fonction, en utilisant son nom et ses paramètres, dans notre code. Le bloc d’instructions de la fonction, déclaré ailleurs, sera alors exécuté comme s’il se trouvait à cet endroit dans le programme.

```python
# Exemple 1
import math
texte = "Nul ne peut se vanter de se passer des autres."

print()
print("Le miel n'est jamais bon dans une seule bouche.")
print(texte)
print("Pi = ", str(math.pi))
print("Note : ", texte, end="\n        ^^^")
```

Dans cet exemple, nous avons appelé une fonction qui se nomme **print** sans paramètres, avec une chaîne de caractères, avec une variable, avec deux paramètres dont une chaîne de caractères et une autre fonction **str**, et, finalement, avec trois paramètres : une chaîne de caractères, une variable, et un paramètre précédé par un mot-clé **end** et un '='. 

Heureusement, il n’y a rien de nouveau pour toi, mais maintenant il faut que tu comprennes ces exemples familiers comme des appels à une fonction avec différents arrangements de paramètres.

### filter (filtrer)

**filter()** est, comme **map()**, utilisé pour la programmation style *functional* en Python. La fonction **filter()** produit, aussi, un *iterator* où ses éléments ont été filtré par une fonction qui teste et accepte ou n'accepte pas chaque élément. La fonction doit, forcement, retourner avec des valeurs booléennes (**True** ou **False**).

Dans l'exemple ci-dessous **filter()** applique la fonction de str *isalpha* sur les éléments de la liste *details*. *isalpha* retourne True pour les éléments qui contiennent uniquement des lettres et False pour les éléments qui ontiennent des caractères non alphabétiques. **map()** applique la fonction de str *captialize* pour rendre la première lettre des éléments de *details* acceptés par **filter()** en majuscule.

```python
# Exemple 8
details = ["boubacar", "bouba@gmail.com", "79087676", 
           "mariam", "mtoure@yahoo.fr", "76202045"]

noms = map(str.capitalize, filter(str.isalpha,details))
list(noms)
```

### all (tous)

Cette fonction teste si tous les éléments d'un itérable sont True. Elle retourne True si tous les éléments sont True ; sinon, elle retourne False.

```python
# Exemple 9
details = ["boubacar", "bouba@gmail.com", "79087676", 
           "mariam", "mtoure@yahoo.fr", "76202045"]

print(all(list(map(str.isalpha, details))))
print(all(list(map(str.isprintable, details))))
```

Dans l'Exemple 9, tous les éléments de *details* ne sont pas composés uniquement de lettres, donc la liste générée par **map()** avec la fonction *is.alpha* ne contient pas exclusivement des valeurs True, et **all()** retourne False. Par contre, tous les éléments de *details* sont printable (capable d'être affichés), donc la liste générée par le deuxième **map()** ne contient que les valeurs true et **all()** retourne True.

```python
# Exemple 10
resultat = all(set([i for i in range(1, 10)]))
print("Valeur de all sur range(1, 10)", resultat)

resultat = all(set([i for i in range(10)]))
print("Valeur de all sur range(10)", resultat)
```

### any

Le contraire de **all()** est **any()** : si au moins un élément est True, le résultat sera True. any() retourne False seulement si tous les éléments sont False.

```python
# Exemple 11
details = ["boubacar", "bouba@gmail.com", "79087676", 
           "mariam", "mtoure@yahoo.fr", "76202045"]

print(any(list(map(str.isalpha, details))))
print(any(list(map(str.isprintable, details))))
print(any(list(map(str.isupper, details))))
```

Nous avons ajouté le test *str.isupper*, qui vérifie que tous les caractères sont des majuscules. Comme ce n'est pas le cas pour chacun des éléments de *details* toute la liste est False et **any()** retourne False.

### Définition

Le processus de création d'une nouvelle fonction est appelé «définition de fonction». Lorsque nous définissons une fonction, nous sommes responsables de la manière dont cette fonction doit se comporter.

Maintenant, comment définir une fonction ? 

Facile : nous utilisons le mot-clé **def** suivi du nom de la fonction, puis des parenthèses, et enfin de deux points.

```python
# Exemple 13

def ma_premiere_fonction():
    print("Hello World ! こんにちは世界 ! Je suis une fonction.")

ma_premiere_fonction()
```

C'était facile : dans l'exemple, nous avons défini une fonction nommée ma_premiere_fonction(), puis appelé cette fonction. Mais pourquoi n’avons-nous pas utilisé print() ici ? Ne t’inquiète pas, nous en discuterons bientôt.

Lors de la définition d’une fonction, nous devons prendre en considération que la fonction est un bloc de code ; par conséquent, elle doit suivre les règles des blocs. La dénomination des fonctions suit les mêmes règles que la dénomination des variables. Une bonne pratique pour nommer une fonction est de choisir un nom qui indique clairement ce que fait la fonction.

```python
# Exemple 14

TERRES = "Il y a environ 6 milliards de planètes semblables\
à la Terre dans l'univers observable."

def message_des_terres():
    "Affiche le nombre estimé de planètes semblables à la Terre. "
    print(TERRES)

message_des_terres()
```

Il existe un type spécial de commentaire pour les fonctions. Dans l’Exemple 14, c’est le texte placé entre guillemets juste sous le **def**. Une fois qu’une fonction est définie, la fonction **help()** en tient compte et affiche cette ligne de commentaire lorsque **help()** est appelée sur ta fonction. Bien sûr, tu peux aussi mettre des commentaires normaux dans la fonction, mais ils ne seront pas affichés par **help()**.

```python
# Exemple 15

help(message_des_terres)
```

Lorsque nous définissons une fonction, nous avons la possibilité de lui passer un ou plusieurs paramètres, comme nous avons fait avec les fonctions intégrées. Les paramètres d'une fonction sont les noms listés dans la définition de la fonction. Une fois que la fonction est exécutée, les paramètres ne sont plus les paramètres, mais les arguments, c'est-à-dire les valeurs réelles passées à la fonction. Rien de grave 

Rien de grave si tu confonds les mots paramètre et argument, tant que tu comprends, sous l’un ou l’autre nom, comment les utiliser.

```python
# Exemple 16

def repeter_quatre_fois(énoncé):
    for _ in range(4):
        print(énoncé)

repeter_quatre_fois("Meuh")
```

Dans cet exemple, *énoncé* est le paramètre et «Meuh» est l'argument. Nous pouvons changer la valeur de Meuh et la fonction affichera cette valeur quatre fois.

### Portée

En Python, il y a des règles qui définissent la portée des variables. Portée ? Les variables sont accessibles dans le programme selon la portée qui s'applique à elles. La portée peut être locale, c’est-à-dire accessible uniquement dans le conteneur (par exemple une fonction) où elles sont définies, ou globale, cela veut dire, accessible partout dans le programme.

Pourquoi a-t-on besoin de la notion de portée ? La raison principale est *name collision* (collision de noms). Supposons que tu aies défini deux fonctions dans ton programme et que chaque fonction possède une variable appelée *my_variable*. Est-ce que les deux instances de *my_variable* représentent la même variable ? Si je change la valeur de *my_variable* dans une fonction, est-ce que la valeur de *my_variable* dans l’autre fonction changera aussi ? Pas si la portée de chaque variable est locale : chaque variable est indépendante de l'autre. 

Cependant, il se peut que les deux fonctions doivent partager les mêmes données, par exemple un grand dictionnaire auquel chaque fonction doit accéder. Dans ce cas, tu dois créer ces données avec une portée globale.

```python
# Exemple 19
# variables globales
entrées = set()

# fonctions
def ajouteEnsemble(membre):
    entrées.add(membre)
def listeEnsemble():
    print(entrées)
def supprimeEnsemble(membre):
    entrées.remove(membre)

# programme principal
while(True) :
    action = input("Ensemble a=ajoute l=liste s=supprime q=quit")
    if(action == 'q'):
        break
    elif(action =='a'):
        élément = input("Entre valeur : ")
        ajouteEnsemble(élément)
    elif(action == 'l'):
        listeEnsemble()
    elif(action == 's'):
        élément = input("Entre valeur : ")
        supprimeEnsemble(élément)
    else:
        print("Entrée invalide")
```

L'Exemple 19 a trois sections, la déclaration des variables globales, les fonctions, et le programme principal. L'ensemble *entrées* est dehors des fonctions et le programme principal et, par conséquent, il est global.
Ensuite, il y a trois définitions de fonctions : ajouteEnsemble, listeEnsemble et supprimeEnsemble, deux avec un paramètre et une sans paramètre. Chacune de ces fonctions accède à l’ensemble global *entrées*.

Le programme principal suit les fonctions ; c’est là que l’exécution du programme commence. Il y a deux variables locales dans le programme principal, *action* et *élément*. Le programme principal n’accède pas directement à la variable globale : il utilise les fonctions pour tout accès et toute manipulation des données globales.

C’est une structure de programme très commune et bénéfique. L’idée est de canaliser l’accès aux données globales vers des fonctions conçues pour cela et d’éviter l’accès direct aux données globales depuis tous les coins du programme, c’est-à-dire d’une manière ingérable. Ce concept est développé plus profondément dans la programmation orientée objet, où il s’appelle l’*encapsulation*.

### Valeur de retour

Jusqu’à présent, nos fonctions n’ont renvoyé aucune valeur explicite. Ou du moins, nous pourrions le penser. En réalité, il existe toujours une valeur de retour : la valeur par défaut 'None'. Les fonctions intégrées **print()** et **help()** renvoient cette valeur par défaut.

Une fonction peut renvoyer sa propre valeur de retour en utilisant le mot-clé **return**, suivi d’une valeur du type approprié. Le plus souvent, **return** se trouve à la fin du bloc de la fonction. Une fois la fonction exécutée, l’appel de la fonction est remplacé par sa valeur de retour, et cette valeur peut être attribuée à une variable ou utilisée comme argument.

De la même manière qu’un programme bien organisé utilise des fonctions ou des méthodes pour accéder aux variables globales, il est préférable de ne pas mettre des **print()** partout dans les fonctions. Il vaut mieux que les fonctions renvoient leurs résultats au programme principal, et que ce dernier gère la communication avec l’utilisateur.

Dans l'Exemple ci-dessous, nous avons écrit deux versions d'une fonction qui teste si un nombre est pair ou impair. La première communique directement en utilisant un **print()** et  renvoie la valeur par défaut None. L'autre n'affiche pas de message, mais, elle renvoie une chaîne de caractères avec le résultat du test.

```python
# Exemple 20

def pair_ou_impair(n):
    print(str(n), "pair_ou_impair ? ", end='')
    print('pair' if n % 2 == 0 else 'impair')

def pair_ou_impair_2(n):
    return "pair" if n % 2 == 0 else "impair"

print("pair_ou_impair valeur de retour (implicite) :", 
      pair_ou_impair(24))
print("pair_ou_impair_2 valeur de retour (explicite) :", 
      pair_ou_impair_2(24))
```

Après l'exécution du mot-clé **return** toutes les lignes suivantes, s'il y en a, ne s'exécutent pas. **return** signifie que la fonction est terminée.

```python
# Exemple 21

def notre_factorielle(n):
    fact = 1
    if(n < 1):
        return 1
    for i in range(1, n+1):
        fact *= i
    return fact

while(True):
    nombre = input("Entre l'entier pour lequel vous\
    voulez son factorielle ou q pour quitter : ")
    if(nombre == 'q'):
        break
    elif(nombre.isdigit() != True):
        print("L'entrée doit être un entier !")
    else:
        print(notre_factorielle(int(nombre)))
```

### Fonctions Lambda

Les fonctions lambda sont de petites fonctions anonymes, c’est-à-dire qu’elles n’ont pas besoin d’identifiant ; leur dénomination est facultative. Ce sont des fonctions que nous pouvons définir directement sur la ligne où elles sont utilisées. Une fonction lambda possède une syntaxe restreinte : elle ne peut contenir qu’une seule expression.

Si notre introduction aux fonctions lambda paraît modeste, il ne faut pas s’y tromper. La fonction lambda est au cœur de la programmation functional et elle joue un rôle important dans l’histoire de l’informatique théorique. Mais nous avons promis de ne pas te tuer avec notre arme préférée : la théorie.

La syntaxe d'une fonction lambda est la suivante :

- mot clé lambda
- suivi de paramètres optionnels, séparés par des virgules s’il y en a plusieurs,
- suivi de deux points et de l'expression souhaitée.

```python
# Exemple 22

carré = lambda param: param**2

print(carré(12))
```

Dans l’Exemple 22, la fonction **lambda** possède un paramètre. Elle calcule le carré de son argument.

```python
# Exemple 23 

puissance_seq = map(lambda x: x**2, [i for i in range(10)])

print([i for i in puissance_seq])
```

## Récursion

Quel est le plus beau rêve qu’un informaticien puisse avoir ? C’est le rêve de la récursion. Si tu veux avoir une petite idée de la beauté de la récursion pour un informaticien, tu pourrais voir le film *Inception*. Ou bien tu pourrais simplement continuer cette leçon.

En termes simples, la récursivité est l’appel d’une fonction à l’intérieur de cette même fonction. La récursion permet souvent de résoudre un problème avec une élégance remarquable. Elle est également un élément important de la programmation fonctionnelle. On retrouve des mécanismes récursifs dans la nature, et certains informaticiens diraient même que la récursion pourrait être le secret de la construction de l’univers.

Comme elle pourrait être le secret de l’univers, on comprend que la récursivité puisse être intimidante au début. Mais tu es arrivé jusqu’ici, donc tu es courageux / courageuse. Même si les débuts peuvent être un peu difficiles, avec le temps la récursion deviendra plus… naturelle.

Voici comment fonctionne la récursion :

- Il faut un cas de base : lorsque ce cas est atteint, le programme s’arrête.
- Il faut un cas récursif : c’est lui qui répète le processus, un peu comme une itération.

```python
# Exemple 26

def notre_factorielle2(n):
    if(n <= 1):
        return 1
    return n * notre_factorielle2(n - 1)

print(notre_factorielle2(5))
```

Dans l’Exemple 27 ci-dessous, nous avons ajouté des **print** afin de produire une sortie qui montre chaque étape de la récursion du programme. Le code lui-même n’est pas le plus important : il faut surtout observer la sortie. Tu peux voir comme chaque appel à *notre_factorielle* ajoute un facteur, mais l'expression factorielle n'est pas complète jusqu'à l'arrivée au cas de base. C'est l'action de "monter" (techniquement, pousse au *stack* (pile) ... mais cela c'est juste pour les curieux😌). Une fois le cas de base atteint, tous les facteurs sont connus et nous redescendons dans les appels : les deux derniers facteurs sont multipliés à chaque étape avant d’exécuter le dernier **return**. Techniquement, on parle alors de pop de la pile.

```python
# Exemple 27
add = str()
tail = str()
def notre_factorielle2(n):
    global add, tail
    if(n <= 1):
        add += "1)))))"
        print(add)
        print("(5 * (4 * (3 * 2)))")
        print("(5 * (4 * 6))")
        print("(5 * 24)")
        return 1
    add += "(" + str(n) + " * "
    tail += ")"
    print(add + "notre_factorielle2(" + str(n-1) + ")" + tail)
    return n * notre_factorielle2(n - 1)

print(add + "notre_factorielle2(" + '5' + ")" )
print(notre_factorielle2(5))
```

Pour savoir si un élément appartient à un itérable en Python, nous avons heureusement l’opérateur **in**. Essayons maintenant de nous entraîner en créant une implémentation de **in** à l’aide de la récursion.

```python
# Exemple 28

val = [25, 416, 961, 103, 834, 634, 623, 178, 50, 529]

def notre_in(lst, val):
    if(len(lst) <= 1):
        return False if len(lst) < 1 else val == lst[0]
    
    if(len(lst) == 2):
        return val == lst[0] or val == lst[1]

    prem, *reste, last = lst

    if(prem == val or last == val):
        return True
    return notre_in(reste, val)

print("Est 18 dans la liste val ?",notre_in(val, 18))
print("Est 103 dans la liste val ?",notre_in(val, 103))
```

Le programme qui cherche un élément dans une liste n’est pas obligé d’effectuer des calculs lorsqu’il revient du cas de base, contrairement au programme de la factorielle qui accumule les multiplications pour produire le résultat final. Dans le cas de la recherche dans une liste, il suffit simplement de transmettre la valeur finale jusqu’à l’étape où le calcul a commencé. Ce type particulier de récursion s’appelle la récursion terminale (tail recursion, ou récursion de queue).

**Question 8.7** Il existe un avantage à la récursion terminale par rapport à la récursion normale. Quel est cet avantage ?

La différence entre la récursion terminale et la récursion normale est très importante dans de nombreux langages de programmation, mais beaucoup moins en Python. Même si Python est très puissant, il ne fait pas vraiment d’optimisation spécifique pour distinguer efficacement ces deux formes de récursion.

Il existe une autre méthode pour déterminer si un élément appartient à une liste : la recherche dichotomique (binary search). Cette méthode est beaucoup plus efficace et elle utilise également une approche récursive. En fait, la recherche dichotomique est un algorithme fondamental en informatique qui mérite d’être étudié et compris parfaitement.


La recherche dichotomique exige que la liste soit ordonnée. Lorsqu’une comparaison est faite entre un élément de la liste et la valeur recherchée, on peut déterminer si la valeur cherchée est plus petite ou plus grande que cet élément. Grâce à cette information, on peut ignorer soit la moitié gauche, soit la moitié droite de la liste pour les comparaisons suivantes. On commence la recherche au milieu de la liste, puis à chaque étape récursive on divise encore la partie restante par deux.

Comme nous sommes extraordinairement gentils, nous avons ajouté des **print** dans le code avec les valeurs des variables clés afin de te permettre de suivre facilement l’exécution du programme.

```python
# Exemple 29
itérations = 0

valeurs = [
      1,  28,  55,  82, 109, 136, 163, 190, 244, 271, 325, 379, 406, 433, 460, 500, 
    509, 514, 541, 568, 595, 622, 676, 784, 811, 838, 865, 892, 919, 946, 973
]
def recherche_dichotomique(lst, valeur, debut, fin):
    """ 
    recherche_dichotomique:
        lst: une liste
        valeur: valeur a chercher
        debut: debut de fragment a chercher
        fin: fin du fragment a chercher
    """
    global itérations
    itérations += 1

    milieu = debut + (fin-debut) // 2

    print("cherche entre index debut", debut, 
          "et index fin", fin, "teste à l'index milieu #",
          milieu, "son valeur:",lst[milieu])
    if(lst[milieu] == valeur):
        return True
    elif(fin == debut):
        return False
    elif(lst[milieu] < valeur):
        debut = milieu + 1

    elif(lst[milieu] > valeur):
        fin = milieu - 1

    return recherche_dichotomique(lst, valeur, debut, fin)     

while (True):
    itérations = 0
    nombre = input("Entre un entier ou q pour quitter ")
    if (nombre == 'q'):
        break
    else:
        if (recherche_dichotomique(valeurs, int(nombre), 0, len(valeurs)-1) == True):
            print(int(nombre),"trouvé !",end=' ')
        else:
            print(int(nombre),"n'est pas dans la liste",end=' ')
        print("(Numéro d'itérations :", itérations, "taille de la liste :", len(valeurs),")")
```
