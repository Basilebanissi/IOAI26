## Utilisation des fonctions

Les fonctions en programmation sont des commandes qui effectue des instructions spécifiques. Les fonctions nous permettent de regrouper un certain nombre d'instructions sous le couvert d'une seule commande. Les fonctions nous permettent de nommer un gros bloc de code pour une réutilisation ultérieure. Le format d'une fonction est généralement le suivant :

- **nom de la fonction**: toutes les fonctions ont un nom qui décrit ce que fait la fonction.
- **(\[paramètres\])**: le nom est directement suivi d'un pair de parenthèses. Entre les parenthèses, il se peut qu'il y a des paramètres. Les parametres sont les données que la fonction utilisera dans ses calculs. Si la fonction n'a pas besoin de paramètres, on garde les parenthèses, mais il n'y a rien entre elles.

"Appeler une fonction" est de placer la fonction, utilisant le nom de la fonction et ses paramètres, le cas échéant, dans notre code. Le bloc d'instructions de notre fonction, déclarée ailleurs, seront exécutées comme si elles se trouvent là-bas.

```python
# Exemple 1
import math
texte = "Nul ne peut se vanter de se passer des autres."

print()
print("Le miel n'est jamais bon dans une seule bouche.")
print(texte)
print("Pi = ", str(math.pi))
print("Note : ", texte, end="\n        ^^^")
```

Dans cet exemple, nous avons appelé un fonction qui se nomme **print** sans parametres, avec une chaîne de caractères, avec une variable, avec deux parametres dont une chaîne de caractères et une autre fonction **str**, et, finalement, avec 3 parametres dont une chaîne de caractères, une variable, et un parametre précédé par un mot-clé **end** et un '='. Heureusement, il y a rien de nouveau pour toi, mais, maintenant, il faut que tu comprennes ces exemples familiers comme appels à une fonction avec des différents arrangements de paramètres.

### map

**map()** est la fonction de *mappage*. Elle applique une fonction à chaque élément d'un itérable. Dans l'exemple au-dessous, **map()** applique la fonction pour trouver la racine d'un nombre à chaque élément d'une liste. Dans l'exemple, nous utilisons **list()** pour générer une liste avec les racines du résultat de l'exécution de **map()**.

Une boucle **for** peut être utilisé, souvent, pour accomplir la même chose que **map()**. Alors, pourquoi **map()**? La réponse est que Python est super-cool. **map()** est utilisé pour la programmation dans le style appelé *functional programming*. ### filter (filtrer)

**filter()** est, comme **map()**, utilisé pour la programmation style *functional* en Python. La fonction **filter()** produit, aussi, un *iterator* où ses éléments ont été filtré par une fonction qui teste et accepte ou n'accepte pas chaque élément. La fonction doit, forcement, revenir avec de valeurs booléennes (True ou False).

Dans l'exemple au-dessous **filter()** applique la fonction de str *isalpha* sur les éléments de la liste *details*. *isalpha* revient avec True pour les éléments qui consiste uniquement des lettres et False pour les éléments qui contient les caractères non-alphabétiques. **map()** applique la fonction de str *captialize* pour rendre les premières lettres des éléments de *details* acceptés par **filter()** en majuscule.

```python
# Exemple 8
details = ["boubacar", "bouba@gmail.com", "79087676", "mariam", "mtoure@yahoo.fr", "76202045"]

noms = map(str.capitalize, filter(str.isalpha,details))
list(noms)
```

### all (tous)

Cette fonction teste si tous les éléments d'un itérable sont True. Cela donnerait un True s'ils sont tous True, sinon cela donnerait un False.

```python
# Exemple 9
details = ["boubacar", "bouba@gmail.com", "79087676", "mariam", "mtoure@yahoo.fr", "76202045"]

print(all(list(map(str.isalpha, details))))
print(all(list(map(str.isprintable, details))))
```

Dans l'Exemple 9, tous les éléments de *details* ne sont pas tous les lettres, donc la liste générée par **map()** avec la fonction *is.alpha* ne contient pas exclusivement les vrais et **all()** est False. Par contre, tous les éléments de *details* sont printable (capable d'être imprimés), donc la liste générée par le deuxième **map()** ne contient que les éléments vrais et **all()** est True.

```python
# Exemple 10
resultat = all(set([i for i in range(1, 10)]))
print("Valeur de all sur range(1, 10)", resultat)

resultat = all(set([i for i in range(10)]))
print("Valeur de all sur range(10)", resultat)
```

### any

Le contraire de «all», si un seul élément est True, le résultat donnerait un True. Cela donnerait False seulement si tout est False.

```python
# Exemple 11
details = ["boubacar", "bouba@gmail.com", "79087676", "mariam", "mtoure@yahoo.fr", "76202045"]

print(any(list(map(str.isalpha, details))))
print(any(list(map(str.isprintable, details))))
print(any(list(map(str.isupper, details))))
```

Nous avons ajouté le test *str.isupper* qui vérifie que toutes les caractères sont les majuscules. Comme ce n'est pas le cas pour chacun des éléments de *details* toute la liste est False et **any()** est False.

### Définition

Le processus de création d'une nouvelle fonction est appelé «définition de fonction». Lorsque nous définissons une fonction, nous sommes responsables de la manière dont cette fonction doit se dérouler.

Maintenant, comment définir une fonction - facile, nous utilisons le mot-clé **def** suivi du nom de la fonction suivi de parenthèses et de deux points.

```python
# Exemple 13

def ma_premiere_fonction():
    print("Hello World ! こんにちは世界 ! Je suis une fonction.")

ma_premiere_fonction()
```

C'était facile, dans l'exemple, nous avons défini une fonction nommée ma_premiere_fonction(), et appelé cette fonction. Mais nous n'avons pas utilisé 'print()' ici pourquoi? Ne t'inquiétes pas, nous en discuterons bientôt.

Lors de la définition d'une fonction, nous devons prendre en considération que la fonction est un bloc de code, par conséquent, elle doit suivre les règles des blocs. La dénomination des fonctions suit la même règle que la dénomination des variables. Une bonne règle pour les noms de fonction est d'avoir des noms qui indiquent ce que fait la fonction.

```python
# Exemple 14

TERRES = "Il y a environ 6 milliards de planètes semblables à la Terre dans l'univers observable."

def message_des_terres():
    "Affiche le nombre estimé de planètes semblables à la Terre. "
    print(TERRES)

message_des_terres()
```

Il y a un spécial type de commentaire pour les fonctions, dans l'Exemple 14 il est le texte entre guillemets au-dessous le **def**. Une fois qu'une fonction est défini, la fonction **help()** sera au courant et elle affichera cette ligne de commentaire quand **help()** est appelée sur ta fonction. Bien sûr, tu peux mettre les commentaires normaux également dans la fonction, mais ils ne sont pas affichés par **help()**. 

```python
# Exemple 15

help(message_des_terres)
```

### Paramètres et Arguments

Lorsque nous définissons une fonction, nous avons la possibilité de lui passer un ou plusieurs paramètres, comme nous avons fait avec les fonctions intégrées. Les paramètres d'une fonction sont les noms listés dans la définition de la fonction. Une fois que la fonction est exécutée, les paramètres ne sont plus les paramètres, mais les arguments, c'est-à-dire les valeurs réelles passées à la fonction. Rien de grave si tu confonds les mots paramètre et argument si tu comprends, par un nom ou l'autre, comment les utiliser.

```python
# Exemple 16

def repeter_quatre_fois(énoncé):
    for _ in range(4):
        print(énoncé)

repeter_quatre_fois("Meuh")
```

Dans cet exemple, *énoncé* est le paramètre et «Meuh» est l'argument. Nous pouvons changer la valeur de Meuh et la fonction affichera cette valeur quatre fois. 

La caractère '\_' après **for** est une variable. '\_' est utilisé souvent dans les boucles où l'on ne fait pas l'utilisation de sa valeur dans le bloc de code, même si '\_' est réglo comme nom de variable et sa valeur est utilisable. ### Portée

En Python, il y a des règles qui définissent la portée des variables. Portée ? Les variables sont accessibles dans le programme selon la portée qui s'applique à elles. La portée peut être locale, cela veut dire, accessible uniquement dans le conteneur (par exemple, une fonction) où elles sont définies, ou globale, cela veut dire, accessible partout dans le programme.

Pourquoi a-t-on besoin de la notion de portée ? La raison principale est *name collision* (collision de noms). Suppose que tu as défini deux fonctions dans ton programme et chaque fonction a une variable *my_variable* ? Est-il que les deux instances de *my_variable* sont la même variable ? Si je change la valeur de *my_variable* dans une fonction, est-il que la valeur de *my_variable* dans l'autre fonction sera la même? Pas si la portée de chaque variable est locale, chaque variable est indépendant de l'autre. Mais, il se peut que les deux fonctions doivent partager les mêmes données, par exemple, un grand dictionnaire que chaque fonction doit accéder. Dans ce cas, tu as besoin de créer ces données avec une portée globale.

```python
# Exemple 19
# variables globales
entrées = set()

# fonctions
def ajouteEnsemble(membre):
    entrées.add(membre)
def listeEnsemble():
    print(entrées)
def supprimeEnsemble(membre):
    entrées.remove(membre)

# programme principal
while(True) :
    action = input("Ensemble a=ajoute l=liste s=supprime q=quit")
    if(action == 'q'):
        break
    elif(action =='a'):
        élément = input("Entre valeur : ")
        ajouteEnsemble(élément)
    elif(action == 'l'):
        listeEnsemble()
    elif(action == 's'):
        élément = input("Entre valeur : ")
        supprimeEnsemble(élément)
    else:
        print("Entrée invalide")
```

L'Exemple 19 a trois sections, la déclaration des variables globales, les fonctions, et le programme principal. L'ensemble *entrées* est dehors des fonctions et le programme principal et, par conséquent, il est global. Puis, il y a trois définitions de fonctions : ajouteEnsemble, listeEnsemble, et supprimeEnsemble, deux avec un paramètre et l'un sans paramètres. Chacun de ces fonctions accède à l'ensemble global *entrées*. Le programme principal suit les fonctions, il est là que l'exécution du programme commence. Il y a deux variables locales dans le programme principal, *action* et *élément*. Le programme principal n'accède pas à la variable globale, il utilise les fonctions pour tout accès et manipulation des données globales. Ceci est une structure de programmes très commune et bénéfique, l'idée est de canaliser l'accès aux données globales aux fonctions conçu pour cela et d'éviter l'accès direct aux données globales de tous les coins de programme, c'est-à-dire, dans une manière ingérable. Le concept est développé plus profondément dans la programmation orienté objet où elle s'appelle l'*encapsulation*.

### Valeur de retour

Jusqu'à présent, nos fonctions n'ont renvoyé aucune valeur. Ou nous pourrions le penser, mais, en effet, il y a une valeur de retour, la valeur de défaut : 'None'. Les fonctions intégrées **print()** et **help()** renvoient la valeur de retour défaut.

Une fonction peut renvoie sa propre valeur de retour en mettant  le mot-clé **return** suivi par une valeur du type approprié,. Le plus souvent, le **return** se trouve en dernière position dans le bloc. Une fois que la fonction est exécutée, l'appel de la fonction est remplacé par sa valeur de retour et cette valeur peut-être attribuée à une variable ou utilisée comme un argument.

Pareille à l'idée qu'un programme bien organisé utilise les fonctions ou les méthodes pour accéder aux variables globales, il est bénéfique de ne pas mettre les **print()** partout dans les fonctions. Pour savoir si un élément est dans un itérable en Python, nous avons, pour notre bonheur, **in**. Laissons-nous nous entraîner en créant une implementation de **in** avec la récursion. 

```python
# Exemple 28

val = [25, 416, 961, 103, 834, 634, 623, 178, 50, 529]

def notre_in(lst, val):
    
    if(len(lst) <= 1):
        return False if len(lst) < 1 else val == lst[0]
    
    if(len(lst) == 2):
        return val == lst[0] or val == lst[1]

    prem, *reste, last = lst

    if(prem == val or last == val):
        return True
    return notre_in(reste, val)

print("Est 18 dans la liste val ?", notre_in(val, 18))
print("Est 103 dans la liste val ?", notre_in(val, 103))
```

Il y a une autre méthode pour découvrir l'appartenance d'un élément dans une liste : la recherche dichotomique (binary search). Cette méthode est beaucoup plus efficace et elle utilise l'approche récursive. En effet, la recherche dichotomique est un algorithme fondamentale dans l'informatique qui vaut la peine de bien étudier et comprendre à la perfection.

La recherche dichotomique exige que la liste soit ordonnée. Quand une comparaison est effectuée sur un élément de la liste contre la valeur recherché, il faut que l'on peut déterminer si la valeur est plus petité ou plus grande que l'élément de la liste. Avec cette détermination, la partie droite ou la partie gauche de la liste peut être ignorée pour les comparaisons subséquentes. On commence la recherche en plein milieu de la liste et avec chaque récursion, on divise encore le reste par moitié.

Parce que nous sommes extraordinairement gentils, nous avons mis des **prints** dans le code avec les valeurs des variables clés pour te permettre de tracer facilement l'exécution du programme. 

```python
# Exemple 29
itérations = 0

valeurs = [
      1,  28,  55,  82, 109, 136, 163, 190, 244, 271, 325, 379, 406, 433, 460, 500, 
    509, 514, 541, 568, 595, 622, 676, 784, 811, 838, 865, 892, 919, 946, 973
]

def recherche_dichotomique(lst, valeur, debut, fin):
    """ 
    recherche_dichotomique:
        lst: une liste
        valeur: valeur a chercher
        debut: debut de fragment a chercher
        fin: fin du fragment a chercher
    """
    global itérations
    itérations += 1

    milieu = debut + (fin-debut) // 2

    print("cherche entre index debut", debut, "et index fin", fin, "teste à l'index milieu #", milieu, "son valeur:", lst[milieu])
    if(lst[milieu] == valeur):
        return True
    elif(fin == debut):
        return False
    elif(lst[milieu] < valeur):
        debut = milieu + 1

    elif(lst[milieu] > valeur):
        fin = milieu - 1

    return recherche_dichotomique(lst, valeur, debut, fin)     

while (True):
    itérations = 0
    nombre = input("Entre un entier ou q pour quitter ")
    if (nombre == 'q'):
        break
    else:
        if (recherche_dichotomique(valeurs, int(nombre), 0, len(valeurs)-1) == True):
            print(int(nombre),"trouvé !",end=' ')
        else:
            print(int(nombre),"n'est pas dans la liste",end=' ')
        print("(Numéro d'itérations :", itérations, "taille de la liste :", len(valeurs),")")
```

```python
# Problème 3

def pascal_triangle(depart, rows):
    for i in range(rows):
        for s in range(rows-i):
            print(format(" ", "^2"), end="")

        for j in range(i+1):
            if(j==0 or i == 0):
                binom = depart
            else:
                binom *= (i-j+1) / j
            print(format(int(binom), "^4"), end="")
        print()
    return

pascal_triangle(7, 10)
```

```python
# Problème 4

HEXSYMB = ['a', 'b', 'c', 'd', 'e', 'f']

def dec_2_bin(n):
    """ 
        Base10 à Base2 -- type(n) ==> int
        Les nombes negatifs et rationels ne sont pas accceptés
    """
    if(n < 0): return None
    
    bi_ = ""

    if(n == 0): bi_ += "0"

    while(n >= 1):
        if(n%2 == 0): bi_ += "0"
        else: bi_ += "1"
        n //= 2
    return bi_[::-1]

def hex_2_bin(n):
    """
        Base16 à Base2 -- type(n) ==> str
    """
    n = list(n)
    for i in n:
        if(i.lower() in HEXSYMB):
            n[n.index(i)] = 10 + HEXSYMB.index(i.lower())
    nibs = []
    for i in n:
        nibs.append(
            f"{dec_2_bin(int(i)):0>4}")
    return "".join(nibs) if None not in nibs else None

def base_10(n):
    """ Base 2 à Base 10"""
    bi = 0
    counter = 0
    for i in n:
        bi += 2 ** ((len(n)-1) - counter) * int(n[counter])  
        counter += 1
    return bi

def base_16(n):
    """ Base2 à Base16 """
    ln = list(n) # Liste de n

    while(len(ln) % 4 != 0): ln.insert(0, 0)
    chars = []
    compteur = 1

    val = ""
    for i in ln:
        if(compteur % 4 != 0): val += str(i)
        if(compteur % 4 == 0):
            val += str(i)
            chars.append(val)
            val = ""
        compteur += 1

    res = [str(base_10(i)) for i in chars]
    compteur = 0

    for i in res:
        if(int(i) >= 10):
            res[compteur] = HEXSYMB[int(i) - 10].upper()
        compteur += 1
    return "".join(res)

def montre_opts(x=None):
    """ remove x[0, 2] from the lists """
    res = ""
    opts = ["Binaire", "Decimal", "Hexadecimal"]
    if(x is not None): opts.pop(x)
    for i,j in enumerate(opts): res += f"{i+1}: {j} "
    return res, opts

def main():
    converter = 1
    while(converter):
        full_opts = ["Binaire", "Decimal", "Hexadecimal"]
        convertir_de = input(f"Convertir de {montre_opts()[0]} :")

        all_opts = montre_opts()[1]
        conv_de = int(convertir_de) - 1
        convertir_a = input(f"Convertir en {montre_opts(conv_de)[0]}")
        valeur = input("Valeur à convertir: ")
        opts = montre_opts(conv_de)[1]
        conv_cas = [full_opts[conv_de], opts[int(convertir_a)-1]]

        # Cas 1 - binaire -> decimale
        if(conv_cas[0] == "Binaire" and conv_cas[1] == "Decimal"):
            print(f"Binaire {valeur} ==> '{base_10(valeur)}' en Decimal")
        elif(conv_cas[0] == "Binaire" and conv_cas[1] == "Hexadecimal"):
            print(f"Binaire {valeur} ==> '{base_16(valeur)}' en Hexadecimal")
        elif(conv_cas[0] == "Decimal" and conv_cas[1] == "Binaire"):
            print(f"Decimal {valeur} ==> '{dec_2_bin(int(valeur))}' en Binaire")
        elif(conv_cas[0] == "Decimal" and conv_cas[1] == "Hexadecimal"):
            print(f"Decimal {valeur} ==> '{base_16(dec_2_bin(int(valeur)))}' en Hexadecimal")
        elif(conv_cas[0] == "Hexadecimal" and conv_cas[1] == "Binaire"):
            print(f"Hexadecimal {valeur} ==> '{hex_2_bin(valeur)}' en Binaire")
        elif(conv_cas[0] == "Hexadecimal" and conv_cas[1] == "Decimal"):
            print(f"Hexadecimal {valeur} ==> '{base_10(hex_2_bin(valeur))}' en Decimal")
        
        converter = input("Quitter [q/c]: ")
        if(converter in "Qq"): break
main()
```
