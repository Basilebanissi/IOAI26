### import

Nous avons déjà abordé **import** dans notre premier exemple. Maintenant, tous les secrets d'**import** te seront dévoilés.

Tous les modules marchent de la même manière en Python. Ils sont externes, et doivent être explicitement inclus dans notre programme pour que nous puissons les utiliser. Pour les inclure, il faut utiliser **import**. import est un mot-clé de Python utilisé pour retrouver les modules et leurs `fonctions\méthodes\constantes\sous-modules`.Il existe plusieurs manières d’importer des modules en Python :

- **import** *nom_du_module*: Importe un module appelé nom_du_module. Ses éléments seront accessible uniquement en utilisant la notation point : *nom_du_module.nom_de_fonction()*

- **from** *nom_du_module* **import**  : Importe tout ce qui se trouve dans le module, rendant tous ses éléments accessibles directement dans ton programme, sans utiliser la notation point.*(Attention : cela peut créer des conflits si des objets locaux portent le même nom que des objets importés du module.)*

- **from** *nom_du_module* **import** *sous_module* : Importe uniquement un sous-module spécifique du module, et rien d’autre.

- **from** *nom_du_module* **import** *(sous_module1, sous_module2, sous_module3, ..., sous_moduleN)* : Importe tous les sous-modules listés entre parenthèses.

- **from** *(chemin relatif)* **import** *nom_du_module* :Ne fonctionne qu’avec des modules locaux et utilise un chemin d’accès relatif valide basé sur des points (./, ../).

#### Aliasing

- \[**from** *valeur*\] **import** *valeur* **as** *alias* : Fonctionne de la même manière que les **import**s précédents, mais donne un surnom (*alias*) au module après le mot-clé **as**, ce qui permet de l’utiliser avec ce nom plus court. L’alias rend souvent le code plus lisible.

### Objets et Programmation orientée objet (OOP)

Nous avons déjà évoqué la programmation orientée objet **(OOP) plusieurs fois et travaillé avec des objets et leurs méthodes avant de présenter une explication complète de ses principes. Python est un langage multi-paradigme, ce qui signifie qu’il intègre des éléments de presque tous les styles de programmation, dont l’OOP, l’un des plus répandus depuis la création de C++ (1985) et Java (1991). Les concepts de la programmation orientée objet ont été développés bien plus tôt : le tout premier langage orienté objet, Simula, a été créé en 1961.

In a nutshell (dans une coquille de noix - en un mot), 
l'idée principale de l'OOP est que l'élément fondamental de la programmation est l’**objet** (object), un type de données défini par un schéma appelé classe (**class**). Les valeurs associées à l'objet, appelées variables d'instance (*instance variables*), font partie de l'objet, tout comme les fonctions pouvant être appliquées à cet objet, appelées méthodes (*methods*). Les objets du programme devraient modéliser les objets du monde réel que le programme traite.

Une idée philosophique centrale de l'OOP est que les langages de programmation devraient permettre aux programmeurs de résoudre des problèmes en utilisant une logique proche de l'intelligence humaine, plutôt que de concevoir les solutions uniquement selon le fonctionnement mécanique d'un ordinateur

Les caractéristiques clés de l’OOP dérivent de cette approche :

- Abstraction : Modéliser des entités ou concepts variés ayant des caractéristiques et traitements communs afin de simplifier et d'unifier leur manipulation
- Encapsulation : Regrouper des données et un ensemble de fonctions logiquement liées dans une "capsule fermée".
- Inheritance (héritage) : Créer des objets dérivés à partir d’autres objets en réutilisant les caractéristiques communes.

Il est normal si ces notions d’OOP te semblent un peu complexes au départ. Ce n’est pas grave : nous avons déjà vu qu’il n’est pas difficile d’utiliser des objets et des méthodes. Maintenant, il s’agit de comprendre comment définir tes propres objets à partir des classes. Dans l’exemple ci-dessous, nous allons créer une classe pour représenter des artistes maliens.

```python
# Exemple 8

class ArtisteMalien:
    """ represente artistes maliens en tout forme d'art """
    
    def __init__(self, nom, vivant):
        self.nom = nom
        self.vivant = vivant

    def get_nom(self):
        return self.nom

    def est_vivant(self):
        return self.vivant

artistes = [ArtisteMalien("Salif Keita", True), 
            ArtisteMalien("Ali Farka Touré", False)]

for artiste in artistes:
    if (artiste.est_vivant() == True):
        vivant = "est toujours en vie."
    else:
        vivant = "n'est plus parmi nous."
    print(f"Artiste {artiste.get_nom()} {vivant}")
```

Nous avons vu les variables d' instance, les variables associées à chaque objet. Il y a aussi les variables de classe, des valeurs qui sont communes à tous les objets d’un type d’une classe. L'exemple ci-dessous illustre ce concept. Il y a deux variables définies dans la portée de la classe, mais pas à l'intérieur d'une méthode de la classe, *\_\_espèce\_\_* et *\_\_nom_scientifique\_\_*. Ce sont des variables de classe : elles s'appliquent à la classe et à chaque objet créé à partir de cette classe.  Leurs valeurs peuvent être obtenues en écrivant le nom de la classe ou d’un objet de type Personne, suivi d’un point et du nom de la variable. Si la valeur d'une variable de classe est changée, elle sera changée pour chaque instance de la classe.

```python
# Exemple 9
class Personne:
    """ informations sur les personnes """
    
    éspèce = "primate"
    nom_scientifique = "homo sapien"
    
    def __init__(self, nom, année, sexe):
        self.nom = nom
        self.naissance = année
        self.sexe = sexe

    def get_nom(self):
        return self.nom
    
    def get_sexe(self):
        return self.sexe

personnes = [Personne("Mariam", 2003, 'f'), 
             Personne("Boubacar", 2005, 'm'), 
             Personne("Hawa", 2007, 'f'), 
             Personne("Mami", 2001, 'f')]

for personne in personnes:
    if personne.get_sexe() == 'f':
        print(personne.get_nom(),"est un", 
              personne.éspèce,"femelle (",
              personne.nom_scientifique,")")

print(Personne.éspèce)
print(Personne.nom_scientifique)
```

Parlons maintenant de l’héritage. Comme son nom l'indique, l'héritage signifie qu'une classe peut hériter de ses attributs d'une autre classe. La classe qui hérite est la classe enfant, et la classe dont elle hérite est la classe parent.

Pour un exemple concret, les musiciens, les photographes et les acteurs maliens sont tous des artistes maliens. Donc, il est logique que la classe ArtisteMalien soit la classe parent des classes enfants MusicienMalien, PhotographeMalien et ActeurMalien. Toutes les variables et toutes les méthodes de la classe ArtisteMalien sont aussi les variables et les méthodes de ses classes enfants. Nous les utilisons avec les classes enfants comme si elles leur étaient propres.

```python
# Example 12

from lesson9 import ArtisteMalien

class MusicienMalien(ArtisteMalien):
    pass

class PhotographeMalien(ArtisteMalien):
    pass

class ActeurMalien(ArtisteMalien):
    pass

musicien = MusicienMalien("Bazoumana Sissoko", False, 1)
photographe = PhotographeMalien("Seydou Keita", False, 1)
acteur = ActeurMalien("Ballamoussa Sissoko", False, 1)

print("Musicien : ", musicien.en_string())
print("Photographe : ", photographe.en_string())
print("Acteur : ", acteur.en_string())
```

Ceci est un exemple simple d'héritage. Pour hériter d'une classe, il suffit de mettre le nom de la classe parent dont tu veux hériter comme paramètre de la classe enfant. Dans l'exemple, les classes *MusicienMalien*, *PhotographeMalien* et *ActeurMalien* héritent de la classe *ArtisteMalien*. Les objets créés à partir de ces classes ont tous les attributs de la classe parent. Ainsi, nous pouvons invoquer la méthode *en_string()* sur chaque objet.

Il serait inutile d'hériter de toutes les fonctionnalités de la classe parent sans ajouter de nouvelles fonctionnalités. Tandis qu'un musicien, un photographe, un acteur et un danseur sont tous des artistes, chaque type d'artiste possède ses particularités. L'héritage nous permet d'hériter de la classe parent, mais aussi d'ajouter d'autres attributs qui appartiennent uniquement à la classe enfant. C'est ce que nous avons fait dans l'exemple ci-dessous où nous avons créé une classe *MusicienMalien* qui hérite des attributs de la classe parent, mais ajoute également des attributs spécifiques à la classe.

```python
# Exemple 13

from lesson9 import ArtisteMalien

class MusicienMalien(ArtisteMalien):
    __albums = 0
    __awards = []

    print()

    def changer_albums(self, albums):
        self.__albums = albums

    def ajouter_awards(self, award_name):
        self.__awards.append(award_name)

    def montre_les_awards(self):
        print(f"\n{self.__nom__} AWARDS - ")
        for i,j in enumerate(self.__awards):
            print(f"{i+1}\t{j}")

    def en_string(self):
        res = f"Artiste {self.__nom__} est du {self._nationalite}, \
est de sexe {self._sexe_en_str()}, est {self._vie_en_str()}."
        msg1 = f"L'artiste est musicien, a fait {self.__albums} albums,\
        et a eu {len(self.__awards)} awards"
        return f"{res} {msg1}"

musicien = MusicienMalien("Amy Koita", True, 0)
musicien.changer_albums(3)
musicien.ajouter_awards("Meilleur Griot de L'annee 2001")
musicien.ajouter_awards("Meilleur Griot de L'annee 2005")
print(musicien.en_string())
musicien.montre_les_awards()
```

La classe *MusicienMalien* a ajouté deux class variables (privées, c'est-à-dire, restreint à l'utilisation uniquement dans la classe, indiquée par le préfixe *\_\_*), *\_\_albums* et la liste *\_\_awards* et trois nouvelles méthodes, *changer_albums()*, *ajouter_awards()*, et *montre_les_awards()*. 

Il y a aussi une nouvelle version d'une méthode trouvée dans la classe *ArtisteMalien*, *en_string()*, cette nouvelle version remplacera la version de la classe parent pour tous objets de type *MusicienMalien*. Pourtant, la nouvelle *en_string()* utilise deux variables héritées de la classe parent, *_sexe_en_str()* et *_vie_en_str()*. Pour rendre le code plus compréhensible, voici un petit dictionnaire des utilitaires utilisés :

- *os.name* - nom du système d'exploitation
- *os.path.exists* - booléen, True si le chemin du fichier existe 

- *os.environ[]* - est un type de dictionnaire qui fonctionne à la fois sur nt/posix mais avec des clés différentes, USERNAME est valide sur nt, mais c'-est USER sur posix

- *os.getuid()* -une méthode disponible uniquement sur posix qui retourne l'ID de l'utilisateur actuel

- *os.system()* - exécute l'argument comme une commande dans l'environnent de ligne de commande du système d'exploitation

```python
# Exemple 15

import os

mon_os = 1 if os.name == "posix" else 0

def __display_header():
    print(f"{'id':15}{'Utilisateur':15}{'Sys. Exploit':15}\
    {'Root':15}\t{'Dossier de Travail':40}".upper())

def __linux_message():
    __display_header()
    print("{:<15}{:15}{:15}{:15}\t{:40}".format(
            os.getuid(), os.environ['USER'], os.name, 
        os.environ['HOME'], os.getcwd()))

def __windows_message():
    __display_header()
    print("{:<15}{:15}{:15}{:15}\t{:40}".format(
            "N/A", os.environ['USERNAME'], 
        os.environ['OS'], os.environ['SYSTEMDRIVE'], os.getcwd()))

command = "echo \"Je suis un Hacker\" > coolfile.txt"

def main():
    print("Mon systeme: ")
    if(mon_os):
        __linux_message()
    else:
        __windows_message()

    print()


    print("Entrain d'ecrire sur un fichier sur le disque...")
    os.system(command)
    fi_path = os.getcwd() + ("/coolfile.txt" if mon_os else r"\coolfile.txt")
    
    fi_write = os.path.exists(fi_path)

    print("Fichiers dans cwd: ", os.listdir())

    if(fi_write):
        print("Tu as sauvegarder un fichier dans ton cwd")
    return

main()
```

```python
# Exemple 17

import sys

print(sys.version)
```

Ici, nous utilisons le module sys pour afficher la version de notre interpréteur Python. 

Le numéro avant les parenthèses est le numéro de version au format X.Y.Z.
- X est le numéro de version majeure,
- Y est le numéro de version mineure,
- Z est le numéro de version micro. 

Connaître le numéro de version de Python en cours d'exécution est pertinent, car il y a des problèmes de compatibilité entre les différentes versions de Python. En fait, nous avons sys.version_info qui donne des données de type tuple contenant des informations sur la version de Python.

```python
# Exemple 18

import sys

ver = sys.version_info

version = float("{}.{}".format(ver.major, ver.minor))

if( version < 2.7 or (version > 2.7 and version < 3.0)):
    sys.stderr.write("ATTENTION: Tu utilises la version {},\
    tu dois effectuer une mise à niveau de Python!".format(
        version))
else:
    sys.stdout.write("SUCCES: Tu as la version {} de Python.\n".format(
        version))

if(sys.platform.startswith('win')):
    sys.stdout.write("Tu dois etre sous windows, les utilisateurs\
    de windows ne sont pas la bienvenue ici!\n")

elif(sys.platform == "linux" or sys.platform == "linux2"):
    sys.stdout.write("Well done! Un utilisateur de Linux, tu es un geant.\n")

else:
    sys.stdout.write("Bienvenue, tout le monde sauf les utilisateurs de Windows.\n")

sys.stdout.write("Mon interpreteur Python se trouve dans: {}\n".format(
    sys.prefix))

sys.exit(0)
```

Le module *sys* nous permet d'écrire des programmes pour différents systèmes. Dans cet exemple, nous montrons comment avec *sys* nous pouvons générer des sorties complètement différentes par système (os et interpréteur python), comme le montrent les images ci-dessous. Les deux sont équivalents :

```python
# Exemple 21

""" CERCLE UNITAIRE - Identités trigonométiques: Tableau 45° intervalle """

import math

angles = [
    0, math.pi/4, math.pi/2, 3*math.pi/4,  math.pi, 5*math.pi/4,
    3*math.pi/2, 7*math.pi/4, 2*math.pi]

angles_verbose = [
    0, 'pi/4', 'pi/2', '3pi/2', 'pi', '5pi/4', '3pi/2', '7pi/4', '2pi']

def ncleanup(value):
    val = str(value)
    if('e' in val):
        zs = val.split('e')[1]
        if(int(zs) > 0): return math.nan
        return 0.0
    return value

def reciprocal(v):
    return ncleanup(1/v) if (v != 0) else math.nan

def main():
    th = "{:>12}{:>12}{:>12}{:>12}{:>12}{:>12}{:>12}".format(
        'Angles(A)', 'sin A', 'cos A', 'tan A', 
        'csc A', 'sec A', 'cot A')

    print(th)

    for i, v in enumerate(angles):
        clns = [ncleanup(j) for j in [math.sin(v), math.cos(v)]]
        sinx, cosx = clns
        rsinx, rcosx = [
            reciprocal(sinx), reciprocal(cosx)]
        
        tr = "{:>12}{:>12.3}{:>12.3}{:>12.3}{:>12.3}{:>12.3}{:>12.3}".format(
            angles_verbose[i],
            sinx, cosx, math.nan if cosx == 0 else sinx/cosx,
            rsinx, rcosx, math.nan if sinx == 0 else cosx/sinx)
        print(tr)

main()
```

Si nous voulons prédire une mutation génétique, le mouvement d’une molécule dans un gaz ou même le résultat d’une élection, il faut prendre en compte une part d’aléa. Lorsque nous écrivons des programmes pour simuler le monde réel ou le comportement humain, nous avons besoin de fonctions permettant d’introduire cet aléa. Le module *random*, pour ton bonheur, possède une liste \_\_all\_\_, ce qui facilite la découverte de ses fonctions avec *random.\_\_all\_\_*. Jetons un coup d’œil à quelques éléments.

```python
# Exemple 23

import random

print(random.__all__, end="\n\n")

def generer_nombres(limite: int=10) -> list:
    return [round(random.random(), 5) for _ in range(limite)]

print(generer_nombres())
```

Puis l’opérateur '=' indique la valeur par défaut (10), c’est-à-dire la valeur utilisée si la fonction est appelée sans cet argument.

Il y a aussi l’annotation de retour *-> list*, qui indique que la fonction renvoie une liste. Cette annotation n’est pas nouvelle pour toi.

Ces fonctionnalités ont été ajoutées à Python à partir de la version 3.5.

Le prochain exemple montre l’utilisation de méthodes utiles du module random :

- *random.randint (a, b)* : Cette méthode renvoie un entier aléatoire x; où a <= x <= b.

###### re

La bombe nucléaire des modules Python utilisés pour le traitement de texte est *re*, les expressions régulières (regex). Les expressions régulières sont, en effet, un langage à part, intégré en Python par son module. Presque tous les langages de programmation auront soit un module similaire à *re*, soit les éléments des expressions régulières intégrées directement dans le langage de base (Perl est un exemple notable). Regex est intégré dans beaucoup d'editeurs de texte dont Word, Notepad++, Atom, et ... Jupyter ! Il y a aussi les outils en ligne et les autres téléchargeables uniquement pour le développement des expressions régulières, p.e., https://regexr.com/ https://pythex.org/

Les expressions régulières sont au cœur de notre sujet préféré, l'informatique théorétique. Ils sont construits autour des automates finis, liés intimement à la théorie de calcul. Il nous faudrait trois ou quatre mois pour expliquer tout cela en détail, mais, encore une fois, we are only going to scratch the surface. La chose à retenir est que, en raison de leur fondement théorique (c'est ce que *régulière* veut dire), les expressions régulières sont super-efficaces et permettent de trouver des motifs dans d’énormes quantités de texte, des opérations qui seraient impossibles sans elles.

Pour plonger directement dans le vif du sujet, voici quelques éléments de base des expressions régulières :

- **.** : Cela signifie que le motif peut correspondre à n'importe quel caractère.
    - Exemple : *a.b* acceptera abc, a7c, aɛc, aùc
- **^** : Le symbole pour la mise en correspondance du début d'une ligne.
    - Exemple : *^Python* acceptera Python uniquement si le mot est au début d'une ligne
- **\$** : Le symbole de correspondance de la fin d'une ligne.
    - Exemple : *a\$* acceptera uniquement la dernière 'a' à la fin de ligne pour le mot 'abracadabra'
- **\[seq\]** : correspond à n'importe quel caractère dans une séquence (seq) de caractères (ordre non important).
    - Exemple : *\[Ll\]\[ae\]* acceptera La, Le, la, ou le
- **\[^seq\]** : correspond à tous les caractères sauf ceux de seq.
    - Exemple : *s\[^aeiou\]c* n'acceptera pas sac, sec, suc mais acceptera sbc, slc, sNc
- **\-** : correspond à un intervalle alphabétique ou numérique 
    - Exemple : *\[a-z\]\[0-9\]\[A-Z\] acceptera a9Z, g3R, mais n'acceptera pas A9Z, gg3, t9b
- **|**: l'operateur *ou* (or), ou la séquence sur la gauche ou la séquence sur la droite peut être valable
    - Exemple : *chat|chien* acceptera chat ou chien
- **(seq)** : groupe la séquence entre parenthèses (permet l'application des opérateurs de répétition là-dessus)
    - Exemple : *(chat){2}* acceptera chatchat

La répétition est effectuée en utilisant les éléments suivants:

- **\*** : (wildcard, signe générique ou joker) - correspond au motif zéro ou plus
    - Exemple : _ma\*rre_ acceptera mrre, marre, maarre, maaarre, maaaarre
- **\?** : correspond à un motif zéro ou une fois
    - Exemple : *ma\?rre* acceptera mrre, marre, mais n'acceptera pas maarre, maaarre, etc.
- **\+** : correspond à un motif une fois ou plus
    - Exemple : *ma\+rre* acceptera marre, maarre, maaarre, maaaarre, mais n'acceptera pas mrre
- \**{s\}** : correspond exactement à un nombre s de fois du modèle.
    - Exemple : *ma\{3\}rre* acceptera uniquement maaarre
    
L'Exemple 27 utilise la fonction *re.compile()*. Les expressions régulières, étant un langage, doivent être compilées avant leur exécution comme tout langage de programmation, c'est l'action de *re.compile()*. 

*re.compile()* produit un objet *pattern* (motif). L'exemple utilise le méthode *match()* (correspondre) de l'objet *pattern*. Si *match()* trouve une correspondance, il renvoie un *match* objet. Finalement, l'exemple utilise la méthode *group()* (groupe) de l'objet *match* pour renvoyer soit une chaîne de caractères, soit un tuple si plusieurs correspondances sont trouvées.

Le prochain (et dernier !) exemple est un véritable tour de force de *pyplot*, montrant sa capacité à tracer des équations variées.

```python
# Exemple 36

import matplotlib.pyplot as plt
import numpy as np
import cmath

# f(x) = 1/x + 1
# f(x) = sqrt(x) + 2
# f(x) = x^2
# f(x) = ]x|

domains = np.arange(-4, 5)

f1 = lambda x: 1/(x*2)
f2 = lambda x: cmath.sqrt(x) + 1
f3 = lambda x: pow(x, 2) + 2
f4 = lambda x: abs(x) + 3

# Fonction de cosinus
cosx = np.linspace(-4, 4, 50)
cosy = np.cos(cosx) + 5

ranges = lambda f: [f(x) for x in domains]

plt.plot(domains, ranges(f1), linestyle="-.", 
         marker='.', color="r", label="f(x) = 1/x")
plt.plot(domains, ranges(f2), linestyle="--", 
         marker='*', color="g", label="sqrt(x) + 2")
plt.plot(domains, ranges(f3), linestyle=":", 
         marker='+', color="m", label="x^2 + 3")
plt.plot(domains, ranges(f4), linestyle="-", 
         marker='|', color="y", label="|x| + 4")
plt.plot(cosx, cosy, linestyle="-.", 
         marker=',', color="k", label="cos(x) + 5")

plt.title("Graphiques de diverses fonctions")

plt.legend()

plt.xlabel("Axe-X")
plt.ylabel("Axe-Y")

plt.show()
```
