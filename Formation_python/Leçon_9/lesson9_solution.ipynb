### import

Nous avons déjà touché sur **import** dans notre premier exemple. Maintenant, tous les secrets d'**import** te seront dévoilés. 

Tous les modules marchent de la même manière en Python. Ils sont externes, et doivent être explicitement inclus dans notre programme pour que nous soyons capable de les utiliser. Pour les inclure, il faut utiliser **import**. import est un mot-clé de Python utilisé pour retrouver les modules et leurs fonctions\méthodes\constantes\sous-modules. Il existe beaucoup de manières pour importer des modules dans Python. Les voici  :

- **import** *nom_du_module*: Importe un module appelé nom_du_module. Ses éléments seront accessible uniquement en utilisant la notation point : *nom_du_module.nom_de_fonction()*

- **from** *nom_du_module* **import** * : Importe tout ce qui se trouve dans le module, en rendant accessible tous ses utilitaires dans ton programme, sans avoir besoin d'utiliser la notation point (Attention, cela peut créer des conflits entre les objets locaux et les objets importés du module portant les mêmes noms).

- **from** *nom_du_module* **import** *sous_module* : Importe spécifiquement un sous-module d'un module, et rien d'autre.

- **from** *nom_du_module* **import** *(sous_module1, sous_module2, sous_module3, ..., sous_moduleN)* : Importe tous les sous-modules se trouvant dans les parenthèses.

- **from** *(chemin relatif)* **import** *nom_du_module* : Ne fonctionne qu'avec des modules locaux, doit être un style de chemin d'accès relatif basé sur des points valide (./, ../)

#### Aliasing
- \[**from** *valeur*\] **import** *valeur* **as** *alias* : Fonctionne de la même manière que les **import**s précédents à part qu'il donne un surnom (*alias*) au module après le mot-clé **as** pour qu'il soit accessible en utilisant *alias* à la place du nom du module. Souvent l'alias est un nom plus court afin de rendre le code plus lisable.

### Objets et Programmation orientée objet

Nous avons fait allusion à la fameuse programmation orientée objet (OOP) déjà plusieurs fois, et nous avons même travaillé avec les objets et leurs méthodes avant de donner une explication complète de ses principes. Python est un langage multi-paradigme, c'est-à-dire que Python comprend des éléments de presque tous les styles de programmation dont l'OOP est, peut-être, la plus répandue depuis la création de C++ (1985) et Java (1991). Les concepts de la programmation orientée objet ont été élaboré bien plus tôt, la toute première langue orientée objet, Simula, a été créé en 1961.

In a nutshell (dans une coquille de noix - en un mot), l'idée principale d'OOP est que l'élément primaire de la programmation est l'*object*, l'objet étant un type de données définis par un schéma qui s'appelle un *class*. Les valeurs associées avec l'objet, les *instance variables*, font partie de l'objet, ainsi que les fonctions qui peut être appliquer pour opérer sur l'objet, appelé, *methods*. Les objets du programme devraient modèlent les objets dans le monde réel traités par le programme. Une idée philosophique centrale dans l'OOP est que les langues de programmation devraient permettre les programmeurs de résoudre les problèmes utilisant un logique qui s'approche l'intelligence humaine plutôt que de concevoir les solutions sur la base du fonctionnement mécanique d'un ordinateur. Des caractéristiques clés de l'OOP qui se dérivent de cette approche sont :

- Abstraction : Modéliser des entités ou concepts variés ayant des caractéristiques et traitements communs afin de simplifier et d'unifier leur manipulation
- Encapsulation : regrouper des données avec un ensemble de fonctions logiquement liées dans une "capsule fermé"
- Inheritance (héritage) : Créer des objets dérivés à partir d'autres objets avec les caractéristiques communes

C'est normale si tu es déjà un peu perdu avec ces notions de l'OOP. C'est pas grave, nous avons déjà vu qu'il n'est pas difficile d'utiliser les objets et les méthodes. Maintenant, il nous faut juste d'essayer de comprendre comment tu peux définir tes propres objets à partir des classes. Dans l'exemple, au-dessous, nous allons créer une classe pour définir les objets qui représenteront les artistes maliens.

```python
# Exemple 8

class ArtisteMalien:
    """ represente artistes maliens en tout forme d'art """
    
    def __init__(self, nom, vivant):
        self.nom = nom
        self.vivant = vivant

    def get_nom(self):
        return self.nom

    def est_vivant(self):
        return self.vivant

artistes = [ArtisteMalien("Salif Keita", True), ArtisteMalien("Ali Farka Touré", False)]

for artiste in artistes:
    if (artiste.est_vivant() == True):
        vivant = "est toujours en vie."
    else:
        vivant = "n'est plus parmi nous."
    print(f"Artiste {artiste.get_nom()} {vivant}")
```

Voila, nous avons créé notre première classe et nous les avons utilisé de créer des objets !

Nous avons vu les instances variables, les variables associées à chaque objet. Il y a aussi les class variables, les valeurs que sont communes à tous les objets d'un type d'une classe. L'exemple, au-dessous, illustre le concept. Il y a deux variables définies dans la portée de la classe, mais pas à l'intérieur d'une méthode de la classe, *\_\_éspèce\_\_* et *\_\_nom_scientifique\_\_*. Elles sont les variables de classe, elles s'appliquent à la classe et elles s'appliquent à chaque objet crée à partir de la classe. Leurs valeurs sont accèdes en apposant un point et le nom de la variable au nom de la classe ou à une instance d'un objet de type *Personne*. Si la valeur d'une variable de classe est changée, elle sera changée pour chaque instance de la classe.

Un objet de type *Personne*, dans l'exemple, est doté d'attributs individuels dont un nom, une année de naissance, et un sexe. Ces attributs appartiennent à l'instance objet, quand ses valeurs sont fixées ou changées pour un objet, il n'y a aucun effet sur les autres objets.

Parlons maintenant de l'héritage. Comme son nom l'indique, l'héritage signifie qu'une classe peut hériter ses attributs d'une autre classe. La classe qui hérite est la classe enfant, et la classe qui la classe enfant a hérité ses attributs est la classe parent.

Pour un exemple concret, les musiciens, les photographes, et les acteurs maliens sont tous les artistes maliens. Donc, il est logique est la classe ArtisteMalien est la classe parent des classes enfants MusicienMalien, PhotographeMalien, et ActeurMalien. Toutes les variables et toutes les méthodes de la classe ArtisteMalien sont, aussi, les variables et les méthodes de ses classes enfants. Nous les utilisons avec les classes enfants comme si elles sont propres à elles.

```python
# Example 12

from lesson9 import ArtisteMalien

class MusicienMalien(ArtisteMalien):
    pass

class PhotographeMalien(ArtisteMalien):
    pass

class ActeurMalien(ArtisteMalien):
    pass

musicien = MusicienMalien("Bazoumana Sissoko", False, 1)
photographeur = PhotographeMalien("Seydou Keita", False, 1)
acteur = ActeurMalien("Ballamoussa Sissoko", False, 1)

print("Musicien : ", musicien.en_string())
print("Photographe : ", photographe.en_string())
print("Acteur : ", acteur.en_string())
```

Ceci est un exemple simple d'héritage. Pour hériter d'une classe, tu mets simplement le nom de la classe parent dont tu veux hériter comme nom du paramètre de la classe enfant. Dans l'exemple, les classes MusicienMalien, PhotographeMalien, et ActeurMaliennous ont hérité de la classe ArtisteMalien. Les objets créés à partir de ces classes ont tous les attributs de la classe parent. Ainsi, nous pouvons invoquer la méthode *en_string* sur chaque objet.

Il serait inutile d'hériter de toutes les fonctionnalités de la classe parente, sans ajouter de nouvelles fonctionnalités. Tandis qu'un musicien, un photographe, un acteur, et un danseur sont tous les artistes, chaque type d'artiste a ses particularités propres à lui. L'héritage nous permet d'hériter de la classe parent, mais aussi d'ajouter d'autres attributs qui appartiennent uniquement à la classe enfant. C'est ce que nous avons fait dans l'exemple au-dessous où nous avons créé une classe MusicienMalien qui hérite des attributs de la classe parent, mais également ajouté des attributs spécifiques à la classe. 

```python
# Exemple 13

from lesson9 import ArtisteMalien

class MusicienMalien(ArtisteMalien):
    __albums = 0
    __awards = []

    print()

    def changer_albums(self, albums):
        self.__albums = albums

    def ajouter_awards(self, award_name):
        self.__awards.append(award_name)

    def montre_les_awards(self):
        print(f"\n{self.__nom__} AWARDS - ")
        for i,j in enumerate(self.__awards):
            print(f"{i+1}\t{j}")

    def en_string(self):
        res = f"Artiste {self.__nom__} est du {self._nationalite}, \
est de sexe {self._sexe_en_str()}, est {self._vie_en_str()}."
        msg1 = f"L'artiste est musicien, a fait {self.__albums} albums, et a eu {len(self.__awards)} awards"
        return f"{res} {msg1}"

musicien = MusicienMalien("Amy Koita", True, 0)
musicien.changer_albums(3)
musicien.ajouter_awards("Meilleur Griot de L'annee 2001")
musicien.ajouter_awards("Meilleur Griot de L'annee 2005")
print(musicien.en_string())
musicien.montre_les_awards()
```

La classe *MusicienMalien* a ajouté deux class variables (privées, c'est-à-dire, restreint à l'utilisation uniquement dans la classe, indiquée par le préfixe *\_\_*), *\_\_albums* et la liste *\_\_awards* et trois nouvelles méthodes, *changer_albums()*, *ajouter_awards()*, et *montre_les_awards()*. Il y a aussi une nouvelle version d'une méthode trouvée dans la classe *ArtisteMalien*, *en_string()*, cette nouvelle version remplacera la version dans la classe parent pour tous objets de type *MusicienMalien*. Pourtant, la nouvelle *en_string()* emploie deux variables héritées de la classe parent, *_sexe_en_str()* et *_vie_en_str()*. Pour rendre le code un peu compréhensible à toi, voici une petite dictionnaire des utilitaires utilisés :

- *os.name* - nom du système d'exploitation
- *os.path.exists* - booléen, True si le chemin de fichier existe

```python
# Exemple 15

import os

mon_os = 1 if os.name == "posix" else 0

def __display_header():
    print(f"{'id':15}{'Utilisateur':15}{'Sys. Exploit':15}{'Root':15}\t{'Dossier de Travail':40}".upper())

def __linux_message():
    __display_header()
    print("{:<15}{:15}{:15}{:15}\t{:40}".format(
            os.getuid(), os.environ['USER'], os.name, os.environ['HOME'], os.getcwd()))

def __windows_message():
    __display_header()
    print("{:<15}{:15}{:15}{:15}\t{:40}".format(
            "N/A", os.environ['USERNAME'], os.environ['OS'], os.environ['SYSTEMDRIVE'], os.getcwd()))

command = "echo \"Je suis un Hacker\" > coolfile.txt"

def main():
    print("Mon systeme: ")
    if(mon_os):
        __linux_message()
    else:
        __windows_message()

    print()


    print("Entrain d'ecrire sur un fichier sur le disque...")
    os.system(command)
    fi_path = os.getcwd() + ("/coolfile.txt" if mon_os else r"\coolfile.txt")
    
    fi_write = os.path.exists(fi_path)

    print("Fichiers dans cwd: ", os.listdir())

    if(fi_write):
        print("Tu as sauvegarder un fichier dans ton cwd")
    return

main()
```

```python
# Exemple 17

import sys

print(sys.version)
```

Ici, nous utilisons le module sys pour afficher la version de notre interpréteur Python. Le numéro avant les parenthèses est le numéro de version au format X.Y.Z. X est le numéro de version majeure, Y est le numéro de version mineure et Z est le numéro de version micro. Connaître le numéro de version du Python en cours d'exécution est pertinent, car il y a des problèmes de compatibilité entre les différentes versions de Python. En fait, nous avons sys.version_info qui donne des données de type tuple contenant des informations sur la version de Python.

```python
# Exemple 18

import sys

ver = sys.version_info

version = float("{}.{}".format(ver.major, ver.minor))

if( version < 2.7 or (version > 2.7 and version < 3.0)):
    sys.stderr.write("ATTENTION: Tu utilises la version {}, tu dois effectuer une mise à niveau de Python!".format(version))
else:
    sys.stdout.write("SUCCES: Tu as la version {} de Python.\n".format(version))

if(sys.platform.startswith('win')):
    sys.stdout.write("Tu dois etre sous windows, les utilisateurs de windows ne sont pas la bienvenue ici!\n")

elif(sys.platform == "linux" or sys.platform == "linux2"):
    sys.stdout.write("Well done! Un utilisateur de Linux, tu es un geant.\n")

else:
    sys.stdout.write("Bienvenue, tout le monde sauf les utilisateurs de Windows.\n")

sys.stdout.write("Mon interpreteur Python se trouve dans: {}\n".format(sys.prefix))

sys.exit(0)
```

Le module *sys* nous permet d'écrire des programmes pour différents systèmes.
```python
# Exemple 21

""" CERCLE UNITAIRE - Identités trigonométiques: Tableau 45° intervalle """

import math

angles = [
    0, math.pi/4, math.pi/2, 3*math.pi/4,  math.pi, 5*math.pi/4,
    3*math.pi/2, 7*math.pi/4, 2*math.pi]

angles_verbose = [
    0, 'pi/4', 'pi/2', '3pi/2', 'pi', '5pi/4', '3pi/2', '7pi/4', '2pi']

def ncleanup(value):
    val = str(value)
    if('e' in val):
        zs = val.split('e')[1]
        if(int(zs) > 0): return math.nan
        return 0.0
    return value

def reciprocal(v):
    return ncleanup(1/v) if (v != 0) else math.nan

def main():
    th = "{:>12}{:>12}{:>12}{:>12}{:>12}{:>12}{:>12}".format(
        'Angles(A)', 'sin A', 'cos A', 'tan A', 
        'csc A', 'sec A', 'cot A')

    print(th)

    for i, v in enumerate(angles):
        clns = [ncleanup(j) for j in [math.sin(v), math.cos(v)]]
        sinx, cosx = clns
        rsinx, rcosx = [
            reciprocal(sinx), reciprocal(cosx)]
        
        tr = "{:>12}{:>12.3}{:>12.3}{:>12.3}{:>12.3}{:>12.3}{:>12.3}".format(
            angles_verbose[i],
            sinx, cosx, math.nan if cosx == 0 else sinx/cosx,
            rsinx, rcosx, math.nan if sinx == 0 else cosx/sinx)
        print(tr)

main()
```

La bombe nucléaire de modules Python utilisés pour le traitement de texte est *re*, les expressions régulières (regex). Les expressions régulières sont, en effet, un langage à part, intégré en Python par son module. Presque tous les langages de programmation auront soit un module pareille à *re* soit les éléments des expressions régulières intégrées directement dans le langage de base (Perl est un exemple notable). Regex est intégré dans beaucoup d'editeurs de texte dont Word, Notepad++, Atom, et ... Jupyter ! Il y a aussi les outils en ligne et les autres téléchargeables uniquement pour le développement des expressions régulières, p.e., https://regexr.com/ https://pythex.org/

Les expressions régulières sont au cœur de notre sujet préféré, l'informatique théorétique. Ils sont construits autour de les automates finis, lié intimement avec la théorie de calcul. Nous voudrions avoir trois ou quatre mois à expliquer tout cela à toi, mais, encore, we are only going to scratch the surface. La chose à retenir est que, en raison de leur fondement théorique (c'est ce que *régulière* veut dire), les expressions régulières sont super-efficaces et il est possible de trouver les motifs en les quantités de texte énorme, les opérations qui seraient impossible sans les expressions régulières.

Pour plonger directement dans le vif de la matière, voici quelques-unes des éléments de base des expressions régulières :

- . : Cela signifie que le motif peut correspondre à n'importe quel caractère.
    - Exemple : *a.b* acceptera abc, a7c, aɛc, aùc
- ^ : Le symbole pour la mise en correspondance du début d'une ligne.
    - Exemple : *^Python* acceptera Python uniquement si le mot est au début d'une ligne
- \$ : Le symbole de correspondance de la fin d'une ligne.
    - Exemple : *a\$* acceptera uniquement la dernière 'a' à la fin de ligne pour le mot 'abracadabra'
- \[seq\] : correspond à n'importe quel caractère dans une séquence (seq) de caractères, n'est pas ordonné.
    - Exemple : *\[Ll\]\[ae\]* acceptera La, Le, la, ou le
- \[^seq\] : correspond à tous les caractères sauf ceux de seq.
    - Exemple : *s\[^aeiou\]c* n'acceptera pas sac, sec, suc mais acceptera sbc, slc, sNc
- \- : correspond à un intervalle alphabétique ou numérique 
    - Exemple : *\[a-z\]\[0-9\]\[A-Z\] acceptera a9Z, g3R, mais n'acceptera pas A9Z, gg3, t9b
- | : l'operateur *ou* (or), ou la séquence sur la gauche ou la séquence sur la droite peut être valable
    - Exemple : *chat|chien* acceptera chat ou chien
- (seq) : groupe la séquence entre parethéses (permet l'application des opérateurs de répétition au-dessous)
    - Exemple : *(chat){2}* acceptera chatchat

La répétition est effectuée en utilisant les éléments suivants:

- \* : (wildcard, signe générique ou joker) - correspond au motif zéro ou plus
    - Exemple : _ma\*rre_ acceptera mrre, marre, maarre, maaarre, maaaarre
- \? : correspond à un motif zéro ou une fois
    - Exemple : *ma\?rre* acceptera mrre, marre, mais n'acceptera pas maarre, maaarre, etc.
- \+ : correspond à un motif une fois ou plus
    - Exemple : *ma\+rre* acceptera marre, maarre, maaarre, maaaarre, mais n'acceptera pas mrre
- \{s\} : correspond exactement à un nombre s de fois du modèle.
    - Exemple : *ma\{3\}rre* acceptera uniquement maaarre
    
L'Exemple 27 utilise la fonction *re.compile()*. Les expressions régulières, étant un langage, doivent être compilées avant leur exécution comme tout langage de programmation, c'est l'action de *re.compile()*. *re.compile()* produit un objet *pattern* (motif). L'exemple utilise le méthode *match()* (correspondre) de l'objet *pattern*. Si *match()* trouve une correspondance, il renvoie un *match* objet. Finalement, l'exemple utilise le méthode *group()* (groupe) de l'objet *match* pour renvoyer soit une chaîne de caractères soit un tuple pour les résultats multiples pour les correspondances.

```python
# Exemple 36

import matplotlib.pyplot as plt
import numpy as np
import cmath

# f(x) = 1/x + 1
# f(x) = sqrt(x) + 2
# f(x) = x^2
# f(x) = ]x|

domains = np.arange(-4, 5)

f1 = lambda x: 1/(x*2)
f2 = lambda x: cmath.sqrt(x) + 1
f3 = lambda x: pow(x, 2) + 2
f4 = lambda x: abs(x) + 3

# Fonction de cosinus
cosx = np.linspace(-4, 4, 50)
cosy = np.cos(cosx) + 5

ranges = lambda f: [f(x) for x in domains]

plt.plot(domains, ranges(f1), linestyle="-.", marker='.', color="r", label="f(x) = 1/x")
plt.plot(domains, ranges(f2), linestyle="--", marker='*', color="g", label="sqrt(x) + 2")
plt.plot(domains, ranges(f3), linestyle=":", marker='+', color="m", label="x^2 + 3")
plt.plot(domains, ranges(f4), linestyle="-", marker='|', color="y", label="|x| + 4")
plt.plot(cosx, cosy, linestyle="-.", marker=',', color="k", label="cos(x) + 5")

plt.title("Graphiques de diverses fonctions")

plt.legend()

plt.xlabel("Axe-X")
plt.ylabel("Axe-Y")

plt.show()
```

```python
# Probleme 1
import string
import math
import time
import itertools
import numpy as np

def mots_francais(anags):
    with open("french.txt") as fp:
        data = fp.readlines()
    data = [i.replace("\n", '') for i in data]
    mset = set(data).intersection(anags)
    return list(mset)

def column_display(anags, x=8):
    for i,j in enumerate(anags):
        print(j, end="\t")
        if((i+1)%x == 0):
            print()
    print()

def time_elapsed(t):
    h, r = divmod(t, 3600)
    m, s = divmod(r, 60)
    stformat = f"{int(h):0>2}:{int(m):0>2}:{int(s):0>2}s"
    return stformat

## Methode 1
def anagram1(text):
    return ["".join(i) for i in list(itertools.permutations(list(text)))][:200]

## Methode 2
def nPr(n,r):
    """ Permutation Counter """
    return math.factorial(n) / math.factorial(n-r)

def permutation_indexes(taille, limit=200):
    """ P-X """
    indexes = list(range(int(taille)))
    perms = nPr(taille, taille)
    compute_range = perms if perms <= limit else limit
    res = []
    for i in range(int(compute_range)):
        idx = np.random.permutation(indexes)
        while(list(idx) in res):
            idx = np.random.permutation(indexes) # Not efficient
        res.append(list(idx))
    return res

def anagram2(chaine):
    chaine = chaine.lower()
    indexes = permutation_indexes(len(chaine))
    chaines = []
    for i in indexes:
        new_chaine = ""
        for j in i:
            new_chaine += chaine[j]
        chaines.append(new_chaine)
    return chaines

## Main
def main():
    chaine = input("Ta chaine: ")

    start_t = time.time()
    chaineperms = anagram2(chaine)
    end_t = time.time()
    ecouler = time_elapsed(end_t-start_t)
    tperms = int(nPr(len(chaine), len(chaine)))

    print(f"Nombre de permutation de '{chaine}': ", tperms)

    if(tperms > 200):
        print(f"200 permutations de {chaine}: ")
        column_display(chaineperms, 4)
    else:
        print(f"Permutations de {chaine}: ")
        column_display(chaineperms, 4)

    frn = mots_francais(chaineperms)

    if(frn):
        print(f"Anagrames de {chaine}: ")
        column_display(frn, 5)

    print(f"Temps écoulé: {ecouler}")

main()
```

```python
# Probleme 4

import math
import time

class PiConvergence(object):

    def __init__(self, prec=None):
        self.prec = prec

    def compute_arcpi(self, prec=None):
        def trouve_nombre():
            for i in range(10):
                pi_calc = math.atan(1)*i
                if round(pi_calc, 2) == 3.14:
                    return pi_calc
        pprec = (prec if prec else self.prec) + 1
        pi_v = str(trouve_nombre())[:pprec+1]
        return float(pi_v)

    def liebniz_series(self, iter, prec):
        denom = 1.0
        sign = 1.0
        pi_v = 0
        compi = 0
        piprec = self.compute_arcpi(prec)
        for i in range(iter+1):
            pi_v += (4*(1.0 / denom)) * sign
            denom += 2
            sign *= -1
            compi = float(str(pi_v)[:prec+2])
            if(piprec == compi):
                return compi
        return None

    def nilakanta_series(self, iter, prec):
        multiplier = 1.0
        start_denom = 2.0
        pi_v = 3
        compi = 0
        piprec = self.compute_arcpi(prec)
        for i in range(1, iter+1):
            pi_v += ((4.0 / (start_denom * (start_denom+1.0) * (start_denom+2.0))) * multiplier)
            start_denom += 2.0
            multiplier *= -1.0
            compi = float(str(pi_v)[:prec+2])
            if(piprec == compi):
                return compi
        return pi_v

    def pi_profile(self, it, f, *args):
        """ 
            f: function
            it: iteration
            args: arguments of f
        """

        temps = []
        for i in range(it):
            start = time.perf_counter()
            v = f(*args)
            end = time.perf_counter()
            temps.append(end-start)
        return sum(temps)/it

    def tableau_temps(self, it1, it2, it3, prec):
        header = f"Pi. Approx.\t{it1}x\t{it2}x\t{it3}x"

        fn = [self.compute_arcpi, self.liebniz_series, self.nilakanta_series]

        rkeys = ["atan", "liebniz", "nilakantha"]

        pi_data = {}

        for i,j in enumerate(fn):
            for k in [it1, it2, it3]:
                if(i == 0): 
                    z = rkeys[0]
                    args = (prec,)
                if(i == 1): 
                    z = rkeys[1]
                    args = (12000000, prec)
                if(i == 2): 
                    z = rkeys[2]
                    args = (100, prec)
                if(z not in pi_data): pi_data[z] = []
                pi_data[z].append(self.pi_profile(k, j, *args))

        print(header)
        print("atan\t\t{:.2f}\t{:.2f}\t{:.2f}".format(*pi_data["atan"]))
        print("nilakantha\t{:.2f}\t{:.2f}\t{:.2f}".format(*pi_data["nilakantha"]))
        print("liebniz\t\t{:.2f}\t{:.2f}\t{:.2f}".format(*pi_data["liebniz"]))

PIV = PiConvergence()
PIV.tableau_temps(5, 8, 10, 6)
```
