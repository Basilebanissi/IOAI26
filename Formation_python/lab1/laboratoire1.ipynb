## Laboratoire I

Les laboratoires sont des exercices qui simulent le travail réel d'un programmeur. Nous te donnons les spécifications d'un programme et vous devez imaginer son organisation et l'écrire tout entier sans ayant un cadre pour débuter. Vous allez aussi le tester et démontrer que le programme marche correctement.

Il faut prendre soin de bien expliquer ton programme dans les commentaires. S'il n'est pas facile pour ton client (les profs) de comprendre ton programme, vous mériterez une réprimande ! Choisir les bons noms pour tes variables. Il faut inclure les résultats de tes tests et ces résultats doivent obligatoirement montrer que tu as rigoureusement testé ton programme et prouver que le programme marche pour tous les cas.

### Programme 1. Déterminer la "mention" d'un étudiant

Une école décerne des certificats à tous les étudiants qui ont réussi leur programme d'études. Ces certificats seront décernés avec des mentions, chacun avec une description textuelle de la mention. La moyenne de tous les points accumulés dans les cours est calculée et l'élève reçoit une mention en fonction de la fourchette dans laquelle se situe sa moyenne. Les mentions et leurs fourchettes sont les suivantes :

- Maître programmeur : Toute personne ayant une moyenne égale ou supérieure à 100 %.

- Très bien : Toute personne ayant une moyenne égale ou supérieure à 90 % et moins que 100 %.

- Bien : Toute personne ayant une moyenne égale ou supérieure à 80 % et moins que 90 %.

- Assez-bien : Toute personne ayant une moyenne égale ou supérieure à 70 % et moins que 80 %.

- Acceptable : Toute personne ayant une moyenne égale ou supérieure à 65 % et moins que 70 %.

- Pas bon: Toute personne ayant une moyenne égale ou supérieure à 0 et moins que 65 %. (Elles ne recevront pas de certificat)

- Invalide : Pour les moyennes inférieures à 0 ou supérieures à 115%.

En raison du nombre d'étudiants, l'administration est débordée par la tâche de calculer les moyennes. Elle a décidé de t'engager pour écrire un programme qui fera les calculs, malgré tes honoraires extrêmement élevés :

- **Pour cette tâche, il faut que tu écrives un programme qui demande à l'utilisateur de taper le nom du étudiant, et cinq moyennes qu'ils ont reçues. Ensuite, tu dois calculer la moyenne totale en entier de l'étudiant, puis émit le message qui correspond au niveau gagné par l'étudiant.**

Par exemple, si une étudiante a reçu : 80, 85, 80, 90, 100, et elle est appelée M'Badiala Doumbia; cette étudiante aura une moyenne totale de 87 comme nombre entier. Le programme doit alors afficher le message suivant :

>>> M'Badiala Doumbia a reçu la mention 'Bien'.

Si M'Badiala Doumbia a reçu : 50, 40, 0, 60, 80, elle aura alors une moyenne totale de 46 ; cela donnerait ce message :

>>> M'Badiala Doumbia n'a pas une moyenne suffisante pour obtenir un certificat.

### Programme 2.  Elever au carré les nombres qui se terminent en 5

En mathématiques, il existe un raccourci pour élever au carré les numbres qui se terminent en cinq. Calculée à la main, pour les nombres de tailles restreintes, cette méthode peut être plus rapide que la méthode classique de réalisation de multiplications. Ton patron croit que cette méthode peut être aussi plus rapide par ordinateur. Il te demande d'implémenter la méthode et de faire un test pour voir laquelle est la plus rapide.

#### L'algorithme:

##### Exemple1
Disons que nous voulons calculer **105<sup>2</sup>**

- Tout d'abord, nous laissons tomber le 5 final. Nous utiliserons le nombre qui reste dans l'étape suivante. Pour cet exemple, après nous avons enlevé le 5, nous avons le 10 qui reste.

- Nous multiplions ensuite le nombre avant le 5 final par ce même nombre incrémenté par 1. Donc 10 * (10+1), pour l'exemple, donne 110.

- puis nous multiplions le nombre obtenu à l'étape 2 par 100 et additionnons 25 au nombre. 110 * 100 + 25 est 11000 + 25, donne le résultat de 11 025.

###### figure1
105<sup>2</sup>    =>  10\*(10+1)

            =>  10*(11)
            =>  110*100 + 25
Résultat      =>  11 025

On peut faire la troisième étape comme une concaténation entre le résultat du nombre de l'étape 2, converti en chaîne de caractères et la chaîne de caractères "25". Pour l'exemple précédent, cela donnera "110" + "25" donnant "11025". Voici un autre exemple, utilisant cette méthode :

##### Exemple2
Calculons **135<sup>2</sup>**

###### figure2

135<sup>2</sup>    => 13\*(13+1)   
            
            => 13*(14)     
            => "182"    +    "25"
Résultat      => 18225

#### Programme

Ta tâche consiste à écrire un programme capable d'effectuer ce calcul. Le programme :

- Doit demander aux utilisateurs de saisir un nombre qui se termine en cinq. **Si le nombre ne se terimine pas en cinq, il faut revenir à l'utilisateur avec un message d'erreur.**

- Effectuer les étapes de l'algorithme utilisant la première méthode

- Le résultat doit ressembler à ceci

> 135 ** 2 = 18225

- Répeter pour la deuxième méthode
