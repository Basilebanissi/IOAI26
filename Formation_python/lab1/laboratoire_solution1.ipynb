### Programme 1. Déterminer la "mention" d'un étudiant

### Programme 2.  Elever au carré les nombres qui se terminent par 5

En mathématiques, il existe un raccourci pour élever au carré les nombres qui se terminent par cinq. Calcule à la main, pour les nombres de tailles restreintes, cette méthode peut être plus rapide que la méthode classique de réalisation de multiplications. Ton patron croit que cette méthode peut être aussi plus rapide par ordinateur. Il te demande d'implémenter la méthode et de faire un test pour voir laquelle est la plus rapide.

#### L'algorithme:

##### Exemple1
Disons que nous voulons calculer **105<sup>2</sup>**

- Tout d'abord, nous laissons tomber le 5 final. Nous utiliserons le nombre qui reste dans l'étape suivante. Pour cet exemple, après nous avons enlevé le 5, nous avons le 10 qui reste.

- Nous multiplions ensuite le nombre avant le 5 final par ce même nombre incrémenté par 1. Donc 10 * (10+1), pour l'exemple, donne 110.

- puis nous multiplions le nombre obtenu à l'étape 2 par 100 et additionnons 25 au nombre. 110 * 100 + 25 est 11000 + 25, donne le résultat de 11 025.

###### figure1
105<sup>2</sup>    =>  10\*(10+1)

            =>  10*(11)
            =>  110*100 + 25
Résultat      =>  11 025

On peut faire la troisième étape comme une concaténation entre le résultat du nombre de l'étape 2, converti en chaîne de caractères et la chaîne de caractères "25". Pour l'exemple précédent, cela donnera "110" + "25" donnant "11025". Voici un autre exemple, utilisant cette méthode :

##### Exemple2
Calculons **135<sup>2</sup>**

###### figure2

135<sup>2</sup>    => 13\*(13+1)   
            
            => 13*(14)     
            => "182"    +    "25"
Résultat      => 18225

#### Programme

Ta tâche consiste à écrire un programme capable d'effectuer ce calcul. Le programme :

- Doit demander aux utilisateurs de saisir un nombre qui se termine par cinq. **Si le nombre ne se terimine pas par cinq, il faut revenir à l'utilisateur avec un message d'erreur.**

- Effectue les étapes de l'algorithme utilisant la première méthode

- Le résultat doit ressembler à ceci

> 135 ** 2 = 18225

- Répète pour la deuxième méthode

C'est bon pour le moment. Nous allons utiliser ton code plus tard pour tester l'hypothèse de ton patron dans une leçon ultérieure. Au plaisir de nouvelles aventures !

```python
# Methode 1: Sans concatenation
nombre_terminant_par5 = int(input("Entre un entier terminant par 5: "))
print(nombre_terminant_par5)

# Constants
FACTEUR = 5
UNITAIRE = 1
DIZAINE = 10
CENTAINE = 100

# Sortie constant
sortie_constant = " ** 2 = "
erreur_msg = "Cet algorithme ne fonctionne qu'avec des entiers qui se terminent par 5."

# Teste si l'entrée est valide
if(nombre_terminant_par5 % DIZAINE == FACTEUR):
    # Obtien tous les valeur sauf le 5
    avant5 = nombre_terminant_par5 // DIZAINE

    # Joins le 25 final avec la valeur multipliér par son increment
    resulat = (avant5 * (avant5 + UNITAIRE)) * CENTAINE + (FACTEUR * FACTEUR)

    # Imprime le resultat finale
    print(str(nombre_terminant_par5) + sortie_constant + str(resulat))

else:
    # Message d'erreur si le nombre entré ne se termine pas par 5
    print(erreur_msg)
```

```python
# Methode 2: Avec concatenation
nombre_terminant_par5 = int(input("Entre un entier terminant par 5: "))

# Constants
FACTEUR = 5
UNITAIRE = 1
DIXAINE = 10
CENTAINE = 100

# Sortie constant
sortie_constant = " ** 2 = "
erreur_msg = "Cet algorithme ne fonctionne qu'avec des entiers qui se terminent par 5."

# Teste si l'entrée est valide
if(nombre_terminant_par5 % DIXAINE == FACTEUR):
    # Obtiens toutes les valeurs sauf le 5
    avant5 = nombre_terminant_par5 // DIXAINE

    # Joins le 25 final en str format avec la valeur multipliée par son increment
    resulat = str((avant5 * (avant5 + UNITAIRE))) + str(FACTEUR * FACTEUR)

    # Imprime le resultat final
    print(str(nombre_terminant_par5) + sortie_constant + str(resulat))

else:
    # Message d'erreur si le nombre entré ne se termine pas par 5
    print(erreur_msg)
```
