```````python
# Programme 8
import math

def poi(s : int, d : int, N :int) -> float : 
        # s la longueur de la chaîne de caractères cherchée
        # d le nombre de symboles dans l'alphabet (ex, chiffres 10, a-z 26)
        # N le nombre de caractères dans l'espace de recherche
        # renvoie : la probabilité
    '''  
    Cette fonction calcule la probabilité de trouver une chaîne de caractères de longueur s
    dans une chaîne de caractères de longueur N avec une distribution aléatoire où le numéro
    de symboles uniques est d. Par exemple, si l'on cherche "1234" dans une chaîne de
    chiffres (0-9) aléatoire d'une longueur de 1 000 000, s=4, d=10, et N=1000000.
    Le calcul est fait utilisant l'approximation d'une distribution binomiale de Poisson.
    p = 1 - e**((N-s) * (log (d**s - 1) - log (d**s)))
    '''    
    p= 1 - math.pow(math.e, ( (N-s) * (math.log(math.pow(d,s) - 1) - math.log(math.pow(d,s)))))  
    return(p)

def lire_fichier(path_fichier : str) -> str:
    # path_fichier une chaîne de caractères avec le chemin et le nom de fichiers à lire
    # renvoie : une chaîne de caractères avec tout le contenu du fichier
    
    # Cette fonction met tout le contenu du fichier dans une chaîne de caractères
    # Elle enlève les nouvelles lignes (\n) et le symbole pour continuation de lignes (\)
    
    with open(path_fichier, 'r') as monfichier:
        mstr = monfichier.read().replace('\n', '')
    return(mstr.replace('\\',''))
```````

Trouver des nombres premiers jusqu'à  un nombre N peut être un calcul intensif s'il est fait en utilisant la méthode de la force brute. Certaines personnes extrêmement intelligentes ont trouvé des techniques malines pour accélérer le processus de recherche de nombres premiers jusqu'à N. Trois de ces méthodes sont:

##### Crible d'Ératosthène

Cette méthode est l'une des plus anciennes approches pour obtenir des nombres premiers jusqu'à N. *Elle sélectionne la première occurrence d'un nombre comme nombre premier, puis marque tous ses multiples comme composites, elle le fait de manière itérative jusqu'à ce que tous les nombres jusqu'à N soient marqué comme prime ou composite*

```
Disons que tu veux des nombres premiers jusqu'à 10.
tu commences par la liste: [2,3,4,5,6,7,8,9,10], 
nous savons que 0 et 1 ne sont pas des nombres premiers.

Itération1 - Nous analysons la liste pour tous les multiples 
de 2 (disons que nous les supprimons):
Après itération1 - Notre liste sera [2,3,5,7,9], nous avons supprimé 4,6,8

Itération2 - Nous analysons la liste pour tous les multiples de 3
Après itération2 - Notre liste sera [2,3,5,7], nous avons suprimé 6,9

Itération3 - Nous analysons la liste pour tous les multiples de 5
Après itération3 - Notre liste sera [2,3,5,7], 
il n'y a pas un multiple de 5 dans la liste

Itération4 - Nous analysons la liste pour tous les multiples de 7
Après itération4 - Notre liste sera [2,3,5,7], 
il n'y a pas un multiple de 7 dans la liste

Alors notre résultat final sera [2,3,5,7]
```

##### Crible de Sundaram

Un peu de temps après Ératosthène, approximativement 2000 ans, un autre savant, Sundaram, a trouvé une autre méthode pour trouver des valeurs premières jusqu'à N. Le tamis de Sundaram est obtenu avec le pseudocode suivant:

```
L'algorithme produit des nombres premiers plus petits que (2 * x + 2) pour un nombre x donné.
Puisque nous voulons des nombres premiers plus petits que N, notre première étape sera de 
réduire N de moitié.

p = N-1 // 2
nList = [0,1,2,3,...,p]

for i=1 to p
    j = i
    while i+j+2*i*j <= p
        retirer nList[i+j+2*i*j]
        j += 1

premiers[0] = 2

for x=1 to taille(nList)
    premiers[x] = 2*nList[x] + 1

premiers donnera les nombres premiers à N.

Disons que tu veux des nombres premiers jusqu'à 10.

p = 4, nList = [0,1,2,3,4]
Après la première boucle nList est [0,1,2,3], nList[4] ayant été retiré
Dans la deuxième boucle nList[1]=1,nList[2]=2,nList[3]=3 est multipliés par 2, 
incrémentés par 1 et apposer à la liste premiers, donnant [2,3,5,7]
```

Le crible de Sundaram tamise les nombres composites tout comme le crible d'Eratosthène, mais les nombres pairs ne sont pas considérés ; le travail d'éliminer les multiples de 2 est effectué par l'étape finale de doubler le nombre et l'incrémenter par un.

##### La méthode de la Racine Carrée

Cette méthode n'obtient pas une liste des nombres premiers jusqu'à N, mais elle est efficace pour tester si un nombre est premier ou non quand le programme itère à travers une plage d'entiers. Cela fonctionne comme ceci :

```
Étant donné un nombre N
```

> r = $\sqrt{N}$

```
si r est un nombre entier, on sait automatiquement que ce n'est pas un nombre premier.
si r n'est pas un nombre entier, nous testons si N est divisible par tous les nombres 
jusqu'à int(r) (c'est-à-dire, coupe les valeurs après la virgule décimale et 
teste uniquement la partie entière du nombre).
```
Évidemment, le nombre de valeurs à tester jusqu'à la partie entière de la racine carrée sera petit, p.ex., la racine carrée de 847 est 29.103, donc nous testerons juste les facteurs jusqu'à 29.
