## Setup a executer en premier

````python
import ipywidgets as widgets
from IPython.display import display, Markdown

score = {"correct": 0, "total": 0}

def qcm(question, options, correct_index, explication):
    radio = widgets.RadioButtons(
        options=options,
        layout=widgets.Layout(width="100%"),
        style={"description_width": "initial"},
    )
    bouton = widgets.Button(description="Valider", button_style="info")
    zone_resultat = widgets.Output()

    def on_click(b):
        if bouton.disabled:
            return
        bouton.disabled = True
        score["total"] += 1
        with zone_resultat:
            if radio.index == correct_index:
                score["correct"] += 1
                display(Markdown(f"**Correct !** {explication}"))
            else:
                bonne_reponse = options[correct_index]
                display(Markdown(f"**Pas tout a fait.** La bonne reponse etait : *{bonne_reponse}*\n\n{explication}"))
            display(Markdown(f"*Score actuel : {score['correct']}/{score['total']}*"))

    bouton.on_click(on_click)
    display(Markdown(f"**{question}**"))
    display(radio, bouton, zone_resultat)


def reinitialiser_score():
    score["correct"] = 0
    score["total"] = 0
    print("Score reinitialise.")


print("Setup pret. Score actuel :", score)
````
